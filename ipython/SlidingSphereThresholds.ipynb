```python
# Load libraries
%matplotlib inline
import glob,gzip,os,operator
import numpy as np
from numpy.random import randn
from itertools import combinations
import random
import pandas as pd
from scipy import stats
from scipy.spatial import KDTree
from sklearn.metrics import confusion_matrix
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 500)
```

```python
# Load header information
header_fname  = 'scripts/sphere_header.txt'
with open(header_fname,'rb') as hfin:
    header = [h.strip() for h in hfin.readlines()]
```

```python
# Load the results file into a pandas data frame
fnames = glob.glob('/Volumes/sivleyrm/pdbmap/results/sliding_sphere_10/split/obs/bystruct/*.txt.gz')
rlist  = [[r.split('-')[0][-4:],r.split('-')[1].split('.')[0]] for r in fnames]

# Read all structures separately
dfs = [(i,pd.read_table(gzip.open(fname,'rb'),delimiter='\t',skiprows=1,names=header)) for i,fname in enumerate(fnames)]

# Add biological assembly identifiers
for i,df in dfs:
    df['structid'],df['biounit'] = [rlist[i][0] for j in range(len(df))],[rlist[i][1] for j in range(len(df))]

# Filter out problem structures (any structure where an entire columns is NULL)
dfs = [df for i,df in dfs if not np.any(df.isnull().sum()==len(df))]

# Merge all dataframes into a single dataframe
df = pd.concat(dfs)
del dfs

# Drop duplicate residues
df.drop_duplicates(['structid','biounit','x','y','z'],inplace=True)
```

```python
# Observed values
info_cols      = header[:21]       # Residue information
count_cols     = header[20:36]     # SNP counts 
cdaf_cols      = header[36:41]     # Cumulative derived allele frequencies
ddaf_cols      = header[41:69:3]   # Mean directional deltaDAF
ddaf_cols     += header[42:70:3]   # 
absddaf_cols   = header[43:71:3]   # Mean absolute deltaDAF
fst_cols       = header[71:82]     # Fst
w_count_cols   = header[82:98]     # Windowed SNP counts
w_cdaf_cols    = header[98:103]    # Windowed Cumulative derived allele frequencies
w_ddaf_cols    = header[103:131:3] # Windowed mean directional deltaDAF
w_ddaf_cols   += header[104:132:3] # 
w_absddaf_cols = header[105:133:3] # Windowed mean absolute deltaDAF
w_fst_cols     = header[133:144]   # Windowed Fst

# Observed p-values
def pval_cols(cols):
    return ["%s_pval"%col for col in cols]
count_pval_cols     = pval_cols(count_cols)
cdaf_pval_cols      = pval_cols(cdaf_cols)
ddaf_pval_cols      = pval_cols(ddaf_cols)
absddaf_pval_cols   = pval_cols(absddaf_cols)
fst_pval_cols       = pval_cols(fst_cols)
w_count_pval_cols   = pval_cols(w_count_cols)
w_cdaf_pval_cols    = pval_cols(w_cdaf_cols)
w_ddaf_pval_cols    = pval_cols(w_ddaf_cols)
w_absddaf_pval_cols = pval_cols(w_absddaf_cols)
w_fst_pval_cols     = pval_cols(w_fst_cols)

# Nested column sets
value_cols  = [count_cols,cdaf_cols,ddaf_cols,absddaf_cols,fst_cols,w_ddaf_cols,w_absddaf_cols,w_fst_cols]
pvalue_cols = [count_pval_cols,cdaf_pval_cols,ddaf_pval_cols,absddaf_pval_cols,fst_pval_cols,w_ddaf_pval_cols,w_absddaf_pval_cols,w_fst_pval_cols]
```

```python
# Shift negative Fst values to 0
for col in fst_cols + w_fst_cols:
    df.ix[df[col]<0,col] = 0
```

```python
# Correct poorly implemented >1 SNP per sphere restriction
# If there are fewer than 2 shared SNPs in the sphere, all measurements
# should be set to 0, and all p-values to 1.0
for i,scount in enumerate(count_cols[-10:]): # Shared SNP counts (spheres)
    df.ix[df[scount]<2,[c[i] for c in value_cols[2:5]]]  = 0.
    df.ix[df[scount]<2,[c[i] for c in pvalue_cols[2:5]]] = 1.

df.ix[df['snpcount']<2,'allpop_fst']      = 0.
df.ix[df['snpcount']<2,'allpop_fst_pval'] = 1.

for i,wcount in enumerate(w_count_cols[-10:]): # Shared SNP counts (windows)
    df.ix[df[wcount]<2,[c[i] for c in value_cols[5:]]]  = 0.
    df.ix[df[wcount]<2,[c[i] for c in pvalue_cols[5:]]] = 1.

df.ix[df['w_snpcount']<2,'w_allpop_fst']      = 0.
df.ix[df['w_snpcount']<2,'w_allpop_fst_pval'] = 1.
```

```python
# Define threshold analysis
def threshold(df,col,pcol,op=operator.lt,maxval=None):
    # Determine threshold operator
    if type(op) == type(operator.lt):
        pass
    elif op == '<':
        op = operator.lt
    elif op == '>':
        op = operator.gt
    elif op == '<=':
        op = operator.le
    elif op == '>=':
        op = operator.ge
    else:
        raise Exception("Invalid operator")
    # Determine threshold intervals
    minval = df[col].min()
    maxval = df[col].max() if not maxval else maxval
    if col==pcol:
        minval = 1
        maxval = -np.log10(df[col].min())
    if df[col].dtype==np.int:
        step = 1
    else:
        step = (maxval-minval)/100.
    # Test threshold intervals
    threshpass,p = [],0.01
    while minval <= maxval:
        if col==pcol:
            p = minval
            tdf = df[-np.log10(df[pcol])>p]
        else:
            tdf = df[(op(df[col],minval)) & (df[pcol]<p)]
        s  = len(tdf)
        ds = len(tdf.groupby('unp'))
        threshpass.append([minval,s,ds])
        minval += step
    res = np.array(threshpass)
    return res
```

```python
# Plot threshold analysis results
def plotthresh(res,label,**kwargs):
    fig,ax1 = plt.subplots(figsize=(10,5))
    sns.regplot(res[:,0],res[:,1],fit_reg=False,label="Spheres",color='darkslateblue',ax=ax1,**kwargs)
    for t1 in ax1.get_yticklabels():
        t1.set_color('darkslateblue')
    max1 = res[:,1].max()
    ax1.set_ylim(0-max1*.1,max1+max1*.1)
    ax1.set_yticks(np.arange(0,max1+max1*.1, int(max1+max1*.1)/10))
    ax2 = ax1.twinx()
    for t1 in ax2.get_yticklabels():
        t1.set_color('darkred')
    sns.regplot(res[:,0],res[:,2],fit_reg=False,label="Proteins",color='darkred',ax=ax2,**kwargs)
    max2 = res[:,2].max()
    ax2.set_ylim(0-max2*.1,max2+max2*.1)
    ax2.set_yticks(np.arange(0,max2+max2*.1, int(max2+max2*.1)/10))
    return ax1,ax2
```

```python
# Threshold analysis: Number of SNPs per sphere
res = threshold(df,'snpcount','allpop_fst_pval','>=')
ax1,ax2 = plotthresh(res,label="Spheres")
# res = threshold(df,'w_snpcount','w_allpop_fst_pval','>=')
# plotthresh(res,label="Windows")
plt.title("Sig Spheres Passing SNP Count Thresholds")
plt.xlabel("SNPs per Sphere")
ax1.set_ylabel("Spheres Passing Threshold",color='darkslateblue')
ax2.set_ylabel("Proteins Passing Threshold",color='darkred')
# plt.legend()
plt.tight_layout()
plt.show()
print "A mandatory SNP minimum of 2 has already been applied"
print "Thus, all significant results pass thresholds <=2 by default"
```

```python
from scipy.spatial import KDTree
def collapse_roi(df,structid,biounit):
    dft = df[ (df["structid"]==structid)
            & (df["biounit"]==biounit)
            & (df['snpcount']>2) 
            & (df['allpop_fst']>.05) 
            & (df['allpop_fst_pval']<0.05) 
            & (df['w_allpop_fst_pval']>0.01)
            & (df['allpop_fst']>df['w_allpop_fst'])]
    dft = dft[["structid","biounit","unp","model","chain","seqid","icode","x","y","z"]]
    if dft.empty:
        return dft
    dft = dft.copy().reset_index() # copy and reset index for significant-only residues
    dft["allpop_fst_roi"] = np.nan    # initialize column for RIDs
    crd = [list(c) for c in np.array(dft[["x","y","z"]],dtype=np.float64)]
    kdt = KDTree(crd)
    cdt = [c for c in crd] # candidate set

    # Define the nearest-neighbor walk
    def nnsearch(c,kdt,crd,frnt):
        nbrs  = [n for n in kdt.query_ball_point(c,15) if crd[n] in frnt]
        for n in nbrs:
            frnt.remove(crd[n]) # remove from candidate set
        return [crd.index(c)]+nbrs+[g for n in nbrs for g in nnsearch(crd[n],kdt,crd,frnt)] 

    # Run the nearest-neighbor walk and assign ROIs
    i = 0
    while len(cdt)>0:
        for n in nnsearch(cdt.pop(0),kdt,crd,cdt):
            dft.ix[n,"allpop_fst_roi"] = i
        i += 1
    return dft[["structid","biounit","unp","model","chain","seqid","icode","allpop_fst_roi"]]
```

```python
dfm = df.copy()
dfm["allpop_fst_roi"] = np.nan
for i,(structid,biounit) in dfm[["structid","biounit"]].drop_duplicates().iterrows():
    dft = collapse_roi(dfm,structid,biounit)
    dfm.combine_first(dft)
print "Number of distinct Fst ROI (of any size): %d"%len(dfm)
```

```python
# Threshold analysis: Number of spheres per ROI
gdf = dfm[-np.isnan(dfm["allpop_fst_roi"])].groupby(["allpop_fst_roi"])
minval = gdf.size().min()
maxval = gdf.size().max()
step   = 1
while minval <= maxval:
    tdf = gdf.filter(lambda x: len(x) >= minval)
    s  = len(tdf)
    threshpass.append([minval,s])
    minval += step
res = np.array(threshpass)
sns.regplot(res[:,0],res[:,1],fit_reg=False,label="ROIs",**kwargs)
plt.title("ROIs Passing Sphere-Count Thresholds")
plt.xlabel("Spheres in ROI")
plt.ylabel("ROI Passing Threshold")
plt.legend()
plt.tight_layout()
plt.show()
```
