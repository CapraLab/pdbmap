```python
# Imports, configurations, and initializations
from pandas import DataFrame,read_table
import pandas as pd
import numpy as np
# Import and initiailize matplotlib parameters
%matplotlib inline
import matplotlib.pylab as plt
import matplotlib as mpl
mpl.rcParams.update({'font.size': 8, 'font.family': 'serif'})
```

```python
def summarize(adf,percentile=99,title='Forgot to pass title'):
    
    # Protect the original dataframe from destructive actions
    df = adf.copy(deep=True)
#     print df.describe()
    
    # Create the figure
    fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(8,8))
    fig.suptitle(title,fontsize=10)
    fig.tight_layout()
    fig.subplots_adjust(top=0.96)
    
    # Reduce coef to real, positive, non-infinite values
    msg = [[],[],[],[]]
    for e,tcoef in enumerate(['npdcoef','dscoef5','dscoef10','dscoef15']):
        coef = df[tcoef].values
        cnt = len(coef)
        msg[e] =  ["%s"%tcoef]
        if len(tcoef)<8: msg[e][0] += '\t'
        msg[e] += [" # Total: %d"%cnt]
        nan = np.count_nonzero(np.isnan(coef))
        coef = coef[~np.isnan(coef)]
        msg[e] += [" # NaN:   %d"%nan]
        inf = np.count_nonzero(np.isinf(coef))
        coef = coef[~np.isinf(coef)]
        msg[e] += [" # Inf:   %d"%inf]
        cnt = len(coef)
        eq0 = cnt - np.count_nonzero(coef)
        coef = coef[coef>0]
        msg[e] += [" # Eq0:   %d"%eq0]
        valid = len(coef)
        msg[e] += [" # Valid: %d"%valid]
        thresh = np.percentile(coef,percentile)
        msg[e] += [" %d%%: %2.4f"%(percentile,thresh)]
        coef = coef[coef<thresh]
    
        # Plot the histogram
        i,j = e / 2, e % 2
        ax = axes[i][j] # <- note: now plotting the top-left w/ updated sig array
        minv = np.nanmin(coef)
        maxv = np.nanmax(coef)
        h1 = ax.hist(coef,bins=50,range=(minv,maxv),color='steelblue')
        ax.set_xlabel('%s Coefficient'%tcoef)
        ax.set_ylabel('Frequency')
    
    # Report statistics for each vector
#     for j in range(len(msg[0])):
#         print '\t'.join([msg[i][j] for i in range(4)])
    
#     # Plot the histogram of dscoef5
#     ax = axes[0][1] # <- note: now plotting the top-left w/ updated sig array
#     minv = np.nanmin(coef)
#     maxv = np.nanmax(coef)
#     h1 = ax.hist(coef,bins=50,range=(minv,maxv),color='steelblue')
#     ax.set_xlabel('DBSCAN Coefficient (Eps=5)')
#     ax.set_ylabel('Frequency')
    
#     # Plot the histogram of dscoef10
#     ax = axes[1][0] # <- note: now plotting the top-left w/ updated sig array
#     minv = np.nanmin(coef)
#     maxv = np.nanmax(coef)
#     h1 = ax.hist(coef,bins=50,range=(minv,maxv),color='steelblue')
#     ax.set_xlabel('DBSCAN Coefficient (Eps=10)')
#     ax.set_ylabel('Frequency')
    
#     # Plot the histogram of dscoef15
#     ax = axes[1][1] # <- note: now plotting the top-left w/ updated sig array
#     minv = np.nanmin(coef)
#     maxv = np.nanmax(coef)
#     h1 = ax.hist(coef,bins=50,range=(minv,maxv),color='steelblue')
#     ax.set_xlabel('DBSCAN Coefficient (Eps=15)')
#     ax.set_ylabel('Frequency')
```

## Empirical distributions

I'm using the structures from the evernote: Top DBSCAN Clustering Coefficient Results
These structures show a lot of diversity, so they should provide a good landscape of empirical results, at least as good as anything else. They also have the convenience being the most easily attainable list of structures as I glance at my notes.
They are:
* 3AZF
* 1VJ5
* 1SJE
* 3NWH
* 3O8E

For each dataset, I also found the top 5 most and least significant structures, with a secondary ordering by the coefficient magnitude. From these 5, I've selected a "good" and "bad" performer for each method.
ENSP00000343377_1 had an infinite score, but a p-value of 0.001998
3P8C had an infinite score, but a p-value of 0.000999 (better than entire empirical distribution)

```python
import subprocess as sp
random_structures = ['3azf','1vj5','1sje','3nwh','3o8e']
best_structures   = ['1QVO','4AOJ','2C6N','ENSP00000343377_1','3P8C']
worst_structures  = ['ENSP00000354505_1','3NR9','1ZMD','ENSP00000244533_1','4L28']
structures = random_structures + best_structures + worst_structures
metric = ['r' for i in range(5)] + ['pd','ds5','ds10','ds15','ds20']*2
status = ['na' for i in range(5)] + ['good' for i in range(5)] + ['bad' for i in range(5)]
# %cd ../..
# for i,s in enumerate(structures):
#     if i in [6,7,8,9,11,12,13,14]:
#         cmd = ["./pdbmap.py","-c","./config/v10.config","visualize",s,"1kg","dbscan","all","%d:2"%(5*(i%5))]
#         print ' '.join(cmd)
#         sp.check_output(cmd)
# %cd ../results/cluster_coefficient_1kg
#     with open("%s_1kg_empirical_distribution.txt"%s.lower(),'wb') as fout:
#         for line in sp.check_output(["grep","^%s"%s.upper(),"1kg_empirical_distribution.txt"]):
#             fout.write(line)
# This set of commands was executed server-side, do not run again
for row in sp.check_output(["wc","-l"] + ["%s_1kg_empirical_distribution.txt"%s.lower() for s in structures]).split('\n'):
    print row
```

```python
# Plot the empirical distributions for each of the random, best, and worst clustering structures
for i,s in enumerate(structures):
    df = read_table('%s_1kg_empirical_distribution.txt'%s.lower(),header=None,sep='\t',error_bad_lines=False,
                names=["structid","nres","snpres","nsnp","eps","mins","pdcoef","npdcoef",
                       "dscoef5","dscoef10","dscoef15","dummy","ndscoef10",
                       "ndscoef15","ndscoef20"])
    try:
        summarize(df,percentile=99,title='Empirical Distributions for %s (Metric=%s; Status=%s)'%(s.upper(),metric[i],status[i]))
    except:
        print "%s fails for some metrics (probably chosen for poor pdcoef and applied to dscoef with <3 SNPs)"%s
```

```python
df = read_table('3azf_1kg_empirical_distribution.txt',header=None,sep='\t',error_bad_lines=False,
                names=["structid","nres","snpres","nsnp","eps","mins","pdcoef","npdcoef",
                       "dscoef5","dscoef10","dscoef15","dummy","ndscoef10",
                       "ndscoef15","ndscoef20"])
print "Significant Pairwise Inverse Distance Coefficient"
summarize(df,percentile=99,title='Empirical Distributions for 3AZF')
```
