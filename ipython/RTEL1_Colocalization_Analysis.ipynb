```python
def snp_coords(coord_file):
    with gzip.open(coord_file,'rb') as fin:
        rows = [line.split()[3:8] for line in fin if line[:4]=="ATOM"]
        try:
            int(rows[0][1])
        except:
            with open(coord_file,'rb') as fin:
                rows = [line.split()[3:9] for line in fin if line[:4]=="ATOM"]
        rtel1_coords = np.array(rows)
        if rtel1_coords.shape[1] == 5:
            df = pd.DataFrame(rtel1_coords,columns=["res","seqid","x","y","z"])
            df["chain"] = "A"
        else:
            df = pd.DataFrame(rtel1_coords,columns=["res","chain","seqid","x","y","z"])
    df[["seqid"]] = df[["seqid"]].astype(np.int)
    df[["x","y","z"]] = df[["x","y","z"]].astype(np.float64)
    coords = np.array(zip(df.groupby(["chain","seqid"])["x"].mean(),df.groupby(["chain","seqid"])["y"].mean(),df.groupby(["chain","seqid"])["z"].mean()))
    coord_df = df[["res","chain","seqid"]].drop_duplicates()
    coord_df["x"] = coords[:,0]
    coord_df["y"] = coords[:,1]
    coord_df["z"] = coords[:,2]
    return coord_df
rtel1_coords = snp_coords(rtel1_coord_file)
```

```python
@np.vectorize
def score(D):
    if 0<D<=8:
        return 1.
    elif 8<D<=16:
        return np.sin((D-8)/5.1+.5*np.pi)
    else:
        return 0.
```

```python
# Predict disease status with a distance-weighted sum
def pathprox(dfq,dcodes=[],bcodes=[0],show_scores=False):
    k = len(dfq)
    dfq["Pathogenic"] = np.array([1 if d in dcodes else -1 for d in dfq["dcode"]]) # Set which dcodes to use
    dfq.ix[dfq["dcode"].isin(bcodes),"Pathogenic"] = 0 # Set the neutral variants
    num_neut = (dfq["Pathogenic"]==0).sum() 
    num_path = (dfq["Pathogenic"]==1).sum()
    pathprox = []
    for i,row in dfq.iterrows():
        # Calculate a "neutral score"
        kdt = KDTree(dfq.ix[dfq["Pathogenic"]==0,["x","y","z"]])
        dist,snps = kdt.query(row[["x","y","z"]],num_neut)
        neutral_score = np.sum(score(dist))/num_neut
        # Calculate a "pathogenic score"
        kdt = KDTree(dfq.ix[dfq["Pathogenic"]==1,["x","y","z"]])
        dist,snps = kdt.query(row[["x","y","z"]],num_path)
        pathogenic_score = np.sum(score(dist))/num_path
        pathprox.append(pathogenic_score/neutral_score)
    # Set all scores to real values (inf->max,nan->0)
    pathprox = np.array(pathprox)
    pathprox[np.isnan(pathprox)] = 0
    pathprox[np.isinf(pathprox)] = np.max(pathprox[np.isfinite(pathprox)])
    return pathprox
```

```python
def path_roc(dfq,dcodes=[0,1],pathcodes=[2],predcol="All_Path",label="",color="red",jitter=False):
    from sklearn.metrics import roc_curve,auc
    jitter = np.random.normal(0,0.01,1)[0] if jitter else 0
    dfq = dfq.sort(predcol,ascending=False)
    dfq["Pathogenic"] = np.array([1 if d in pathcodes else -1 for d in dfq["dcode"]])
    dfq.ix[dfq["dcode"]==0,"Pathogenic"] = 0 # Set to netrual variants
    dfq = dfq[dfq["Pathogenic"]>-1]
    fpr,tpr,thresh = roc_curve(dfq.ix[dfq["dcode"].isin(dcodes),"Pathogenic"],dfq.ix[dfq["dcode"].isin(dcodes),predcol])
    roc_auc = auc(fpr,tpr)
    fpr += jitter
    tpr += jitter
    plt.plot(fpr-0.01,tpr+0.01,label='%s (AUC = %0.2f)'%(label,roc_auc),color=color)
```

```python
# Plot the ROC for predicting RTEL1 IPF SNPs from RTELPATH
plt.figure(figsize=(10,10))
plt.title('Predicting IPF nsSNPs from known Pulmonary Fibrosis and Telomere Syndrome nsSNPs')
# ROC with Pulmonary Fibrosis Background
path_roc(coords,dcodes=[0,2],pathcodes=[2],predcol="PF_Path",label="IPF from PF",color="Gold",jitter=True)
# ROC with Telomere Syndrome Background
path_roc(coords,dcodes=[0,2],pathcodes=[2],predcol="Tel_Path",label="IPF from Telomere Syndrome",color="Blue",jitter=True)
# ROC with Pulmonary Fibrosis + Telomere Syndrome Background
path_roc(coords,dcodes=[0,2],pathcodes=[2],predcol="RTEL1_Path",label="IPF from RTEL1",color="Black",jitter=True)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.savefig("ipf_from_ipf-ts.pdf")
plt.show()
```

```python
# Plot the ROC for predicting RTEL1 IPF SNPs from different pathogenic backgrounds
plt.figure(figsize=(10,10))
plt.title('Predicting Telomere Syndrome Pathogenic nsSNPs from Different Pathogenic SNP Sets')
# ROC with Pulmonary Fibrosis background
path_roc(coords,dcodes=[0,1],pathcodes=[1],predcol="PF_Path",label="Telomere Syndrome from PF",color="red")
# ROC with Telomere Syndrome background
path_roc(coords,dcodes=[0,1],pathcodes=[1],predcol="Tel_Path",label="Telomere Syndrome from Telomere Syndrome",color="orange")
# ROC with Pulmonary Fibrosis + Telomere Syndrome Background
path_roc(coords,dcodes=[0,1],pathcodes=[1],predcol="RTEL1_Path",label="Telomere Syndrome from RTEL1",color="yellow")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```
