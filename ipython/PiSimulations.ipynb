```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import itertools
from scipy.spatial.distance import hamming
%matplotlib inline
```

```python
def pi(genos):
    g = genos.T
    return np.mean([hamming(g[i],g[j]) for i,j in itertools.combinations(xrange(g.shape[0]),2)])
```

```python
# For 1000 permutations
res = []
for i in xrange(100):
    # Generate sets of 2-10 SNPs with varying frequency, bias towards low frequencies
    mafs = np.random.standard_exponential(np.random.randint(2,10))/15.
    nsnp = len(mafs)
    cmaf = mafs.sum()
    mmaf = mafs.mean()
    # Generate a random set of 1000 genotypes over these 10 SNPs
    genos = np.array([np.random.random(1000)<maf for maf in mafs])
    mypi = pi(genos)
    res.append([nsnp,cmaf,mmaf,mypi])
res = np.array(res)
```

```python
from sklearn import linear_model
clf = linear_model.LinearRegression()
clf.fit(res[:,:-1],res[:,-1])
nsnpB,cmafB,mmafB = clf.coef_
intercept = clf.intercept_
print "%2.2f*nSNP + %2.2f*cMAF + %2.2f*mMAF + %2.2f"%(nsnpB,cmafB,mmafB,intercept)
```

```python
clf = linear_model.LinearRegression()
clf.fit(res[:,-2].reshape(100,1),res[:,-1])
mmafB = clf.coef_
intercept = clf.intercept_
print "%2.2f*mMAF + %2.2f"%(mmafB,intercept)
```

```python
# As a function of cumulative allele frequency
plt.scatter(res[:,0],res[:,-1])
plt.xlabel("nSNP")
plt.ylabel("Pi")
```
