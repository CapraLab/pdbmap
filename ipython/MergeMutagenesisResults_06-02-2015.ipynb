```python
# Load libraries
%matplotlib inline
import glob,gzip,os
import numpy as np
from numpy.random import randn
from itertools import combinations
import random
import pandas as pd
from scipy import stats
from scipy.spatial import KDTree
from sklearn.metrics import confusion_matrix
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 500)
```

```python
# Read the mutation disease classifications
dclass = {'WT':-1}
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/1kg3.txt','rb') as fin:
  dclass.update(dict((m.strip(),0) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/unclassified.txt','rb') as fin:
  dclass.update(dict((m.strip(),1) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/dyskeratosis.txt','rb') as fin:
  dclass.update(dict((m.strip(),2) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/published_pathogenic.txt','rb') as fin:
  dclass.update(dict((m.strip(),3) for m in fin))
dcolor = {-2:"orange",-1:"black",0:"darkgray",1:"green",2:"blue",3:"red"}
```

```python
# Reference dictionary for structure size
nres = {"ENSP00000424307_1":751,"ENSP00000353332_1":727,"ENSP00000359035_2":727,"ENSP00000359035_1":695,
        "ENSP00000353332_2":695,"ENSP00000359020_1":336,"ENSP00000424307_2":275,"ENSP00000359020_2":259,
        "ENSP00000353332_4":259,"ENSP00000353332_3":251,"ENSP00000359035_3":251,"ENSP00000424307_3":242,
        "ENSP00000425576_1":117,"ENSP00000425576_2":84,"ENSP00000353332_5":58,"ENSP00000359020_3":58,
        "ENSP00000353332_6":33,"ENSP00000359020_4":33}
```

```python
resdir  = "/Volumes/sivleyrm/pdbmap/results/Q9NZ71_mutagenesis_20150603-PROBLEMS"
names   = ["Mutation","Unrelaxed_Score","Unrelaxed_Rep","Min_Score","Mean_Score","Std_Score","Min_MT_RMSD","Mean_MT_RMSD","Std_MT_RMSD","rWT_RMSD"]
results = []
# bad_structs = ["ENSP00000353332_1","ENSP00000353332_4","ENSP00000359020_1","ENSP00000359035_2"]
bad_structs = ["ENSP00000353332_1","ENSP00000359035_2"]
valid_structs = ["ENSP00000353332_1","ENSP00000359035_2","ENSP00000353332_2","ENSP00000359035_1"]
for f in glob.glob("%s/*.results"%resdir):
  # Record the structid and biounit
  fname = os.path.basename(f)
  if fname in ["score.results","wt-rmsd.results","rwt-rmsd.results","delta.results"]: continue
  structid,biounit,_ = fname.split('.')[0].split('_')
  # Read the Wild-Type and Relaxed Wild-Type scores for this structure
  with open(f,'rb') as fin:
    wt  = float(fin.readline().split(':')[-1].strip())
    rwt = float(fin.readline().split(':')[-1].strip())
  # Extract results for this structure
  try:
    df = pd.read_csv(f,sep='\t',skiprows=3,names=names) # first three rows are comments and header
  except:
    print "\nException: %s\n"%f
    raise
  # Add structid and biounit columns
  sid = "%s_%s"%(structid,biounit)
  if sid not in valid_structs: continue
  df["Structid"] = [sid]*len(df)
  df["nres"]     = [nres[sid]]*len(df)
  df["Quality"]  = [int(sid not in bad_structs)]*len(df)
  # Reduce to minimum scores and RSMDs
#   df = df[["Mutation","Structid","Quality","Unrelaxed_Score","Min_Score","Min_MT_RMSD","rWT_RMSD"]]
  # Calculate ddG
  df["dREU"]  = (df["Unrelaxed_Score"] - rwt)*df["nres"]
  df["rdREU"] = (df["Min_Score"]       - rwt)*df["nres"]
  # Record the WT and rWT scores
  df = df.append({"Mutation":"WT","Unrelaxed_Score":rwt,"Structid":sid},ignore_index=True)
  # Index by structid, biounit, and mutation
  df.set_index(["Structid","Mutation"],inplace=True)
  results.append(df)
# Merge results
df = pd.concat(results)
df = df.reset_index()
nummut = len(df)
df = df[df["Mutation"].isin(dclass)] # Filter out incorrectly mapped SNPs
print "%d invalid mutations removed"%(nummut - len(df))
# Group structures into reliable and unreliable, and sort by decreasing size in each subgroup
good_structs = [sid for sid in df["Structid"].drop_duplicates() if sid not in bad_structs]
bad_structs.sort(key=lambda x: nres[x],reverse=True)
good_structs.sort(key=lambda x: nres[x],reverse=True)
order = bad_structs + good_structs
order = [o for o in order if o in valid_structs]
print "Sizes: \n",'\n'.join([str((o,len(df))) for o in order])
print "SNPs: \n",'\n'.join([str((o,len(df[df["Structid"]==o]))) for o in order])
print "Number of Mutation+Structure Pairs Before Pruning: %d"%len(df)
print "Number of mutations in reliable structures:        %d"%len(df[df["Quality"]==1])
print "Number of mutations in unreliable structures:      %d"%len(df[df["Quality"]==0])
```
