# Summary
This is a comparison of which genes are marked as significant by various binning methods. A bin will be called significant if it contains at least 3 nsSNPs and has an Fst greater than 0.05. A second comparison will require an Fst > 0.15.

```python
def conf_matrix(ax,grp1,grp2,lab1,lab2):
    # Calculate the confusion matrix
    grp1,grp2 = grp1.align(grp2,join="outer",axis=0)
    grp1 = grp1.fillna(False)
    grp2 = grp2.fillna(False)
    if not (grp1 & grp2).sum():
        cm = np.zeros((2,2))
        cm[0][0] = len(grp1)
    else:
        cm = confusion_matrix(grp1,grp2)
    ax.text(0,0,'{0:,}'.format(cm[0][0]),fontsize=40,horizontalalignment='center',verticalalignment='center',color='white',path_effects=[PathEffects.withStroke(linewidth=3,foreground="black")])
    ax.text(0,1,'{0:,}'.format(cm[1][0]),fontsize=40,horizontalalignment='center',verticalalignment='center',color='white',path_effects=[PathEffects.withStroke(linewidth=3,foreground="black")])
    ax.text(1,0,'{0:,}'.format(cm[0][1]),fontsize=40,horizontalalignment='center',verticalalignment='center',color='white',path_effects=[PathEffects.withStroke(linewidth=3,foreground="black")])
    ax.text(1,1,'{0:,}'.format(cm[1][1]),fontsize=40,horizontalalignment='center',verticalalignment='center',color='white',path_effects=[PathEffects.withStroke(linewidth=3,foreground="black")])
    cm[0][0] = 0 # improve visualization of significant agreement/disagreement
    # Show confusion matrix in a separate window
    ax.imshow(cm,cmap=plt.cm.jet,interpolation='nearest')
#     ax.set_title("%s vs %s"%(lab1,lab2),fontsize=35)
    ax.set_ylabel(lab1,fontsize=30)
    ax.set_xlabel(lab2,fontsize=30)
    ax.set_xlim([-0.5,1.5])
    ax.set_ylim([-0.5,1.5])
    ax.set_xticks([0,1])
    ax.set_yticks([0,1])
    ax.set_xticklabels(["Not Significant","Significant"])
    ax.set_yticklabels(["Not Significant","Significant"],rotation='vertical')
    ax.grid(False,'both','both')
```

```python
def unlist_col(col):
    return [str(c).split(',') for c in col]
```

```python
resdir = "/Volumes/sivleyrm/pdbmap/results/fst_analysis_20150616"
shcols = ["fst","nsnp","snplist","maflist","fstlist","unpac","hgnc"]
def load_fst(metric,unq_cols):
    df = pd.read_csv("%s/fst_by_%s.txt"%(resdir,metric),sep=" ",header=None,names=unq_cols+shcols,encoding="utf-8")
    df["snplist"] = unlist_col(df["snplist"])
    df["maflist"] = unlist_col(df["maflist"])
    df["fstlist"] = unlist_col(df["fstlist"])
    df.sort("unpac")
    print "%d %s bins"%(len(df),metric)
    return df
```

```python
# For each gene, determine if it contains a significant bin
# sig_grp = [fst.groupby("unpac").apply(lambda x: len(x[x["fst"]>0.05])>0) for fst in fst_grp]
# Require at least 3 criteria-passing spheres
sig_grp = [fst.groupby("unpac").apply(lambda x: len(x[x["fst"]>0.05])>0) for fst in fst_grp[:-1]]
spheres = fst_grp[-1]
spheres = spheres.groupby("unpac").apply(lambda x: len(x[x["fst"]>0.05])>2)
sig_grp.append(spheres)
```

```python
# Generate an 8x8 confusion matrix
# Calculate the confusion matrix
fig,axes = plt.subplots(len(sig_grp),len(sig_grp),figsize=(50,50))
for i,grp1 in enumerate(sig_grp):
    for j,grp2 in enumerate(sig_grp):
        if j <= i:
            conf_matrix(axes[i][j],grp1,grp2,fst_labs[i],fst_labs[j])
        else:
            axes[i][j].set_frame_on(False)
            axes[i][j].axes.get_xaxis().set_visible(False)
            axes[i][j].axes.get_yaxis().set_visible(False)
plt.tight_layout()
```
