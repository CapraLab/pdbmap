```python
# Load the results file into a pandas data frame
results_fname = '/Volumes/sivleyrm/pdbmap/results/sliding_sphere_10/split/obs/bystruct/sliding_sphere_%s-%d.txt.gz'%(structid,biounit)
header_fname  = 'scripts/sphere_header.txt'
with open(header_fname,'rb') as hfin:
    header = [h.strip() for h in hfin.readlines()]
with gzip.open(results_fname,'rb') as obs_fin:
    df = pd.read_table(obs_fin,delimiter='\t',skiprows=1,names=header)
    df.drop_duplicates(['x','y','z'],inplace=True)
```

```python
# Correct coding issue present in some results files
for col in mean_ddaf_cols+absmean_ddaf_cols+w_mean_ddaf_cols+w_absmean_ddaf_cols:
    pcol = col+'_pval'
    df.ix[np.isinf(df[col]),pcol] = np.nan
    df.ix[np.isinf(df[col]),col]  = 0
    df.ix[np.isnan(df[col]),pcol] = np.nan
    df.ix[np.isnan(df[col]),col]  = 0
```

```python
# Report the number of spheres with >0 SNPs in each population or combination of populations
print "Number of spheres with >0 SNPs:"
print (df[count_cols+combcount_cols] > 0).sum()
# Report the number of spheres with >1 SNP in each population or combination of populations
print "\nNumber of spheres with >1 SNPs:"
print (df[count_cols+combcount_cols] > 1).sum()
```

```python
# Set sphere metrics with <2 SNPs contributing to 0
# Set sphere p-values with <2 SNPs contributing to 1
# Single-Population Metrics
for count in count_cols:
    if count < 2:
        val,pval1,pval2 = count.split('_')[0]+"_cdaf",count+"_pval",count.split('_')[0]+"_cdaf_pval"
        df.ix[:,val] = 0
        df.ix[:,[pval1,pval2]] = [1,1]
# Multi-Population Metrics (Including Windowed Metrics)
for count in combcount_cols:
    if count < 2:
        comb = '_'.join(count.split('_')[:-1])
        val1,pval1   = comb+"_meanddaf",comb+"_meanddaf_pval"
        val2,pval2   = '_'.join(comb.split('_')[::-1])+"_meanddaf",'_'.join(comb.split('_')[::-1])+"_meanddaf_pval"
        val3,pval3   = ''.join(comb.split('_'))+"_meanddaf",''.join(comb.split('_'))+"_meanddaf_pval"
        wval1,wpval1 = "w_"+val1,"w_"+pval1
        wval2,wpval2 = "w_"+val2,"w_"+pval2
        wval3,wpval3 = "w_"+val3,"w_"+pval3
        df.ix[:,[val1,val2,val3,wval1,wval2,wval3]] = [0,0,0,0,0,0]
        df.ix[:,[pval1,pval2,pval3,wpval1,wpval2,wpval3]] = [1,1,1,1,1,1]
```

```python
# Load the permutation file
perm_fname = '/Volumes/sivleyrm/pdbmap/results/sliding_sphere_10/split/perm/bystruct/sliding_sphere_perm_%s-%d.npz'%(structid,biounit)
with np.load(perm_fname) as npzfile:
    npz = npzfile.items()[0][1]
# Flatten the permutation results and store as a pandas data frame
flatten = (npz.shape[0]*npz.shape[1],npz.shape[2])
dfp = pd.DataFrame(npz.reshape(flatten),columns=df.columns[:npz.shape[2]])
```

```python
# Correct coding issue present in some of the results files
for col in mean_ddaf_cols+absmean_ddaf_cols+w_mean_ddaf_cols+w_absmean_ddaf_cols:
    df.ix[np.isinf(df[col]),col]  = 0
    df.ix[np.isnan(df[col]),col]  = 0
```

```python
chains = df.chain.unique()
chains.sort()
minseq = [df[df.chain==c].seqid.min() for c in chains]
maxseq = [df[df.chain==c].seqid.max() for c in chains]
print chains,minseq,maxseq
```

```python
METRIC  = 'eurafr_meanddaf'
WMETRIC = 'w_eurafr_meanddaf'
PVALUE  = 'eurafr_meanddaf_pval'
WPVALUE = 'w_eurafr_meanddaf_pval'
fig, axes = plt.subplots(len(chains),figsize=(max(maxseq)/4,5*len(chains)))
if len(chains) < 2:
    axes = [axes]
for i,chain in enumerate(chains):
    # For each residue in the sequence, show the permutation boxplot and observed value
    dft    = df[(df.chain==chain) & (df.seqid>=minseq[i]) & (df.seqid<=maxseq[i])]
    dfpt   = dfp[(dfp.chain==chain) & (df.seqid>=minseq[i]) & (dfp.seqid<=maxseq[i])]
    cdict  = {np.nan:'blue',0:'blue',1:'red'} # Significance coloring
    wcdict = {np.nan:'black',0:'black',1:'red'}
    sns.boxplot(dfpt[METRIC],groupby=dfpt.seqid,ax=axes[i])
    axes[i].scatter(dft.seqid-minseq[i],dft[WMETRIC],color=[wcdict[x<0.05] for x in dft[WPVALUE]],linewidths=2,marker='x',s=150)
    axes[i].scatter(dft.seqid-minseq[i],dft[METRIC],color=[cdict[x<0.05] for x in dft[PVALUE]],marker='o',s=75)
    stat  = METRIC.split('_')[-1][:-4]
    statp = stat+'p'
    axes[i].set_title("Observed Sphere Values in Permutation Context\n %s[%d].%s deltaDAF (%s) EUR vs AFR"%(structid,biounit,chain,stat),fontsize=15)
    plt.setp(axes[i].get_xticklabels(), rotation=30, horizontalalignment='right')
    # Generate Chimera attribute files
    with open("temp/%s-%d_%s.attr"%(structid,biounit,stat),'wb') as fout:
        fout.write("attribute: %s\nmatch mode: 1-to-1\nrecipient: residues\n\n"%stat)
        for _,row in df[['chain','seqid',METRIC]].iterrows():
            fout.write("\t:%d.%s\t%.6f\n"%(row[1],row[0],row[2]))
    with open("temp/%s-%d_%s.attr"%(structid,biounit,statp),'wb') as fout:
        fout.write("attribute: %s\nmatch mode: 1-to-1\nrecipient: residues\n\n"%statp)
        for _,row in df[['chain','seqid',METRIC]].iterrows():
            fout.write("\t:%d.%s\t%.6f\n"%(row[1],row[0],row[2]))
plt.tight_layout()
savefig('temp/deltaDAF_Distribution_%s-%d.pdf'%(structid,biounit))
```
