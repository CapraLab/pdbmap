```python
def cohens_kappa(v1,v2):
    if len(v1) != len(v2): raise Exception("Vectors must be of equal length")
    N   = float(len(v1))
    v1T = v1.sum() / N
    v2T = v2.sum() / N
    pe  = (v1T*v2T) + ((1-v1T)*(1-v2T))
    pa  = (v1 == v2).sum() / N
    return (pa-pe)/(1-pe)
```

```python
def jaccard_index(cm):
    return float(cm[1][1]) / (cm[1][0]+cm[0][1]+cm[0][0])
```

```python
# Identifies overlapping spheres forming an ROI
# and assigns a biounit-unique RID to the set
def collapse_roi(df,structid,biounit,pmetric,metric):
    """ Significant spheres must have a p-value < 0.01 and an observed value >0.05 """
    dft = df.ix[(df[pmetric]<0.01) & (df[metric]>0.05) & (df.structid==structid) & (df.biounit==biounit),
                ["structid","biounit","model","chain","seqid","icode","x","y","z"]]
    if dft.empty: return dft # biounit contains no significant spheres
    dft = dft.copy().reset_index() # copy and reset index for significant-only residues
    crd = [list(c) for c in np.array(dft[["x","y","z"]],dtype=np.float64)]
    kdt = KDTree(crd)
    cdt = [c for c in crd] # candidate set
    dft["%s_roi"%pmetric] = np.nan    # initialize column for RIDs
    # Define the nearest-neighbor walk
    def nnsearch(c,kdt,crd,frnt):
        nbrs  = [n for n in kdt.query_ball_point(c,15) if crd[n] in frnt]
        for n in nbrs:
            frnt.remove(crd[n]) # remove from candidate set
        return [crd.index(c)]+nbrs+[g for n in nbrs for g in nnsearch(crd[n],kdt,crd,frnt)] 
    # Run the nearest-neighbor walk and assign ROIs
    i = 0
    while len(cdt)>0:
        for n in nnsearch(cdt.pop(0),kdt,crd,cdt):
            dft.ix[n,"%s_roi"%pmetric] = i
        i += 1
    # Return the ROI-annotated copy of df
    return dft

for pcolset in pvalue_cols[3:5]: # abs(dDAF) and Fst
    for pcol in pcolset:
        col = '_'.join(pcol.split('_')[:-1])
        dfroi = []
        for i,(sid,biounit) in enumerate(rlist):
            print "\rROI determination for %20s %3d%% complete (%s.%s)"%(pcol,int(float(i)/len(rlist)*100),sid,biounit),
            dfroi.append(collapse_roi(df,sid,biounit,pcol,col))
        print "\rROI determination for %20s 100% complete"%(pcol)
        dfroi  = pd.concat(dfroi)
        if not dfroi.empty:
            print "Largest significant %s ROI"%col
            dfroig = dfroi.groupby(["structid","biounit","%s_roi"%pcol])
            head = 5 if len(dfroig) > 5 else len(dfroig)
            for _,group in sorted(dfroig,key=lambda x: len(x[1]),reverse=True)[:head]:
                print "%s.%s[%d]; %d spheres"%(group["structid"].iloc[0],group["biounit"].iloc[0],group["%s_roi"%pcol].iloc[0],len(group))
            print ''
    print "=================\n"
```

```python
def top_scores(df,metrics,N=100,order=None,mins=None,maxs=None,ascending=[False]):
    dfc = df.copy()
    # Enforce minimums
    for i,m in enumerate(mins):
        if m:
            dfc = dfc[dfc[metrics[i]]>m]
    # Enforce maximums
    for i,m in enumerate(maxs):
        if m:
            dfc = dfc[dfc[metrics[i]]<m]
    # Define sort order
    order = order if order else range(len(metrics))
    keys  = [metrics[o] for o in order]
    # Ensure numeric column types
    for key in keys:
        dfc[key] = dfc[key].astype(np.float64)
    # Sort the dataframe
    dfc = dfc.sort(keys,ascending=ascending)
    print "Number of spheres meeting criteria: %d"%len(dfc)
    if len(dfc)<1: return []

    # Identify the top N structures containing
    top_structs,i,sids = dict(),0,[]
    while len(top_structs) < N and i<len(dfc):
        sid = dfc.iloc[i].structid
        if sid not in top_structs:
            top_structs[sid] = 1
            sids.append(sid)
        else:
            top_structs[sid] += 1
        i+= 1
    
    # Top N structures
    # Structures ordered by the rank of their highest-ranking sphere
    # Counts are for all spheres with a higher rank than the final sphere considered
    for sid in sids:
        print sid,top_structs[sid]

    # Top N spheres
    dfc = dfc.head(N)

    cols = ["structid",'seqid','chain']+metrics
    print ' \t'+'\t'.join(cols)
    for row in dfc.ix[:,cols].itertuples():
        print '\t'.join(str(r) for r in row)
#     print "\n%s  %s"%(metrics[0],'\n'.join(["%s:%d.%s: %2.2f, p≤%1.1e res=%d afr=%d eur=%d"%( \
#                                         sid,seqid,chain,val,pval,res,afr,eur) \
#                                         for _,sid,seqid,chain,pval,val,res,afr,eur in dfc.ix[:,cols].itertuples()]))
    return [(sid,top_structs[sid]) for sid in sids]
```

```python
from scipy.spatial import KDTree
struct = ("3OG6","1")
dft = df.ix[(df["allpop_fst_pval"]<0.01) & (df["allpop_fst"]>0.1) & (df["structid"]==struct[0]) & (df["biounit"]==struct[1]),
            ["structid","biounit","model","chain","seqid","icode","allpop_fst","allpop_fst_pval","x","y","z","snpcount"]]
dft = dft.reset_index()
pos  = np.array(dft[["x","y","z"]],dtype=np.float64)
pos  = [list(p) for p in pos] # Map from KDTree to coordinates
kdt  = KDTree(pos)            # KDTree for queries
rois = []                     # Lists of overlapping spheres
frnt = [p for p in pos]       # Frontier of ungrouped spheres

def nnsearch(coord,kdt,pos,frnt):
    nbrs  = [n for n in kdt.query_ball_point(coord,15) if pos[n] in frnt]
    for n in nbrs:
        frnt.remove(pos[n]) # global removal
    return nbrs+[g for n in nbrs for g in nnsearch(pos[n],kdt,pos,frnt)]   

while len(frnt)>0:
    rois.append(nnsearch(frnt.pop(0),kdt,pos,frnt))

sig_rois = [dft.ix[r,["model","chain","seqid","icode"]].values for r in rois]

colors = ['1,0,0,0.5','0,1,0,0.5','0,0,1,0.5']#,'orange','brown','yellow','pink']
for i,roi in enumerate(sig_rois):
    print "ROI %d Sphere-Creation Statements"%i
    print ';'.join("shape sphere radius 10 mesh true center #0.%d:%d.%s@CA color %s"%(r[0]-1,r[2],r[1],colors[i]) for r in roi)
    
```
