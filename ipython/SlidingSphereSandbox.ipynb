```python
# Load the results file into a pandas data frame
import pandas as pd
results_fname = 'results/sliding_sphere_10/sliding_sphere_1KG_10A.txt'
df = pd.read_table(results_fname,delimiter='\t')
```

```python
count_cols    = ['amr_count','asn_count','eur_count','afr_count']
cdaf_cols     = ['amr_cdaf','asn_cdaf', 'eur_cdaf', 'afr_cdaf']
sumddaf_cols  = ['amr_asn_sumddaf', 'amr_eur_sumddaf', 'amr_afr_sumddaf', 'asn_eur_sumddaf', 'asn_afr_sumddaf', 'eur_afr_sumddaf']
meanddaf_cols = ['amr_asn_meanddaf','amr_eur_meanddaf','amr_afr_meanddaf','asn_eur_meanddaf','asn_afr_meanddaf','eur_afr_meanddaf']
minddaf_cols  = ['amr_asn_minddaf', 'amr_eur_minddaf', 'amr_afr_minddaf', 'asn_eur_minddaf', 'asn_afr_minddaf', 'eur_afr_minddaf']
maxddaf_cols  = ['amr_asn_maxddaf', 'amr_eur_maxddaf', 'amr_afr_maxddaf', 'asn_eur_maxddaf', 'asn_afr_maxddaf', 'eur_afr_maxddaf']
nest_cols     = [count_cols,cdaf_cols,sumddaf_cols,meanddaf_cols,minddaf_cols,maxddaf_cols]
count_pval_cols    = ['amr_count_pval','asn_count_pval','eur_count_pval','afr_count_pval']
cdaf_pval_cols     = ['amr_cdaf_pval','asn_cdaf_pval', 'eur_cdaf_pval', 'afr_cdaf_pval']
sumddaf_pval_cols  = ['amr_asn_sumddaf_pval', 'amr_eur_sumddaf_pval', 'amr_afr_sumddaf_pval', 'asn_eur_sumddaf_pval', 'asn_afr_sumddaf_pval', 'eur_afr_sumddaf_pval']
meanddaf_pval_cols = ['amr_asn_meanddaf_pval','amr_eur_meanddaf_pval','amr_afr_meanddaf_pval','asn_eur_meanddaf_pval','asn_afr_meanddaf_pval','eur_afr_meanddaf_pval']
minddaf_pval_cols  = ['amr_asn_minddaf_pval', 'amr_eur_minddaf_pval', 'amr_afr_minddaf_pval', 'asn_eur_minddaf_pval', 'asn_afr_minddaf_pval', 'eur_afr_minddaf_pval']
maxddaf_pval_cols  = ['amr_asn_maxddaf_pval', 'amr_eur_maxddaf_pval', 'amr_afr_maxddaf_pval', 'asn_eur_maxddaf_pval', 'asn_afr_maxddaf_pval', 'eur_afr_maxddaf_pval']
nest_pval_cols     = [count_pval_cols,cdaf_pval_cols,sumddaf_pval_cols,meanddaf_pval_cols,minddaf_pval_cols,maxddaf_pval_cols]
nest_all = [nest_cols,nest_pval_cols]
```

```python
# Correct error in pvalue calculations pval=1-pval
df.loc[:,count_pval_cols]    = 1 - df.loc[:,count_pval_cols]
df.loc[:,cdaf_pval_cols]     = 1 - df.loc[:,cdaf_pval_cols]
df.loc[:,sumddaf_pval_cols]  = 1 - df.loc[:,sumddaf_pval_cols]
df.loc[:,meanddaf_pval_cols] = 1 - df.loc[:,meanddaf_pval_cols]
df.loc[:,minddaf_pval_cols]  = 1 - df.loc[:,minddaf_pval_cols]
df.loc[:,maxddaf_pval_cols]  = 1 - df.loc[:,maxddaf_pval_cols]

# Mask values and pvalues where value is 0
# for i,col_stat in enumerate(nest_cols):
#     for j,col in enumerate(col_stat):
#         col_pval = nest_pval_cols[i][j]
#         df.loc[(df[col] == 0),col_pval] = np.nan
#         df.loc[(df[col] == 0),col]      = np.nan
```

```python
# Boxplot for each count, cDAF, or ΔDAF statistic, stratified by population
for colset in nest_cols:
    plt.figure(figsize=(15,6))
    sns.set_style("whitegrid")
    sns.set_palette("hls")
    sns.boxplot(df.ix[:,colset])
    plt.tight_layout()
    plt.show()
```

```python
# Boxplot for each pvalues for count, cDAF, or ΔDAF statistic, stratified by population
for colset in nest_pval_cols:
    plt.figure(figsize=(15,6))
    sns.set_style("whitegrid")
    sns.set_palette("hls")
    sns.boxplot(-np.log(df.ix[:,colset]))
    plt.hlines(2,0,10,'red',linestyles='dotted')
    plt.hlines(1.3,0,10,'blue',linestyles='dotted')
    plt.tight_layout()
    plt.show()
```

```python
# Boxplot for the number of significant spheres for count, cDAF, or ΔDAF statistic, stratified by population, grouped protein
for colset in nest_pval_cols:
    ndf = []
    plt.figure(figsize=(15,6))
    for col in colset:
        sns.set_style("whitegrid")
        sns.set_palette("hls")
        dfc = df[df[col]<0.05].groupby('structid').size()
        ndf.append(dfc)
    sns.boxplot(ndf,names=colset)
    plt.tight_layout()
    plt.show()
```

```python
# Boxplot for significant sphere values for count, cDAF, or ΔDAF statistic, stratified by population, grouped protein
for i,colset in enumerate(nest_pval_cols):
    ndf = []
    plt.figure(figsize=(15,6))
    for j,col in enumerate(colset):
        sns.set_style("whitegrid")
        sns.set_palette("hls")
        dfc = df[df[col]<0.05][nest_cols[i][j]]
        ndf.append(dfc)
    sns.boxplot(ndf,names=nest_cols[i])
    plt.tight_layout()
    plt.show()
```

```python
# Boxplot for mean value of significant spheres for count, cDAF, or ΔDAF statistic, stratified by population, grouped protein
for i,colset in enumerate(nest_pval_cols):
    ndf = []
    plt.figure(figsize=(15,6))
    for j,col in enumerate(colset):
        sns.set_style("whitegrid")
        sns.set_palette("hls")
        dfc = df[df[col]<0.05].groupby('structid')[nest_cols[i][j]].mean()
        ndf.append(dfc)
    sns.boxplot(ndf,names=nest_cols[i])
    plt.tight_layout()
    plt.show()
```

```python
# List the proteins with the *greatest number of* significant spheres for count, cDAF, and ΔDAF, stratified by population
for colset in nest_pval_cols:
    ndf = []
    for col in colset:
        dfc = df[df[col]<0.05].groupby('structid').size()
        dfc.sort()
        print "%s  %s"%(col,' | '.join(["%s: %.2f"%(sid,cnt) for sid,cnt in dfc.tail(5).iteritems()][::-1]))
    print ''
```

```python
# List the proteins with the *most extremely* significant spheres for count, cDAF, and ΔDAF, stratified by population
for i,colset in enumerate(nest_pval_cols):
    ndf = []
    for j,col in enumerate(colset):
        dfc = df[df[col]<0.05].groupby('structid').min().sort([col,nest_cols[i][j]],ascending=[True,False]).head(5)
        print "%s  %s"%(col,' | '.join(["%s: %2.2f, p≤%1.1e"%(sid,val,pval) for sid,_,pval,val in dfc.ix[:,["structid",col,nest_cols[i][j]]].itertuples()]))
    print ''
```
