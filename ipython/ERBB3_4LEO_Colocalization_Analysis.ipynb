```python
# Load libraries
%matplotlib inline
import glob,gzip,os
import numpy as np
from numpy.random import randn
from itertools import combinations
import random
import pandas as pd
from scipy import stats
from scipy.spatial import KDTree
from sklearn.metrics import confusion_matrix
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 500)
```

```python
# Read the mutation disease classifications
dclass = {}
# Contain variants in the HGVS form R###A
with open('/Volumes/sivleyrm/pdbmap/temp/erbb3/4leo_exac.txt','rb') as fin:
  dclass.update(dict((m.strip(),1) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/erbb3/4leo_1kg3.txt','rb') as fin:
  dclass.update(dict((m.strip(),0) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/erbb3/4leo_cosmic.txt','rb') as fin:
  dclass.update(dict((m.strip(),2) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/erbb3/humsavar.txt','rb') as fin:
  dclass.update(dict((m.strip(),3) for m in fin))
# Only one "candidate" variant to test
pos2snp = dict((int(key[1:-1]),key) for key in dclass.keys())
dclass["R391W"] = 4
# There is 1 overlapping Cosmic mutation (R391L)
# There is 1 overlapping ExAC mutation (R391W)
# Force 391 to map to the candidate allele
pos2snp[391] = "R391W"
cls2code = {"1000 Genomes":0,"ExAC":1,"Cosmic":2,"Humsavar":3,"R391W":4}
```

```python
coord_file = "data/rcsb/biounit/coordinates/all/4leo.pdb1.gz"
```

```python
def snp_coords(coord_file,chains=[],resrenumber=None):
    chains = chains if type(chains)==type([]) else [chains]
    with gzip.open(coord_file,'rb') as fin:
        rows = [line.split()[3:9] for line in fin if line[:4]=="ATOM"]
        coords = np.array(rows)
        df = pd.DataFrame(coords,columns=["res","chain","seqid","x","y","z"])
        # Reduce the coordinates to relevant chains
        df = df[df["chain"].isin(chains)]
    df[["seqid"]] = df[["seqid"]].astype(np.int)
    print "Sequence range:",df["seqid"].min(),"-",df["seqid"].max()
    if resrenumber:
        df["seqid"] = df["seqid"].apply(lambda x: x+resrenumber)
        print "Renumbered:",df["seqid"].min(),"-",df["seqid"].max()
    df[["x","y","z"]] = df[["x","y","z"]].astype(np.float64)
    coord_df = df[["res","chain","seqid"]].drop_duplicates()
    # Simplify residue coordinates to the center of mass
    coords = np.array(zip(df.groupby(["chain","seqid"])["x"].mean(),df.groupby(["chain","seqid"])["y"].mean(),df.groupby(["chain","seqid"])["z"].mean()))
    coord_df["x"] = coords[:,0]
    coord_df["y"] = coords[:,1]
    coord_df["z"] = coords[:,2]
    return coord_df
coords = snp_coords(coord_file,chains=["C"],resrenumber=19) # the first observed residue (8) is actually residue 27
coords = coords[coords["seqid"].isin(pos2snp.keys())]
```

```python
coords["dcode"] = coords["seqid"].apply(lambda x: dclass[pos2snp[x]])
```

```python
# Predict disease status with a kNN approach
def WAP(nq,nr,dqr,t=10.):
    return nq*nr*np.exp(-dqr**2/(2*t**2))
def pathprox(dfq,bcodes=[],dcodes=[]):
    k = len(dfq)
    dfq["Pathogenic"] = np.array([1 if d in dcodes else -1 for d in dfq["dcode"]]) # Label the pathogenic variants
    dfq.ix[dfq["dcode"].isin(bcodes),"Pathogenic"] = 0                                # Label the neutral variants
    num_neut = (dfq["Pathogenic"]==0).sum() 
    num_path = (dfq["Pathogenic"]==1).sum()
    pathprox = []
    for i,row in dfq.iterrows():
        # Calculate a "neutral score"
        kdt = KDTree(dfq.ix[dfq["Pathogenic"]==0,["x","y","z"]])
        dist,snps = kdt.query(row[["x","y","z"]],num_neut)
        # test:
        neutral_score = np.sum([WAP(1,1,d) for d in dist if d>0])
#         neutral_score = np.sum(1./dist[dist>0])/num_neut
        # Calculate a "pathogenic score"
        kdt = KDTree(dfq.ix[dfq["Pathogenic"]==1,["x","y","z"]])
        dist,snps = kdt.query(row[["x","y","z"]],num_path)
        #test:
        pathogenic_score = np.sum([WAP(1,1,d) for d in dist if d>0])
#         pathogenic_score = np.sum(1./dist[dist>0])/num_path
        pathprox.append(pathogenic_score/neutral_score)
#         pathprox.append(neutral_score/pathogenic_score)
    return np.array(pathprox)
```

```python
def path_roc(dfq,bcodes=[0,1],pcodes=[2],predcol="1000G_Cosmic",label="",color="red"):
    from sklearn.metrics import roc_curve,auc
    jitter = np.random.normal(0,0.02,1)[0]
    dfq = dfq.sort(predcol,ascending=False)
    dfq["Pathogenic"] = np.array([1 if d in pcodes else -1 for d in dfq["dcode"]]) # Label pathogenic and excluded
    dfq.ix[dfq["dcode"].isin(bcodes),"Pathogenic"] = 0                                # Label benign
    dfq = dfq[dfq["Pathogenic"]>-1]
    fpr,tpr,thresh = roc_curve(dfq.ix[dfq["dcode"].isin(bcodes+pcodes),"Pathogenic"],dfq.ix[dfq["dcode"].isin(bcodes+pcodes),predcol])
    roc_auc = auc(fpr,tpr)
    fpr += jitter
    tpr += jitter
    plt.plot(fpr-0.01,tpr+0.01,label='%s (AUC = %0.2f)'%(label,roc_auc),color=color)
```

```python
# Plot the ROC for predicting Cosmic SNPs from different benign backgrounds
plt.figure(figsize=(10,10))
plt.title('Predicting IPF nsSNPs from Different Pathogenic SNP Sets')
# ROC for 1000G vs Cosmic
path_roc(coords,bcodes=[0],pcodes=[2],predcol="1000G_Cosmic",label="Cosmic from 1000G",color="red")
# ROC for ExAC vs Cosmic
path_roc(coords,bcodes=[0,1],pcodes=[2],predcol="ExAC_Cosmic",label="Cosmic from ExAC",color="orange")
# ROC for 1000G vs Humsavar
path_roc(coords,bcodes=[0],pcodes=[3],predcol="1000G_Humsavar",label="Humsavar from 1000G",color="blue")
# ROC for ExAC vs Humsavar
path_roc(coords,bcodes=[0,1],pcodes=[3],predcol="ExAC_Humsavar",label="Humsavar from ExAC",color="purple")

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```
