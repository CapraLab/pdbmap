```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
@np.vectorize
def fst(af1,af2):
    afs  = [af1,af2]
    gaf  = np.mean(afs)
    nhat = np.sum(((gaf-afs)**2)/len(afs))
    dhat = gaf*(1-gaf)
    return nhat/dhat

@np.vectorize
def nhat(af1,af2):
    afs = [af1,af2]
    gaf = np.mean(afs)
    return np.mean((gaf-afs)**2)

@np.vectorize
def dhat(af1,af2):
    afs = [af1,af2]
    gaf = np.mean(afs)
    
    return gaf*(1-gaf)
```

```python
af = np.arange(0,1000,dtype=np.float64)/10000
af1,af2 = np.meshgrid(af,af)
fmat = fst(af1,af2)
nmat = nhat(af1,af2)
dmat = dhat(af1,af2)
print "Minimum Fst = %.4f"%np.nanmin(fmat)
print "Maximum Fst = %.4f"%np.nanmax(fmat)
print "##"
print "Minimum Nhat = %.4f"%np.nanmin(nmat)
print "Maximum Nhat = %.4f"%np.nanmax(nmat)
print "##"
print "Minimum Dhat = %.4f"%np.nanmin(dmat)
print "Maximum Dhat = %.4f"%np.nanmax(dmat)
```

### Setup the same protocol for 2-SNP simulations

```python
af = np.arange(0,1000,10,dtype=np.float64)/1000
af1,af2 = np.meshgrid(af,af)
fmat = fst(af1,af2)
nmat = nhat(af1,af2)
dmat = dhat(af1,af2)
print "Minimum Nhat = %.4f"%np.nanmin(nmat)
print "Maximum Nhat = %.4f"%np.nanmax(nmat)
print "##"
print "Minimum Dhat = %.4f"%np.nanmin(dmat)
print "Maximum Dhat = %.4f"%np.nanmax(dmat)

# Now we're going to meshgrid these with themselves, to get multi-SNP sets
nmat1,nmat2 = np.meshgrid(nmat,nmat)
dmat1,dmat2 = np.meshgrid(dmat,dmat)
nmat = nmat1 + nmat2
dmat = dmat1 + dmat2
fmat = nmat / dmat
```

```python
# Quantitative
ticks = range(0,10000,1000)
labs  = ["%.2f"%(x/10000.) for x in ticks]
fig, ax = subplots(figsize=(10,10))
cax = ax.imshow(fmat,origin='lower',cmap='Reds')
plt.xticks(ticks,labs)
plt.yticks(ticks,labs)
ax.set_xlabel("MAF 1 of SNP A (major), MAF 1 of SNP B (minor)")
ax.set_ylabel("MAF 2 of SNP A (major), MAF 2 of SNP B (minor)")
fig.colorbar(cax)
plt.grid(False)
plt.savefig("/Volumes/sivleyrm/pdbmap/temp/trippy_fst_plot.tif")

# Boolean matrix demonstrating what combinations of 
# allele frequencies over 2 SNPs can produce an Fst
# greater than 0.1
ticks = range(0,10000,1000)
labs  = ["%.2f"%(x/10000.) for x in ticks]
fig, ax = subplots(figsize=(10,10))
boolmat = fmat>0.1
cax = ax.imshow(boolmat,origin='lower',cmap='Reds')
plt.xticks(ticks,labs)
plt.yticks(ticks,labs)
ax.set_xlabel("MAF 1 of SNP A (major), MAF 1 of SNP B (minor)")
ax.set_ylabel("MAF 2 of SNP A (major), MAF 2 of SNP B (minor)")
fig.colorbar(cax)
plt.grid(False)
plt.savefig("/Volumes/sivleyrm/pdbmap/temp/trippy_fst-gt-01_plot.tif")
```

```python
def fst_plot(p1,p2,binsize,label):
    # What is the overall Fst of SNP B when SNP A is a singleton?
    af = np.arange(0,1000,dtype=np.float64)/1000
    af1,af2 = np.meshgrid(af,af)
    nmat  = nhat(af1,af2)
    dmat  = dhat(af1,af2)
    anhat = nhat(p1,p2)
    adhat = dhat(p1,p2)
    afst  = fst(p1,p2)

    print "Confirming that the %s Fst was calculated correctly:"%label.lower()
    print "%s allele frequency Pop1 => %g"%(label,0.00091575)
    print "%s allele frequency Pop2 => %g"%(label,0.0)
    print ""
    print "%s Nhat => %g"%(label,anhat)
    print "%s Dhat => %g"%(label,adhat)
    print "%s Fst  => %g"%(label,afst)
    print "Fst for %s SNP A: %2.6f/%2.6f => %2.6f"%(label.lower(),anhat,adhat,afst)

    ## What if we group a single ? with SNP B?
    ticks    = [f*1000 for f in af if not f*1000%100]
    labs     = [str(f) for f in af if not f*1000%100]
    fig,axes = subplots(2,3,figsize=(15,10))
    
    # Calculate the SNP B matrix
    ax   = axes[0][0]
    fmat = nmat/dmat
    im  = ax.imshow(fmat,origin='lower',cmap='bwr',vmin=-1,vmax=1)
    ax.autoscale(False)
    fvec = np.array([np.argmin(np.abs(f[range(i+1)]-afst)) for i,f in enumerate(fmat.T)])
    ax.plot(fvec,lw=3)
    fvec = np.array([np.argmin(np.abs(f[range(i,1000)]-afst))+i for i,f in enumerate(fmat.T)])
    ax.plot(fvec,lw=3)
    ax.set_title("Fst(SNP B,%s) Matrix"%label)
    ax.set_ylabel("Frequency of SNP B, Population 2")
    ax.set_xticks(ticks,labs)
    ax.set_yticks(ticks,labs)
    # Calculate the SNP B Matrix given that SNP A is a ?
    ax   = axes[0][1]
    fmat = (nmat + anhat) / (dmat + adhat)
    im  = ax.imshow(fmat,origin='lower',cmap='bwr',vmin=-1,vmax=1)
    ax.autoscale(False)
    fvec = np.array([np.argmin(np.abs(f[range(i+1)]-afst)) for i,f in enumerate(fmat.T)])
    ax.plot(fvec,lw=3)
    fvec = np.array([np.argmin(np.abs(f[range(i,1000)]-afst))+i for i,f in enumerate(fmat.T)])
    ax.plot(fvec,lw=3)
    ax.set_title("Fst(SNP B,%s) Matrix"%label)
    ax.set_ylabel("Frequency of SNP B, Population 2")
    ax.set_xticks(ticks,labs)
    ax.set_yticks(ticks,labs)
    # Calculate the difference between the SNP B matrix with and without the ?
    ax    = axes[0][2]
    fmatD = fmat - nmat/dmat
    im   = ax.imshow(fmatD,origin='lower',cmap='bwr',vmin=-1,vmax=1)
    ax.autoscale(False)
    fvec = np.array([np.argmin(np.abs(f[range(i+1)]-afst)) for i,f in enumerate(fmat.T)])
    ax.plot(fvec,lw=3)
    fvec = np.array([np.argmin(np.abs(f[range(i,1000)]-afst))+i for i,f in enumerate(fmat.T)])
    ax.plot(fvec,lw=3)
    ax.set_title("Fst(SNP B,%s) - Fst(SNP B)"%label)
    ax.set_xticks(ticks,labs)
    ax.set_yticks(ticks,labs)

    ## Now what if we group # ? with SNP B?
    # Calculate the SNP B matrix
    ax   = axes[1][0]
    fmat = nmat/dmat
    im  = ax.imshow(fmat,origin='lower',cmap='bwr',vmin=-1,vmax=1)
    ax.autoscale(False)
    fvec = np.array([np.argmin(np.abs(f[range(i+1)]-afst)) for i,f in enumerate(fmat.T)])
    ax.plot(fvec,lw=3)
    fvec = np.array([np.argmin(np.abs(f[range(i,1000)]-afst))+i for i,f in enumerate(fmat.T)])
    ax.plot(fvec,lw=3)
    ax.set_title("Fst(SNP B) Matrix")
    ax.set_xlabel("Frequency of SNP B, Population 1")
    ax.set_ylabel("Frequency of SNP B, Population 2")
    ax.set_xticks(ticks,labs)
    ax.set_yticks(ticks,labs)
    # Calculate the SNP B Matrix given that SNP A is a ?
    ax   = axes[1][1]
    fmat = (nmat + anhat*binsize) / (dmat + adhat*binsize)
    im  = ax.imshow(fmat,origin='lower',cmap='bwr',vmin=-1,vmax=1)
    ax.autoscale(False)
    fvec = np.array([np.argmin(np.abs(f[range(i+1)]-afst)) for i,f in enumerate(fmat.T)])
    ax.plot(fvec,lw=3)
    fvec = np.array([np.argmin(np.abs(f[range(i,1000)]-afst))+i for i,f in enumerate(fmat.T)])
    ax.plot(fvec,lw=3)
    ax.set_title("Fst(SNP B,%ss (%d)) Matrix"%(label,binsize))
    ax.set_xlabel("Frequency of SNP B, Population 1")
    ax.set_ylabel("Frequency of SNP B, Population 2")
    ax.set_xticks(ticks,labs)
    ax.set_yticks(ticks,labs)
    # Calculate the difference between the SNP B matrix with and without the ?
    ax    = axes[1][2]
    fmatD = fmat - nmat/dmat
    im    = ax.imshow(fmatD,origin='lower',cmap='bwr',vmin=-1,vmax=1)
    ax.autoscale(False)
    fvec = np.array([np.argmin(np.abs(f[range(i+1)]-afst)) for i,f in enumerate(fmat.T)])
    ax.plot(fvec,lw=3)
    fvec = np.array([np.argmin(np.abs(f[range(i,1000)]-afst))+i for i,f in enumerate(fmat.T)])
    ax.plot(fvec,lw=3)
    ax.set_title("Fst(SNP B,%ss (%d)) - Fst(SNP B)"%(label,binsize))
    ax.set_xlabel("Frequency of SNP B, Population 1")
    ax.set_xticks(ticks,labs)
    ax.set_yticks(ticks,labs)

    # Set the ticks and ticklabels for all axes
    plt.setp(axes, xticks=ticks, xticklabels=labs,
            yticks=ticks, yticklabels=labs)
    plt.tight_layout()
    cax = fig.add_axes([0.99, 0.1, 0.03, 0.8])
    fig.colorbar(im,cax=cax)

    plt.show()
```
