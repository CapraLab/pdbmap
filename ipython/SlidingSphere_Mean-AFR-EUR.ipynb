```python
# Load libraries
%matplotlib inline
import glob,gzip
import numpy as np
from numpy.random import randn
import random
import pandas as pd
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 500)
```

```python
# Load the results file into a pandas data frame
fnames = glob.glob('results/sliding_sphere_10/split/obs/bystruct/*.txt.gz')
rlist  = [[r.split('-')[0][-4:],r.split('-')[1].split('.')[0]] for r in fnames]
rdf    = pd.DataFrame(rlist,columns=['structid','biounit'])
dfs,dfsigs = [],[]
# cols =  ['structid','chain','seqid','afr_count','eur_count','afr_cdaf']
# cols += ['eur_cdaf','afr_eur_meanddaf','afr_eur_meanddaf_pval']
cols = [10,11,13,14,15,16,20,21,24,25,77,145]
for fname in fnames:
    # Load only the relevant columns
    df    = pd.read_table(gzip.open(fname,'rb'),delimiter='\t',header=0,usecols=cols)
    df.drop_duplicates(subset=['x','y','z'],inplace=True) # remove duplicate residues
    # Create one dataframe containing only the significant spheres
    dfsig = df[df['eur_afr_meanddaf_pval']<0.05]
    # Sample a maximum of 1,500 (99th percentile) observed spheres from the structure
    if len(df) > 1500:
        df    = df.ix[random.sample(df.index,1500)]
    dfs.append(df)
    dfsigs.append(dfsig)
# Review final df for correctness
# Create the final observation dataframe
df    = pd.concat(dfs)
df    = pd.merge(df,rdf,on='structid')
# Create the final significant dataframe
dfsig = pd.concat(dfsigs)
dfsig = pd.merge(dfsig,rdf,on='structid')
# Cleanup the temporary lists
del dfs,dfsigs
```

```python
# Convert the sort-columns to Z-scores
metric = 'eur_afr_meanddaf'
zmetric = (df[metric] - df[metric].mean())/df[metric].std(ddof=0)
# Maximum possible unique variants in sphere (sum of possibly overlapping counts)
varcnt = np.maximum(df['eur_count'],df['afr_count'])
print varcnt.shape
# Z-normalize and add to the dataframe
zvarcnt = (varcnt - varcnt.mean())/varcnt.std(ddof=0)
del varcnt
```

```python
plt.figure(figsize=(20,6))
x = zmetric * zvarcnt
x.sort()
plt.plot(x.unique(),color='blue')
x = zmetric + zvarcnt
x.sort()
# From this plot, I think that zmeteric*zvarcnt does a better job at distinguishably ranking the results
# It also more heavily weights spheres with large numbers of variants in the sphere
plt.plot(x.unique(),color='red')
```

```python
# Top scoring - EUR selected
metric = 'eur_afr_meanddaf'
pval   = 'eur_afr_meanddaf_pval'
zsorteur = zmetric * zvarcnt # larger positive values of zmetric indicate European selection
df['zsorteur'] = zsorteur

# dfc = df[df[pval]<0.05].sort([metric,'afr_count','eur_count'],ascending=[True,False,False])
dfc = df[df[pval]<0.05].sort('zsorteur',ascending=False)

top_20_structs,i,keys = dict(),0,[]
while len(top_20_structs) < 20:
    sid = dfc.iloc[i].structid
    if sid not in top_20_structs:
        top_20_structs[sid] = 1
        keys.append(sid)
    else:
        top_20_structs[sid] += 1
    i+= 1
# Structures ordered by the rank of their highest-ranking sphere
# Counts are for all spheres with a higher rank than the final sphere considered
for key in keys:
    print key,top_20_structs[key]

dfc = dfc.head(100)

print "%s\n%s"%(metric,'\n'.join(["%s:%d.%s: %2.4f, p≤%1.1e afr=%d eur=%d"%(sid,seqid,chain,val,pval,afr,eur) for _,sid,seqid,chain,pval,val,afr,eur in dfc.ix[:,["structid",'seqid','chain',pval,metric,'afr_count','eur_count']].itertuples()]))
```
