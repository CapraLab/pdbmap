```python
def secure_query(query,qvars=None,cursorclass='SSDictCursor'):
    """ Executes queries using safe practices """
    c = connect()
    filterwarnings('ignore', category = db.Warning)
    try:
      if qvars: c.execute(query,qvars)
      else:     c.execute(query)
    except Exception as e:
      msg  = "ERROR (PDBMapIO) Secure query failed; Exception: %s; "%str(e)
      msg += " Provided query: %s"%query
      msg += " Provided args: %s"%str(qvars)
      if "_last_executed" in dir(c):
        msg += " Executed Query: %s"%c._last_executed
      raise Exception(msg)
    resetwarnings()
    for row in c:
      yield row
```

```python
def load_structure(structid,biounit=0):
  global query
  q = secure_query(query,qvars=(structid,biounit),cursorclass='DictCursor')
  # Construct the Bio.PDB.Structure
  s = None
  for row in q:
    if not s:
        s = Bio.PDB.Structure.Structure(row['structid'])
        s.biounit    = row['biounit']
        s.resolution = row['resolution'] if row['resolution'] else None
        s.mpqs       = row['mpqs'] if row['mpqs'] else None
        s.method     = row['smethod'] if row['smethod'] else row['mmethod']
    m = row['model']
    if m not in s: s.add(Bio.PDB.Model.Model(m))
    m = s[m]
    c = row['chain']
    if c not in m: m.add(Bio.PDB.Chain.Chain(c))
    c = m[c]
    c.unp = row['unp']
    r = (' ',row['rid'], ' ')
    c.add(Bio.PDB.Residue.Residue(r,row['rescode'],' '))
    r = c[r]
    r.coords = (row['x'],row['y'],row['z'])
    print row['structid'],row['biounit'],row['model'],row['chain'],row['unp'],row['rid']
```
