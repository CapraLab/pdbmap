```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
@np.vectorize
def glf(t,A=1.,K=0.,B=0.2,v=0.05,Q=1.,C=1.):
      return A + (K-A)/((C+Q*np.exp(-B*t))**(1./v))
```

```python
x     = np.arange(0,50,0.5)
x_glf = glf(x)
```

```python
plt.figure(figsize=(20,7))
plt.xlabel("Distance",fontsize=25)
plt.ylabel("Weight",fontsize=25)
plt.ylim(0,1.01)
plt.plot(x,x_glf,lw=3,c="darkblue")
plt.show()
```
