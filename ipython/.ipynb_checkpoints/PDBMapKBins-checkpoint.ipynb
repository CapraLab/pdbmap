```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import itertools,random
%matplotlib inline
```

```python
with open("/Volumes/sivleyrm/pdbmap/data/fst/fst_analysis_data_nosingletons.txt",'rb') as fin:
    df = pd.read_csv(fin,sep='\t',header=0)
```

```python
df = df.dropna(subset=["name"]).reset_index()
print "Number of SNPs: %d"%len(df)
print "Maximum index:  %d"%df.index.max()
```

```python
def kbin(df,k):
    for kcomb in itertools.combinations(np.random.permutation(df.index),k):
        # Sample from 5% of the combination space
        if random.random() < 0.05:
            yield df.iloc[list(kcomb),:]
```

```python
def fst(df,bins):
    bins[int(np.around(df["allpop_Nhat"].sum()/df["allpop_Dhat"].sum(),decimals=2)*100)] += 1
    return
```

```python
bins2 = np.zeros(100)
try:
    np.array([fst(dfk,bins2) for dfk in kbin(df,2)])
except:
    pass
```

```python
# Normalize to a % of total nsSNPs
print "Total 2-SNP bins measured: %d"%bins2.sum()
kbins = bins2 / bins2.sum()
fig,axes = plt.subplots(1,2,figsize=(20,10))
ax = axes[0]
ax.bar(range(100),kbins)
# sns.barplot(bins,ax=ax)
ax.set_xticks(np.arange(0,100,5))
ax.set_xticklabels(np.arange(0,1,0.05))
ax.set_title("Fst Emprical Distribution")
ax.set_xlabel("2-SNP Bin Fst")
ax.set_ylabel("% of Observations")
ax = axes[1]
ax.bar(range(9,100),kbins[9:])
ax.set_xticks(np.arange(0,100,5))
ax.set_xticklabels(np.arange(0,1,0.05))
ax.set_title("Fst>0.01 Empirical Distribution")
ax.set_xlabel("2-SNP Bin Fst")
ax.set_ylabel("% of Observations")
plt.show()
```
