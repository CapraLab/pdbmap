```python
# Load libraries
%matplotlib inline
import glob,gzip,os
import numpy as np
from numpy.random import randn
from itertools import combinations
import random
import pandas as pd
from scipy import stats
from scipy.spatial import KDTree
from sklearn.metrics import confusion_matrix
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 500)
```

```python
# Read the mutation disease classifications
dclass = {'WT':-1}
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/1kg3.txt','rb') as fin:
  dclass.update(dict((m.strip(),0) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/unclassified.txt','rb') as fin:
  dclass.update(dict((m.strip(),1) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/dyskeratosis.txt','rb') as fin:
  dclass.update(dict((m.strip(),2) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/published_pathogenic.txt','rb') as fin:
  dclass.update(dict((m.strip(),3) for m in fin))
dcolor = {-2:"orange",-1:"black",0:"darkgray",1:"green",2:"blue",3:"red"}
```

```python
seqident = {
"ENSP00000353332_1":20.00,
"ENSP00000353332_2":30.00,
"ENSP00000353332_3":29.00,
"ENSP00000353332_4":16.00,
"ENSP00000353332_5":22.00,
"ENSP00000353332_6":39.00,
"ENSP00000359020_1":13.00,
"ENSP00000359020_2":16.00,
"ENSP00000359020_3":22.00,
"ENSP00000359020_4":39.00,
"ENSP00000425576_1":33.00,
"ENSP00000425576_2":22.00,
"c2vsfA":21.00,
"c2vl7A":22.00,
"c3crw1":25.00,
"c2xgjA":27.00,
"c1c4oA":17.00,
"c1oywA":20.00,
"c3tmiA":15.00,
"c4ddvA":21.00,
"c3l9oA":24.00,
"c2v1xB":24.00,
"c4f92B":22.00,
"c4bujA":25.00,
"c2va8A":22.00,
"c4cdgA":19.00,
"c4bgdA":20.00,
"c1gm5A":20.00,
"c4a4zA":23.00,
"c1wp9D":17.00,
"c1gl9B":17.00,
"c4fqnC":19.00,
"c1xtkA":21.00
}
```

```python
# resdir  = "/Volumes/sivleyrm/pdbmap/results/Q9NZ71_mutagenesis_20150603"
resdir = "/Volumes/sivleyrm/pdbmap/results/RTEL1_Mutagenesis"
names   = ["Mutation","Unrelaxed_Score","Unrelaxed_Rep","Min_Score","Mean_Score","Std_Score","Min_MT_RMSD","Mean_MT_RMSD","Std_MT_RMSD","Relaxed_Rep","rWT_RMSD"]
results = []
# bad_structs = ["ENSP00000353332_1","ENSP00000353332_4","ENSP00000359020_1","ENSP00000359035_2"]
bad_structs = []
# valid_structs = ["ENSP00000353332_1","ENSP00000359035_2","ENSP00000353332_2","ENSP00000359035_1"]
valid_structs = []
for f in glob.glob("%s/*.results"%resdir):
  # Record the structid and biounit
  fname = os.path.basename(f)
  if fname in ["score.results","wt-rmsd.results","rwt-rmsd.results","delta.results"]: continue
  fsplit = fname.split('.')[0].split('_')
  if len(fsplit)==1:
    structid,biounit = fsplit[0],0
  elif len(fsplit)==2 and not fsplit[1]:
    structid,biounit   = fsplit[0],0
  elif fsplit[-1] == "msa":
    structid,biounit = '_'.join(fsplit),0
  else:
    structid,biounit = '_'.join(fsplit[:2]),fsplit[2]
  # Read the Wild-Type and Relaxed Wild-Type scores for this structure
  with open(f,'rb') as fin:
    wt  = float(fin.readline().split(':')[-1].strip())
    rwt = float(fin.readline().split(':')[-1].strip())
  # Extract results for this structure
  try:
    df = pd.read_csv(f,sep='\t',skiprows=3,names=names) # first three rows are comments and header
  except:
    print "\nException: %s\n"%f
    raise
  # Add structid and biounit columns
#   sid = "%s_%s"%(structid,biounit)
  sid = structid
  # Reduce structures to ModBase and Phyre models w/ sequence identity information
  if sid not in seqident: continue
  valid_structs.append(sid)
#   if sid not in valid_structs: continue
  df["Structid"] = [sid]*len(df)
  df["nres"]     = [nres[structid]]*len(df)
  df["seqident"] = [seqident[structid]]*len(df)
  df["Quality"]  = [int(sid not in bad_structs)]*len(df)
  # Reduce to minimum scores and RSMDs
#   df = df[["Mutation","Structid","Quality","Unrelaxed_Score","Min_Score","Min_MT_RMSD","rWT_RMSD"]]
  # Calculate ddG
  # Multiply nres to un-normalize the dREU and rdREU values
  df["dREU"]  = (df["Unrelaxed_Score"] - rwt)*df["nres"]
  df["rdREU"] = (df["Min_Score"]       - rwt)*df["nres"]
  # Record the WT and rWT scores
#   df = df.append({"Mutation":"WT","Unrelaxed_Score":rwt,"Structid":sid},ignore_index=True)
  # Index by structid, biounit, and mutation
  df.set_index(["Structid","Mutation"],inplace=True)
  results.append(df)
# Merge results
df = pd.concat(results)
df = df.reset_index()
df = df.drop_duplicates(subset=["Structid","Mutation"],take_last=True).dropna(subset=["Structid","Mutation"])
nummut = len(df)
df = df[df["Mutation"].isin(dclass)] # Filter out incorrectly mapped SNPs
print "%d invalid mutations removed"%(nummut - len(df))
# Group structures into reliable and unreliable, and sort by decreasing alignment quality in each subgroup
good_structs = [sid for sid in df["Structid"].drop_duplicates() if sid not in bad_structs]
bad_structs.sort(key=lambda x: seqident[x],reverse=True)
good_structs.sort(key=lambda x: seqident[x],reverse=True)
order = bad_structs + good_structs
order = [o for o in order if o in valid_structs]
thresh30 = np.argmin(np.array([df.ix[df["Structid"]==o,"seqident"].values[0] for o in order])>=30)+.5
thresh20 = np.argmin(np.array([df.ix[df["Structid"]==o,"seqident"].values[0] for o in order])>=20)+.5
# print "Sizes: \n",'\n'.join([str((o,len(df))) for o in order])
# print "SNPs: \n",'\n'.join([str((o,len(df[df["Structid"]==o]))) for o in order])
print "Number of Mutation+Structure Pairs Before Pruning: %d"%len(df)
print "Number of mutations in reliable structures:        %d"%len(df[df["Quality"]==1])
print "Number of mutations in unreliable structures:      %d"%len(df[df["Quality"]==0])
```

```python
# Pivot by structure
df = df.reindex_axis(sorted(df.columns),axis=1)

cls = [dclass[d] if d[0]!='r' else dclass[d[1:]] for d in df["Mutation"]]
labels = ["1000 Genomes","Unclassified","Dyskeratosis","IPF","WT"]
df["DiseaseClass"] = [labels[c] for c in cls]
```

```python
def def_attr(df,sid,attr,avg=False):
  if avg:
    # Filter to mutations in this structure (may contain additional structures)
    dft = df.groupby("Mutation").filter(lambda x: sid in x["Structid"].values)
  else:
    # Filter to only mutations in this structure
    dft = df[df["Structid"]==sid]
  # Calculate mean over all included structures
  dft = dft.groupby("Mutation")[attr].mean()
  # Initialize generator
  attrs = dft.iteritems()
  msg  = "attribute: %s\n"%attr
  msg += "match mode: 1-to-1\n"
  msg += "recipient: residues\n"
  for pos,val in attrs:
    msg += "\t:%s\t%s\n"%(pos[2:-1],val)
  return msg
```
