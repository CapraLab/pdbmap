# Variant Burden in 1000 Genomes by Ancestral Populations

Determine the variant burden within each protein structure in PDBMap for each of the four 
ancestral populations in 1000 genomes, Phase III, Version 3.

Once Version 5 of 1000 Genomes has been loaded into PDBMap, we should recompute these values.

```python
# Ensure that we are in the main pdbmap directory and not a subdirectory or elsewhere
%cd ../../pdbmap
```

```python
# Load the master file into a pandas data frame
import pandas as pd
master_fname = 'temp/pdbmap_v10_1kg_master.txt'
df = pd.read_table(master_fname,delimiter='\t')
```

```python
# Determine how many SNPs were observed in each population
print "Number of residues mapped to observed SNPs (ALL): %d"%len(df)
print "Number of residues mapped to observed SNPs (AMR): %d"%len(df[df.amr_af>0])
print "Number of residues mapped to observed SNPs (ASN): %d"%len(df[df.asn_af>0])
print "Number of residues mapped to observed SNPs (AFR): %d"%len(df[df.afr_af>0])
print "Number of residues mapped to observed SNPs (EUR): %d"%len(df[df.eur_af>0])
```

```python
# Initialize imports
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_dist(df,pop):
    """ Plots the number of variant residues and unique variants per structure
        stratified by the specified 1000 genomes population """
    tdf = df[df["%s_af"%pop.lower()]>0].groupby(('structid',)).size()
    fig,ax = plt.subplots(2,2,figsize=(8,8))
    # Residues mapped to SNPs
    ax[0][0].hist(tdf[tdf<10])
    ax[0][1].hist(tdf[(tdf>9) & (tdf<100)])
    print "Number of structures with <10 variant residues  (%s): %d"%(pop.upper(),np.sum(tdf<10))
    print "Number of structures with >=10 variant residues (%s): %d"%(pop.upper(),np.sum(tdf>9))
    # Unique SNPs
    tdf = df[(df["%s_af"%pop.lower()]>0)].groupby(('structid',))['name'].nunique()
    ax[1][0].hist(tdf[tdf<10])
    ax[1][1].hist(tdf[(tdf>9) & (tdf<100)])
    print "Number of structures with <10 SNPs  (%s): %d"%(pop.upper(),np.sum(tdf<10))
    print "Number of structures with >=10 SNPs (%s): %d"%(pop.upper(),np.sum(tdf>9))
```

```python
plot_dist(df,'amr')
```

```python
plot_dist(df,'asn')
```

```python
plot_dist(df,'afr')
```

```python
plot_dist(df,'eur')
```
