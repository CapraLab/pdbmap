```python
# Tuple w/ (name,dREU rank)
ngl1_tup = [(ngl1[i],i) for i in range(len(ngl1))]
ngl2_tup = [(ngl2[i],i) for i in range(len(ngl2))]
ngl3_tup = [(ngl3[i],i) for i in range(len(ngl3))]
```

```python
# Sort by name, extract rank
ngl1R = [x[1] for x in sorted(ngl1_tup)]
ngl2R = [x[1] for x in sorted(ngl2_tup)]
ngl3R = [x[1] for x in sorted(ngl3_tup)]
```

```python
# Calculate rank correlation between each model
print "ngl1,ngl2: %.3f (p=%.3f)"%spearmanr(ngl1R,ngl2R)
print "ngl2,ngl3: %.3f (p=%.3f):"%spearmanr(ngl2R,ngl3R)
print "ngl1,ngl3: %.3f (p=%.3f):"%spearmanr(ngl1R,ngl3R)
```

```python
# Compare rank correlation 2D
fig,ax = plt.subplots(1,3,figsize=(30,10))
ax[0].scatter(ngl1R,ngl2R,s=50)
ax[0].set_xlabel("ngl1 rank")
ax[0].set_ylabel("ngl2 rank")
ax[1].scatter(ngl2R,ngl3R,s=50)
ax[1].set_xlabel("ngl2 rank")
ax[1].set_ylabel("ngl3 rank")
ax[2].scatter(ngl1R,ngl3R,s=50)
ax[2].set_xlabel("ngl1 rank")
ax[2].set_ylabel("ngl3 rank")
plt.tight_layout()
plt.show()
```

```python
# Compare rank correlation 3D
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(ngl1R,ngl2R,ngl3R)
ax.set_xlabel("ngl1 rank")
ax.set_ylabel("ngl2 rank")
ax.set_zlabel("ngl3 rank")
plt.show()
```
