```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
%matplotlib inline
import numpy as np
from collections import defaultdict
```

```python
# Domain definitions from the ERJ-00401-2015_supplemental alignment
hd1_a = (5,140)
fes   = (140,220)
hd1_b = (220,265)
arch  = (265,460)
hd1_c = (460,505)
hd2   = (505,760)
hrm1  = (890,970)
hrm2  = (1060,1135)
# Where did they get the sequence ranges for the harmonin, PIP, and RING domains?
```

```python
neut = [int(l.strip()[1:-1]) for l in open('/Volumes/sivleyrm/pdbmap/data/udn/rtel1/benign.txt','rb')]
# There are no benign variants specified. Add the 1000 Genomes variants.
neut += [int(l.strip()[1:-1]) for l in open('/Volumes/sivleyrm/pdbmap/data/udn/rtel1/1kg3.txt','rb')]
path = [int(l.strip()[1:-1]) for l in open('/Volumes/sivleyrm/pdbmap/data/udn/rtel1/pathogenic.txt','rb')]
vus  = [int(l.strip()[1:-1]) for l in open('/Volumes/sivleyrm/pdbmap/data/udn/rtel1/vus.txt','rb')]
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig = plt.figure(figsize=(45,2.25),dpi=300)
ax  = fig.add_subplot(111)
ax.grid(False)
ax.set_xlim(-1,1221)
ax.set_ylim(-0.1,1.1)
ax.set_xticks(range(1,1219,100))
ax.set_yticks([])
ax.tick_params(axis='both', which='major', labelsize=30)
ax.set_axis_bgcolor('white')
ax.axhline(1.,c="k",lw=2)
ax.axhline(0.,c="k",lw=2)
ax.axvline(1.,c="k",lw=2)
ax.axvline(1219.,c="k",lw=2)

# Plot the known pathogenic missense variants
bpath = defaultdict(lambda:0,{})
binId = lambda x: int(x/10.)*10
for p in path:
    bpath[binId(p)] += 1
for b,p in bpath.iteritems():
    for i in range(p)[::-1]:
        ax.arrow(b,1.,0,i*-0.2-.01,lw=1.5,head_width=10, head_length=0.2, fc='darkred', ec='k')
    
# for p in path:
#     ax.arrow(p,1.,0,-0.2,lw=1.5,head_width=10, head_length=0.2, fc='orangered', ec='white')

# Plot the neutral missense variants
bneut = defaultdict(lambda:0,{})
binId = lambda x: int(x/10.)*10
for n in neut:
    bneut[binId(n)] += 1
for b,n in bneut.iteritems():
    for i in range(n)[::-1]:
        ax.arrow(b,0.,0,i*0.2+0.01,lw=1.5,head_width=10, head_length=0.2, fc='blue', ec='k')

# for n in neut:
#     ax.arrow(n,0.,0,0.2,lw=1.5,head_width=10, head_length=0.2, fc='steelblue',ec='white')

# Plot the novel VUS

bvus = defaultdict(lambda:0,{})
binId = lambda x: int(x/10.)*10
for v in vus:
    bvus[binId(v)] += 1
for b,v in bvus.iteritems():
    for i in range(v)[::-1]:
        ax.arrow(b,1.,0,i*-0.2-0.01,lw=1.5,head_width=10, head_length=0.2, fc='gold', ec='k')

        
dlabels = "\n%s FeS %s Arch %s Helicase II %s Harmonin I %s Harmonin II %s \n\n %s Helicase I %s"
ax.set_xlabel(dlabels%(' '*15,' '*10,' '*14,' '*13,' '*1,' '*5,' '*0,' '*59),fontsize=40,fontweight="bold")

# for v in vus:
#     ax.arrow(v,1.,0,-0.2,lw=1.5,head_width=10, head_length=0.2, fc='gold', ec='white')

domains = [
    patches.Rectangle(
        (hd1_a[0], 0.), hd1_a[1]-hd1_a[0], 1.,
        facecolor="mediumslateblue"
    ),
    patches.Rectangle(
        (fes[0], 0.), fes[1]-fes[0], 1.,
        facecolor="sandybrown"
    ),
    patches.Rectangle(
        (hd1_b[0], 0.), hd1_b[1]-hd1_b[0], 1.,
        facecolor="mediumslateblue"
    ),
    patches.Rectangle(
        (arch[0], 0.), arch[1]-arch[0], 1.,
        facecolor="beige"
    ),
    patches.Rectangle(
        (hd1_c[0], 0.), hd1_c[1]-hd1_c[0], 1.,
        facecolor="mediumslateblue"
    ),
    patches.Rectangle(
        (hd2[0], 0.), hd2[1]-hd2[0], 1.,
        facecolor="lightsteelblue"
    ),
    patches.Rectangle(
        (hrm1[0], 0.), hrm1[1]-hrm1[0], 1.,
        facecolor="violet"
    ),
    patches.Rectangle(
        (hrm2[0], 0.), hrm2[1]-hrm2[0], 1.,
        facecolor="violet"
    ),
]
for d in domains:
    ax.add_patch(d)
    
plt.savefig("/Users/mikesivley/Google Drive/Vanderbilt/Research/UDN/RTEL1/Final_Analysis/Manuscript/Figures/RTEL1_SNV_Mapping_Sequence.png",bbox_inches='tight')
plt.show()
```
