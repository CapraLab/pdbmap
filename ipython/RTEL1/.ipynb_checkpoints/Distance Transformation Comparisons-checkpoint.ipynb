```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
@np.vectorize
def glf(d,A=1.,K=0.,B=0.2,v=0.05,Q=1.,C=1.):
      return A + (K-A)/((C+Q*np.exp(-B*d))**(1./v))
```

```python
@np.vectorize
def nw(d,lb=4.,ub=11.4):
    if d <= lb:
        return 1.
    elif d >= ub:
        return 0.
    else:
        return 0.5*(np.cos(np.pi*(d-lb)/(ub-lb))+1)
```

```python
x     = np.arange(0,50,0.01)
x_glf = glf(x)
x_nw  = nw(x,lb=4.,ub=16.)
```

```python
plt.figure(figsize=(20,7))
plt.xlabel("Distance",fontsize=25)
plt.xlim(0,45.01)
plt.xticks(range(0,50,5),fontsize=15)
plt.yticks(np.arange(0,1.1,0.2),fontsize=15)
plt.ylabel("Weight",fontsize=25)
plt.ylim(0,1.01)
plt.plot(x,x_glf,lw=3,c="black",ls='--',label='glf')
# plt.plot(x,nw(x,lb=4.,ub=11.4),lw=3,c="darkred",label='nw(4,11.4)')
# plt.plot(x,nw(x,lb=4.,ub=16.),lw=3,label='nw(4,16)')
# plt.plot(x,nw(x,lb=4.,ub=24.),lw=3,label='nw(4,24)')
# plt.plot(x,nw(x,lb=8.,ub=16.),lw=3,label='nw(8,16)')
plt.plot(x,nw(x,lb=8.,ub=24.),lw=3,label='nw(8,24)')
plt.plot(x,nw(x,lb=6.,ub=28.),lw=3,label='nw(6,28)') # closest approximation to glf
plt.legend(loc="upper right",fontsize=20)
plt.show()
```

```python
# Final Parameterization
plt.figure(figsize=(20,7))
plt.xlabel("Distance",fontsize=25)
plt.xlim(0,32,.001)
plt.xticks(range(0,32,8),fontsize=15)
plt.yticks(np.arange(0,1.1,0.2),fontsize=15)
plt.ylabel("Neighbor Weight",fontsize=25)
plt.ylim(-0.01,1.01)
plt.plot(x,nw(x,lb=8.,ub=24.),lw=4,c="k",label='Neighbor Weight(LB=8, UB=24)')
plt.legend(loc="upper right",fontsize=20)
plt.savefig("/Users/mikesivley/Google Drive/Vanderbilt/Research/UDN/RTEL1/Final_Analysis/Manuscript/Figures/neighbor_weight_fxn.png",dpi=300)
plt.show()
```
