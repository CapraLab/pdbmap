```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy.stats import spearmanr,pearsonr
```

```python
# Load the results from the Fan et al colocalization analysis
df = pd.read_csv("/Volumes/sivleyrm/pdbmap/results/RTEL1_Colocalization_2016-08-24/RTEL1_N-terminus_Fan_data.txt",
                 header=0,sep='\t')
df["mutation"]    = df["ref"].str.cat(df["pos"].astype(str)).str.cat(df["alt"])
df["ATP%WT"]      = df["ATPase %wt"].astype(np.float64)
df["Helicase%WT"] = df["Helicase %wt"].astype(np.float64)
```

```python
# Plot the ATPase correlation
tdf = df.dropna(subset=["ATP%WT"])
print "Pearson  Correlation: % .2f (p=%.1g)"%pearsonr(tdf["ATP%WT"],tdf["pathprox"])
print "Spearman Correlation: % .2f (p=%.1g)"%spearmanr(tdf["ATP%WT"],tdf["pathprox"])
plt.figure(figsize=(20,7))
sns.regplot(df["pathprox"],df["ATP%WT"],scatter_kws={"s":50,"alpha":0.5,"c":"darkred"})
plt.ylabel("ATPase Activity (%WT)",fontsize=18)
plt.xlabel("Pathogenic Proximity",fontsize=18)
plt.axhline(0.,ls="--",color="darkgrey",lw=3)
plt.axvline(0.,ls="--",color="darkgrey",lw=3)
plt.tight_layout()
plt.show()
```

```python
# Plot the Helicase correlation
tdf = df.dropna(subset=["Helicase%WT"])
print "Pearson  Correlation: % .2f (p=%.1g)"%pearsonr(tdf["Helicase%WT"],tdf["pathprox"])
print "Spearman Correlation: % .2f (p=%.1g)"%spearmanr(tdf["Helicase%WT"],tdf["pathprox"])
plt.figure(figsize=(20,7))
sns.regplot(df["pathprox"],df["Helicase%WT"],scatter_kws={"s":50,"alpha":0.5,"c":"darkred"})
plt.ylabel("Helicase Activity (%WT)",fontsize=18)
plt.xlabel("Pathogenic Proximity",fontsize=18)
plt.axhline(0.,ls="--",color="darkgrey",lw=3)
plt.axvline(0.,ls="--",color="darkgrey",lw=3)
plt.tight_layout()
plt.show()
```

```python
# Plot the ssDNA KD correlation
tdf = df.dropna(subset=["KD%WT"])
df = tdf[tdf["pos"]!=679]
print "Pearson  Correlation: % .2f (p=%.1g)"%pearsonr(tdf["KD%WT"],tdf["pathprox"])
print "Spearman Correlation: % .2f (p=%.1g)"%spearmanr(tdf["KD%WT"],tdf["pathprox"])
print "Not shown: residue 679 with 463-fold reduction in ssDNA-binding."
plt.figure(figsize=(20,7))
plt.title("Correlation of PathProx with Kuper et al ssDNA-Binding (KD) %WT Results",fontsize=25)
sns.regplot(df["pathprox"],df["KD%WT"],scatter_kws={"s":50,"alpha":0.5,"c":"darkred"})
plt.ylabel("ssDNA-Binding Affinity (KD %WT)",fontsize=18)
plt.xlabel("Pathogenic Proximity",fontsize=18)
plt.axhline(0.,ls="--",color="darkgrey",lw=3)
plt.axvline(0.,ls="--",color="darkgrey",lw=3)
plt.tight_layout()
plt.show()
```

```python
# Combine results
# Load the results from the Fan et al colocalization analysis
df1 = pd.read_csv("/Volumes/sivleyrm/pdbmap/results/RTEL1_Colocalization_2016-08-24/RTEL1_N-terminus_Fan_data.txt",
                 header=0,sep='\t')
df2 = pd.read_csv("/Volumes/sivleyrm/pdbmap/results/RTEL1_Colocalization_2016-08-24/RTEL1_N-terminus_Kuper_data.txt",
                 header=0,sep='\t')
df = df1.append(df2)
df["mutation"] = df["ref"].str.cat(df["pos"].astype(str)).str.cat(df["alt"])
df["ATP%WT"]   = df["ATPase %wt"].astype(np.float64)
df["Helicase%WT"] = df["Helicase %wt"].astype(np.float64)
```
