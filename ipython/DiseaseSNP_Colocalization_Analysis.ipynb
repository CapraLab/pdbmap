```python
# Load libraries
%matplotlib inline
import glob,gzip,os
import numpy as np
from numpy.random import randn
from itertools import combinations
import random
import pandas as pd
from scipy import stats
from scipy.spatial import KDTree
from sklearn.metrics import confusion_matrix
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 500)
```

```python
# Read the mutation disease classifications
rtel1_dclass = {}
saxpd_dclass = {}
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/1kg3.txt','rb') as fin:
  rtel1_dclass.update(dict((m.strip()[1:],0) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/candidate.txt','rb') as fin:
  rtel1_dclass.update(dict((m.strip()[1:],3) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/telomere_syndrome.txt','rb') as fin:
  rtel1_dclass.update(dict((m.strip()[1:],1) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/pulmonary_fibrosis.txt','rb') as fin:
  rtel1_dclass.update(dict((m.strip()[1:],2) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/xp_snps.txt','rb') as fin:
  saxpd_dclass.update(dict((m.strip()[1:],4) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/xpcs_snps.txt','rb') as fin:
  saxpd_dclass.update(dict((m.strip()[1:],5) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/ttd_snps.txt','rb') as fin:
  saxpd_dclass.update(dict((m.strip()[1:],6) for m in fin))
dcolor = {0:"darkgray",1:"green",2:"blue",3:"red",4:"orange",5:"yellow",6:"purple"}
rtel1_pos2snp = dict((int(key[1:-1]),key) for key in rtel1_dclass.keys())
saxpd_pos2snp = dict((int(key[1:-1]),key) for key in saxpd_dclass.keys())
```

```python
rtel1_coord_file = "/Volumes/sivleyrm/pdbmap/temp/rtel1/RTEL1_Extended.coords"
saxpd_coord_file = "/Volumes/sivleyrm/pdbmap/temp/rtel1/SaXPD.coords"
```

```python
def snp_coords(coord_file):
    with open(coord_file,'rb') as fin:
        rows = [line.split()[3:8] for line in fin if line[:4]=="ATOM"]
        try:
            int(rows[0][1])
        except:
            with open(coord_file,'rb') as fin:
                rows = [line.split()[3:9] for line in fin if line[:4]=="ATOM"]
        rtel1_coords = np.array(rows)
        if rtel1_coords.shape[1] == 5:
            df = pd.DataFrame(rtel1_coords,columns=["res","seqid","x","y","z"])
            df["chain"] = "A"
        else:
            df = pd.DataFrame(rtel1_coords,columns=["res","chain","seqid","x","y","z"])
    df[["seqid"]] = df[["seqid"]].astype(np.int)
    df[["x","y","z"]] = df[["x","y","z"]].astype(np.float64)
    coords = np.array(zip(df.groupby(["chain","seqid"])["x"].mean(),df.groupby(["chain","seqid"])["y"].mean(),df.groupby(["chain","seqid"])["z"].mean()))
    coord_df = df[["res","chain","seqid"]].drop_duplicates()
    coord_df["x"] = coords[:,0]
    coord_df["y"] = coords[:,1]
    coord_df["z"] = coords[:,2]
    return coord_df
rtel1_coords = snp_coords(rtel1_coord_file)
saxpd_coords = snp_coords(saxpd_coord_file)

# print "RTEL1:"
# print rtel1_coords
# print "\nSaXPD:"
# print saxpd_coords
```

```python
# print dclass
cls2code = {"1000 Genomes":0,"Telomere Syndrome":1,"Pulmonary Fibrosis":2,"Candidate":3,"XP":4,"XP/CS":5,"TDD":6}
rtel1_coords["dcode"] = rtel1_coords["seqid"].apply(lambda x: rtel1_dclass[rtel1_pos2snp[x]])
saxpd_coords["dcode"] = saxpd_coords["seqid"].apply(lambda x: saxpd_dclass[saxpd_pos2snp[x]])
coords = pd.concat((rtel1_coords,saxpd_coords))
```

```python
# Predict disease status with a kNN approach
def pathprox(dfq,dcodes=[]):
    k = len(dfq)
    dfq["Pathogenic"] = np.array([1 if d in dcodes else -1 for d in dfq["dcode"]]) # Set which dcodes to use
    dfq.ix[dfq["dcode"]==0,"Pathogenic"] = 0 # Set the neutral variants
    num_neut = (dfq["Pathogenic"]==0).sum() 
    num_path = (dfq["Pathogenic"]==1).sum()
    pathprox = []
    for i,row in dfq.iterrows():
        # Calculate a "neutral score"
        kdt = KDTree(dfq.ix[dfq["Pathogenic"]==0,["x","y","z"]])
        dist,snps = kdt.query(row[["x","y","z"]],num_neut)
        neutral_score = np.sum(1./dist[dist>0])/num_neut
        # Calculate a "pathogenic score"
        kdt = KDTree(dfq.ix[dfq["Pathogenic"]==1,["x","y","z"]])
        dist,snps = kdt.query(row[["x","y","z"]],num_path)
        pathogenic_score = np.sum(1./dist[dist>0])/num_path
        pathprox.append(pathogenic_score/neutral_score)
    return np.array(pathprox)
```

```python
all_coords = coords.copy()
coords["Tel_Path"]   = pathprox(coords,[1])
coords["IPF_Path"]   = pathprox(coords,[2])
coords["RTEL1_Path"] = pathprox(coords,[1,2])
coords["XPCS_Path"]  = pathprox(coords,[4,5])
coords["TDD_Path"]   = pathprox(coords,[6])
coords["SaXPD_Path"] = pathprox(coords,[4,5,6])
coords["All_Path"]   = pathprox(coords,[1,2,4,5,6])
print coords[["res","seqid","dcode","IPF_Path","Tel_Path","RTEL1_Path","XPCS_Path","TDD_Path","SaXPD_Path","All_Path"]].sort("All_Path",ascending=False)
coords.to_csv("/Users/mikesivley/Dropbox/Vanderbilt/Research/UDN/RTEL1/RTEL1_SaXPD_Pathogenicity_Predictions.txt",sep="\t")
```

```python
def path_roc(dfq,dcodes=[0,1],pathcodes=[2],predcol="All_Path",label="",color="red"):
    from sklearn.metrics import roc_curve,auc
    jitter = np.random.normal(0,0.02,1)[0]
    dfq = dfq.sort(predcol,ascending=False)
    dfq["Pathogenic"] = np.array([1 if d in pathcodes else -1 for d in dfq["dcode"]])
    dfq.ix[dfq["dcode"]==0,"Pathogenic"] = 0 # Set to netrual variants
    dfq = dfq[dfq["Pathogenic"]>-1]
    fpr,tpr,thresh = roc_curve(dfq.ix[dfq["dcode"].isin(dcodes),"Pathogenic"],dfq.ix[dfq["dcode"].isin(dcodes),predcol])
    roc_auc = auc(fpr,tpr)
    fpr += jitter
    tpr += jitter
    plt.plot(fpr-0.01,tpr+0.01,label='%s (AUC = %0.2f)'%(label,roc_auc),color=color)
```

```python
# Plot the ROC for predicting RTEL1 IPF SNPs from different pathogenic backgrounds
plt.figure(figsize=(10,10))
plt.title('Predicting IPF nsSNPs from Different Pathogenic SNP Sets')
# ROC with Pulmonary Fibrosis background
path_roc(coords,dcodes=[0,2],pathcodes=[2],predcol="IPF_Path",label="IPF from IPF",color="red")
# ROC with Telomere Syndrome background
path_roc(coords,dcodes=[0,2],pathcodes=[2],predcol="Tel_Path",label="IPF from Telomere Syndrome",color="orange")
# ROC with Pulmonary Fibrosis + Telomere Syndrome Background
path_roc(coords,dcodes=[0,2],pathcodes=[2],predcol="RTEL1_Path",label="IPF from RTEL1",color="yellow")
# ROC with XP + XP/CS background
path_roc(coords,dcodes=[0,2],pathcodes=[2],predcol="XPCS_Path",label="IPF from XP/CS",color="green")
# ROC with TDD background
path_roc(coords,dcodes=[0,2],pathcodes=[2],predcol="TDD_Path",label="IPF from TDD",color="blue")
# ROC with XP + XP/CS + TDD background
path_roc(coords,dcodes=[0,2],pathcodes=[2],predcol="SaXPD_Path",label="IPF from SaXPD",color="purple")
# ROC with All known pathogenic SNPs as background
path_roc(coords,dcodes=[0,2],pathcodes=[2],predcol="All_Path",label="IPF from RTEL1 + SaXPD",color="black")

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```
