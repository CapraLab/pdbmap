```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import qvalue
```

```python
df = pd.read_csv("/Volumes/sivleyrm/pdbmap/results/specialK_analysis_2016-05-17/cosmic/recurrence_K_summary.txt",delimiter='\t')
```

```python
plt.figure(figsize=(20,7))

# Plot the unadjusted p-value histogram
p = df["Kp"].values
p = p[~np.isnan(p)]
np.savetxt("/Users/mikesivley/Downloads/pvalues.1.txt",p,delimiter='\t')
plt.hist(p,bins=100,normed=True,color='mediumblue',alpha=0.5)

# Calculate and plot the unadjusted cublic spline
q,lam,pi = qvalue.estimate(p[~np.isnan(p)])
plt.plot(lam,pi,lw=5,c="darkred",ls='dashdot',label="un-adjusted")

# Calculate and plot the p-value-truncated cubic spline
p = p[p<1.0]
q,lam,pi = qvalue.estimate(p[~np.isnan(p)])
plt.plot(lam,pi,lw=5,c="darkorange",ls='dashed',label="truncated")

# Plot details
plt.title("Lambda by Pi - Cubic Spline")
plt.legend(loc="upper center",fontsize=18)
plt.show()
```
