```python
def prep_split(dfx,weighted=False):
    df  = dfx.copy()
    df['fdr']  = 'All (N=%4d)'%len(df)
    if weighted:
        dft = dfx[dfx["wKq"]<0.1].copy()
    else:
        dft = dfx[dfx["Kq"]<0.1].copy()
    dft['fdr'] = "{:<20}".format('FDR < 0.1 (N=%d)'%len(dft))
    if weighted:
        df = pd.concat((df,dft)).dropna(subset=["wKz"])
    else:
        df = pd.concat((df,dft)).dropna(subset=["Kz"])
    df['dummy'] = ''
    return df
```

```python
def plot_dist(df,title,ax,pal,weighted=False,strip=False):
    if weighted:
        split = (df["wKq"]<0.1).sum()>0
    else:
        split = (df["Kq"]<0.1).sum()>0
    tdf = df.copy()
    if strip:
        if weighted:
            tdf.ix[tdf['fdr'].str.contains('FDR < 0.1'),'wKz'] = np.nan
        else:
            tdf.ix[tdf['fdr'].str.contains('FDR < 0.1'),'Kz'] = np.nan
    if weighted:
        sns.violinplot(x='wKz',y='dummy',hue='fdr',data=tdf,inner='quart',color=pal,split=split,scale='width',cut=0,ax=ax)
    else:
        sns.violinplot(x='Kz',y='dummy',hue='fdr',data=tdf,inner='quart',color=pal,split=split,scale='width',cut=0,ax=ax)
    if strip:
        df.ix[~df['fdr'].str.contains('FDR < 0.1'),'Kz'] = np.nan
        if weighted:
            sns.stripplot(x='wKz',y='dummy',hue='fdr',data=df,jitter=True,linewidth=1,color=pal,orient='h',split=True,edgecolor='white',size=10,lw=2,ax=ax)
        else:
            sns.stripplot(x='Kz',y='dummy',hue='fdr',data=df,jitter=True,linewidth=1,color=pal,orient='h',split=True,edgecolor='white',size=10,lw=2,ax=ax)
    ax.set_ylabel(title,fontsize=25,rotation='horizontal',ha='right')
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(fontsize=18,handles=handles[:2],loc='center right')
```

```python
# Setup the plot
fig,ax = plt.subplots(5,1,figsize=(20,10),sharex=True,facecolor='white')
plt.subplots_adjust(wspace=0, hspace=0)
pal = list(sns.color_palette("Set1", n_colors=5, desat=.5))

# Plot the FDR-significant K for solvent accessibility
df = prep_split(rsa,weighted=True)
plot_dist(df,"Solvent\nAccessibility",ax[0],pal[0],weighted=True,strip=False)

# Plot the FDR-significant K for TonyCons
df = prep_split(con,weighted=True)
plot_dist(df,"Evolutionary\nConservation",ax[1],pal[1],weighted=True,strip=False)

# Plot the FDR-significant K for ExAC sSNVs
df = prep_split(syn)
plot_dist(df,"ExAC\nSynonymous",ax[2],pal[2],strip=True)

# Plot the FDR-significant K for ExAC nsSNVs
df = prep_split(ns)
plot_dist(df,"ExAC\nMissense",ax[3],pal[3],strip=False)

# Plot the FDR-significant K for ClinVar pathogenic
df = prep_split(cv)
plot_dist(df,"ClinVar\nPathogenic",ax[4],pal[4],strip=False)


plt.rcParams['xtick.labelsize'] = 20
plt.xlabel("Dispersion%sClustering%s"%(' '*70,' '*28),fontsize=23)
for x in ax:
    x.axvline(x=0,c='black',ls='dashed')
plt.xlim([-15,30])
plt.suptitle("The Spatial Patterns of Structural Properties and Variation",fontsize=30,y=0.935)
plt.show()
```
