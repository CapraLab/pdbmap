```python
import MySQLdb as mysql
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from scipy.stats import f_oneway,spearmanr
%matplotlib inline
```

```python
con = mysql.connect("chgr2.accre.vanderbilt.edu","sivleyrm","global-trifecta","ripleysK_results")
ec  = ["Oxidoreductases","Transferases","Hydrolases","Lyases","Isomerases","Ligases"]
```

```python
sql = """SELECT a.unp,a.pdbid,a.chain,nresidues,nsnvs,t,k,p,z,q,id1,id2
            FROM pdb_exac_missense_univariate a 
            inner join pdbmap_supp.enzyme b
            on a.unp=b.unp"""
df = pd.read_sql(sql,con)
df.sort_values(["id1","id2"],inplace=True)
df["id"] = df["id1"].str.cat(df["id2"],sep='.')
# df["id"] = df["id"].astype(int)
ln   = len(df.drop_duplicates(['pdbid','chain']))
df   = df.groupby("id").filter(lambda x: len(x)>=10)
msg  = "%d chains were removed because they were in a "
msg += "EC class of size <10. %d remaining."
msg  = msg%(ln-len(df.drop_duplicates(['pdbid','chain'])),
                len(df.drop_duplicates(['pdbid','chain'])))

print msg
ecid = list(df["id"].drop_duplicates())
print "Analyzing EC classification bias for %d ExAC results."%(len(df))
```

```python
# Setup the plot
fig,ax = plt.subplots(1,len(ecid),figsize=(20,7),sharey=True,facecolor='white')
plt.subplots_adjust(wspace=0, hspace=0)
pal = list(sns.color_palette("Set1", n_colors=6, desat=.5))

ecid = list(df["id"].drop_duplicates())
subs = [df.ix[df["id"]==ecid[i],"nresidues"] for i in range(len(ecid))]
subs,ecid = zip(*sorted(zip(subs,ecid),key=lambda x: np.median(x[0]),reverse=True))
for i,s in enumerate(subs):
    # Boxplot for each EC class
    sns.boxplot(s,color=pal[int(ecid[i][0])-1],ax=ax[i],orient='vertical')
    ax[i].set_xlabel("%s (%d)"%(ecid[i],len(s)),fontsize=18,rotation='vertical')
    ax[i].set_ylabel("")
    
plt.rcParams['xtick.labelsize'] = 20
ax[0].set_ylabel("Number of Residues",fontsize=23)
for x in ax:
    x.axhline(y=0,c='k',ls='dashed',lw=3)
plt.suptitle("ExAC: Number of Residues by EC Classification",fontsize=28,y=0.95)
plt.legend(handles=[mpatches.Patch(color=pal[i], label=ec[i]) for i in range(len(ec))],
           fontsize=15,loc="upper right")
plt.show()

print "ANOVA: F=%g; p=%.3g"%f_oneway(*subs)
```

```python
# Setup the plot
fig,ax = plt.subplots(1,len(ecid),figsize=(20,7),sharey=True,facecolor='white')
plt.subplots_adjust(wspace=0, hspace=0)
pal = list(sns.color_palette("Set1", n_colors=6, desat=.5))

subs = [df.ix[df["id"]==ecid[i],"z"] for i in range(len(ecid))]
for i,s in enumerate(subs):
    # Boxplot for each EC class
    sns.boxplot(s,color=pal[int(ecid[i][0])-1],ax=ax[i],orient='vertical')
    ax[i].set_xlabel("%s (%d)"%(ecid[i],len(s)),fontsize=18,rotation='vertical')
    ax[i].set_ylabel("")
    ax[i].set_ylim([-4,7])
    
plt.rcParams['xtick.labelsize'] = 20
ax[0].set_ylabel("Protein Summary Z-Score",fontsize=23)
for x in ax:
    x.axhline(y=0,c='k',ls='dashed',lw=3)
plt.suptitle("ExAC: Protein Summary Z-Score by EC Classification",fontsize=28,y=0.95)
plt.legend(handles=[mpatches.Patch(color=pal[i], label=ec[i]) for i in range(len(ec))],
           fontsize=15,loc="upper right")
plt.show()

# print "\n~1 outliers not shown."
print "ANOVA: F=%g; p=%.3g"%f_oneway(*subs)
```
