```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import cycle
%matplotlib inline
import numpy as np
import pandas as pd
pd.set_option('max_rows',1000)
import os
from qvalue import estimate
from glob import iglob
```

```python
# Load the ExAC nsSNV results
ns  = pd.read_csv("/Volumes/sivleyrm/pdbmap/results/specialK_analysis_2016-05-27/exac/daf_K_summary.txt",delimiter='\t')
ns  = ns.drop_duplicates(["structid","chain"])
print "Total proteins analyzed:     %4d"%len(ns)
ns.ix[~ns[ "Kp"].isnull(), "Kq"], K_lam, K_pi = estimate(ns.ix[~ns[ "Kp"].isnull(), "Kp"].values)
ns.ix[~ns["wKp"].isnull(),"wKq"],wK_lam,wK_pi = estimate(ns.ix[~ns["wKp"].isnull(),"wKp"].values)
print "FDR Significant Un-Weighted: %4d"%(ns["Kq"]<0.1).sum()
print "FDR Significant    Weighted: %4d"%(ns["wKq"]<0.1).sum()
# P-Value Distributions
pal = list(sns.color_palette("Set1", n_colors=3, desat=.5))
fig,ax = plt.subplots(3,1,figsize=(20,10))
plt.subplots_adjust(wspace=0, hspace=0)
plt.suptitle("P-Value Distributions for ExAC Univariate K",fontsize=25,y=1.02)
ax[0].set_ylabel("Unweighted K p-value",fontsize=20)
ax[0].hist(ns["Kp"],bins=np.arange(0,1.02,0.02),color=pal[0],normed=True)
ax[0].plot(K_lam,K_pi,lw=3,c='black')
ax[1].set_ylabel("Weighted K p-value",fontsize=20)
ax[1].hist(ns["wKp"],bins=np.arange(0,1.02,0.02),color=pal[1],normed=True)
ax[1].plot(wK_lam,wK_pi,lw=3,c='black')
sns.violinplot(ns["N"],color=pal[2],inner='quart',ax=ax[2])
ax[2].set_ylabel("Number of SNVs",fontsize=20)
plt.tight_layout()
plt.show()
```

```python
# Load the Alzheimer's control results
ctrl = pd.read_csv("/Volumes/sivleyrm/pdbmap/results/specialK_analysis_2016-06-24/geis_ad_control/maf_K_summary.txt",delimiter='\t')
ctrl = ctrl.drop_duplicates(["structid","chain"])
print "Total proteins analyzed:     %4d"%len(ctrl)
ctrl.ix[~ctrl[ "Kp"].isnull(), "Kq"], K_lam, K_pi = estimate(ctrl.ix[~ctrl[ "Kp"].isnull(), "Kp"].values)
ctrl.ix[~ctrl["wKp"].isnull(),"wKq"],wK_lam,wK_pi = estimate(ctrl.ix[~ctrl["wKp"].isnull(),"wKp"].values)
print "FDR Significant Un-Weighted: %4d"%(ctrl["Kq"]<0.1).sum()
print "FDR Significant    Weighted: %4d"%(ctrl["wKq"]<0.1).sum()
# P-Value Distributions
pal = list(sns.color_palette("Set1", n_colors=3, desat=.5))
fig,ax = plt.subplots(3,1,figsize=(20,10))
plt.subplots_adjust(wspace=0, hspace=0)
plt.suptitle("P-Value Distributions for Alzheimer's Control Univariate K",fontsize=25,y=1.02)
ax[0].set_ylabel("Unweighted K p-value",fontsize=20)
ax[0].hist(ctrl["Kp"],bins=np.arange(0,1.02,0.02),color=pal[0],normed=True)
ax[0].plot(K_lam,K_pi,lw=3,c='black')
ax[1].set_ylabel("Weighted K p-value",fontsize=20)
ax[1].hist(ctrl["wKp"],bins=np.arange(0,1.02,0.02),color=pal[1],normed=True)
ax[1].plot(wK_lam,wK_pi,lw=3,c='black')
sns.violinplot(ctrl["N"],color=pal[2],inner='quart',ax=ax[2])
ax[2].set_ylabel("Number of SNVs",fontsize=20)
plt.tight_layout()
plt.show()
```

```python
# Load the Alzheimer's case results
case = pd.read_csv("/Volumes/sivleyrm/pdbmap/results/specialK_analysis_2016-06-24/geis_ad_case/maf_K_summary.txt",delimiter='\t')
case = case.drop_duplicates(["structid","chain"])
print "Total proteins analyzed:     %4d"%len(case)
case.ix[~case[ "Kp"].isnull(), "Kq"], K_lam, K_pi = estimate(case.ix[~case[ "Kp"].isnull(), "Kp"].values)
case.ix[~case["wKp"].isnull(),"wKq"],wK_lam,wK_pi = estimate(case.ix[~case["wKp"].isnull(),"wKp"].values)
print "FDR Significant Un-Weighted: %4d"%(case["Kq"]<0.1).sum()
print "FDR Significant    Weighted: %4d"%(case["wKq"]<0.1).sum()
# P-Value Distributions
pal = list(sns.color_palette("Set1", n_colors=3, desat=.5))
fig,ax = plt.subplots(3,1,figsize=(20,10))
plt.subplots_adjust(wspace=0, hspace=0)
plt.suptitle("P-Value Distributions for Alzheimer's Case Univariate K",fontsize=25,y=1.02)
ax[0].set_ylabel("Unweighted K p-value",fontsize=20)
ax[0].hist(case["Kp"],bins=np.arange(0,1.02,0.02),color=pal[0],normed=True)
ax[0].plot(K_lam,K_pi,lw=3,c='black')
ax[1].set_ylabel("Weighted K p-value",fontsize=20)
ax[1].hist(case["wKp"],bins=np.arange(0,1.02,0.02),color=pal[1],normed=True)
ax[1].plot(wK_lam,wK_pi,lw=3,c='black')
sns.violinplot(case["N"],color=pal[2],inner='quart',ax=ax[2])
ax[2].set_ylabel("Number of SNVs",fontsize=20)
plt.tight_layout()
plt.show()
```

```python
def prep_split(dfx):
    df  = dfx.copy()
    df['fdr']  = '      All (N=%4d)'%len(df)
    dft = dfx[dfx["Kq"]<0.1].copy()
    dft['fdr'] = 'FDR < 0.1 (N=%4d)'%len(dft)
    df = pd.concat((df,dft)).dropna(subset=["Kz"])
    df['dummy'] = ''
    return df
```

```python
# Setup the plot
fig,ax = plt.subplots(3,1,figsize=(20,8),sharex=True,facecolor='white')
plt.subplots_adjust(wspace=0, hspace=0)
pal = list(sns.color_palette("Set1", n_colors=3, desat=.5))

# Plot the FDR-significant K for ExAC nsSNVs
df = prep_split(ns)
split = (df["Kq"]<0.1).sum()>0
sns.violinplot(x='Kz',y='dummy',hue='fdr',data=df,inner='quart',color=pal[0],split=split,scale='width',cut=0,ax=ax[0])
df.ix[~df['fdr'].str.contains('FDR < 0.1'),'Kz'] = np.nan
sns.stripplot(x='Kz',y='dummy',hue='fdr',data=df,jitter=True,linewidth=1,color=pal[0],orient='h',split=True,edgecolor='white',size=10,lw=2,ax=ax[0])
ax[0].set_ylabel("ExAC\nMissense",fontsize=25,rotation='horizontal',ha='right')
handles, labels = ax[0].get_legend_handles_labels()
ax[0].legend(fontsize=20,handles=handles[:2])

# Plot the FDR-significant K for ExAC sSNVs
df = prep_split(ctrl)
split = (df["Kq"]<0.1).sum()>0
tdf = df.copy()
tdf.ix[tdf['fdr']=='FDR < 0.1','Kz'] = np.nan
sns.violinplot(x='Kz',y='dummy',hue='fdr',data=tdf,inner='quart',color=pal[1],split=split,scale='width',cut=0,ax=ax[1])
df.ix[~df['fdr'].str.contains('FDR < 0.1'),'Kz'] = np.nan
sns.stripplot(x='Kz',y='dummy',hue='fdr',data=df,jitter=True,linewidth=1,color=pal[1],orient='h',split=True,edgecolor='white',size=10,lw=2,ax=ax[1])
ax[1].set_ylabel("Alzheimer's\nControl",fontsize=25,rotation='horizontal',ha='right')
handles, labels = ax[1].get_legend_handles_labels()
ax[1].legend(fontsize=20,handles=handles[:2])

# Plot the FDR-significant K for COSMIC somatic nsSNVs
df = prep_split(case)
split = (df["Kq"]<0.1).sum()>0
tdf = df.copy()
tdf.ix[tdf['fdr']=='FDR < 0.1','Kz'] = np.nan
sns.violinplot(x='Kz',y='dummy',hue='fdr',data=tdf,inner='quart',color=pal[2],split=split,scale='width',cut=0,ax=ax[2])
df.ix[~df['fdr'].str.contains('FDR < 0.1'),'Kz'] = np.nan
sns.stripplot(x='Kz',y='dummy',hue='fdr',data=df,jitter=True,linewidth=1,color=pal[2],orient='h',split=True,edgecolor='white',size=10,lw=2,ax=ax[2])
ax[2].set_ylabel("Alzheimer's\nCase",fontsize=25,rotation='horizontal',ha='right')
handles, labels = ax[2].get_legend_handles_labels()
ax[2].legend(fontsize=20,handles=handles[:2])

plt.rcParams['xtick.labelsize'] = 20
plt.xlabel("Dispersion%sClustering%s"%(' '*70,' '*50),fontsize=23)
for x in ax:
    x.axvline(x=0,c='black',ls='dashed')
plt.xlim([-5,20])
plt.suptitle("Un-Weighted Spatial Patterns Between Alzheimer's Cases and Controls",fontsize=30,y=0.95)
plt.show()
```
