```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from scipy.stats import f_oneway
pd.set_option('max_rows',1000)
import seaborn as sns
import os
from glob import iglob
from sklearn.metrics import auc
```

```python
def plot_roc(fpr,tpr,ax,c='black'):
    rocauc = auc(fpr,tpr)
    ax.plot(fpr,tpr,linewidth=1,alpha=0.5,color=c)
    return rocauc
```

```python
def plot_pr(prec,rec,ax,c='black'):
    prec = np.maximum.accumulate(prec)
    prauc = auc(rec,prec)
    ax.plot(rec,prec,linewidth=1,alpha=0.5,color=c)
    return prauc
```

```python
path_rocauc = []
pph2_rocauc = []
sift_rocauc = []
cons_rocauc = []

def set_limits(ax):
    ax.set_xlim([0.,1.])
    ax.set_ylim([0.,1.])
def plot_diag(ax):
    ax.plot([0,1],[0,1],'k--')
fig,ax = plt.subplots(2,2,figsize=(20,20),sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        plot_diag(ax[i][j])
for i,j in [(i,j) for i in range(2) for j in range(2)]:
    set_limits(ax[i][j])
ax[0][0].set_ylabel("True Positive Rate",fontsize=25)
ax[1][0].set_ylabel("True Positive Rate",fontsize=25)
ax[1][0].set_xlabel("False Positive Rate",fontsize=25)
ax[1][1].set_xlabel("False Positive Rate",fontsize=25)

# Plot the Colocalization ROC and PR curves
ax[0][0].set_title("PathProx",fontsize=28)
i=0
for f in iglob("/Volumes/doux/pdbmap/results/multiROC_08302016/*clinvar-exac_roc*.txt.gz"):
    i+=1
    df = pd.read_csv(f,sep='\t',header=None,names=["fpr","tpr"])
    rocauc = plot_roc(df["fpr"],df["tpr"],ax=ax[0][0])
    path_rocauc.append(rocauc)
print "\nPathogenic Proximity:"
print "Number of proteins:  %3d"%i
print "Mean ROC AUC:       %.2f"%(sum(path_rocauc)/float(i))

# Plot the Conservation ROC and PR curves
ax[0][1].set_title("Conservation",fontsize=28)
i=0
for f in iglob("/Volumes/doux/pdbmap/results/multiROC_08302016/*clinvar-exac_cons_roc*.txt.gz"):
    i+=1
    df = pd.read_csv(f,sep='\t',header=None,names=["fpr","tpr"])
    rocauc = plot_roc(df["fpr"],df["tpr"],ax=ax[0][1],c="orchid")
    cons_rocauc.append(rocauc)
print "\nConservation:"
print "Number of proteins:  %3d"%i
print "Mean ROC AUC:       %.2f"%(sum(cons_rocauc)/float(i))

# Plot the SIFT ROC and PR curves
ax[1][0].set_title("SIFT",fontsize=28)
i=0
for f in iglob("/Volumes/doux/pdbmap/results/multiROC_08302016/*clinvar-exac_sift_roc*.txt.gz"):
    i+=1
    df = pd.read_csv(f,sep='\t',header=None,names=["fpr","tpr"])
    rocauc = plot_roc(df["fpr"],df["tpr"],ax=ax[1][0],c="orange")
    sift_rocauc.append(rocauc)
print "\nSIFT:"
print "Number of proteins:  %3d"%i
print "Mean ROC AUC:       %.2f"%(sum(sift_rocauc)/float(i))

# Plot the PolyPhen2 ROC and PR curves
ax[1][1].set_title("PolyPhen",fontsize=28)
i=0
for f in iglob("/Volumes/doux/pdbmap/results/multiROC_08302016/*clinvar-exac_polyphen_roc*.txt.gz"):
    i+=1
    df = pd.read_csv(f,sep='\t',header=None,names=["fpr","tpr"])
    rocauc = plot_roc(df["fpr"],df["tpr"],ax=ax[1][1],c="steelblue")
    pph2_rocauc.append(rocauc) 
print "\nPolyPhen2:"
print "Number of proteins:  %3d"%i
print "Mean ROC AUC:       %.2f"%(sum(pph2_rocauc)/float(i))

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
fig,ax = plt.subplots(1,4,figsize=(20,7),sharey=True)
ax[0].set_ylabel("ROC%s\nAUC%s"%(' '*10,' '*10),fontsize=28,rotation='horizontal')
sns.boxplot(path_rocauc,ax=ax[0],orient='vertical',color="grey")#,inner='quart',cut=0)
ax[0].set_title("PathProx",fontsize=25,y=1.01)
ax[0].set_xlabel("Mean = 0.52",fontsize=20)
sns.boxplot(pph2_rocauc,ax=ax[1],orient='vertical',color="steelblue")#,inner='quart',cut=0)
ax[1].set_title("PolyPhen",fontsize=25,y=1.01)
ax[1].set_xlabel("Mean = 0.73",fontsize=20)
sns.boxplot(sift_rocauc,ax=ax[2],orient='vertical',color="orange")#,inner='quart',cut=0)
ax[2].set_title("SIFT",fontsize=25,y=1.01)
ax[2].set_xlabel("Mean = 0.71",fontsize=20)
sns.boxplot(cons_rocauc,ax=ax[3],orient='vertical',color="orchid")#,inner='quart',cut=0)
ax[3].set_title("Conservation",fontsize=25,y=1.01)
ax[3].set_xlabel("Mean = 0.70",fontsize=20)
fig.subplots_adjust(hspace=0,wspace=0)
plt.show()
print "ANOVA: F=%.2f, p=%.3g"%f_oneway(path_rocauc,pph2_rocauc,sift_rocauc,cons_rocauc)
```
