```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
pd.set_option('max_rows',1000)
import seaborn as sns
import os
from glob import iglob
from sklearn.metrics import auc
```

```python
def plot_roc(fpr,tpr,ax):
    plot_roc.sum_rocauc += auc(fpr,tpr)
    ax.plot(fpr,tpr,linewidth=1,alpha=0.5,color='black')
plot_roc.sum_rocauc = 0
```

```python
def plot_pr(prec,rec,ax):
    prec = np.maximum.accumulate(prec)
    plot_pr.sum_prauc += auc(rec,prec)
    ax.plot(rec,prec,linewidth=1,alpha=0.5,color='black')
plot_pr.sum_prauc = 0
```

```python
plot_roc.sum_rocauc = 0
plot_pr.sum_prauc   = 0
fig,ax = plt.subplots(1,2,figsize=(20,10))
plt.suptitle("Spatial Proximity Differentiates ClinVar from ExAC",fontsize=35,y=1.05)
ax[0].set_title("ROC",fontsize=20)
ax[1].set_title("PR",fontsize=20)
ax[0].plot([0,1],[0,1],'k--')
ax[0].set_xlim([0.,1.])
ax[0].set_ylim([0.,1.])
ax[1].set_xlim([0.,1.])
ax[1].set_ylim([0.,1.])
ax[0].set_xlabel("False Positive Rate",fontsize=20)
ax[0].set_ylabel("True Positive Rate",fontsize=20)
ax[1].set_xlabel("Recall",fontsize=20)
ax[1].set_ylabel("Precision",fontsize=20)

# Plot the ROC curves
i=0
for f in iglob("/Volumes/sivleyrm/pdbmap/results/multiROC_analysis_06-29-2016/clinvar-exac/*roc*.txt.gz"):
    i+=1
    df = pd.read_csv(f,sep='\t',header=None,names=["fpr","tpr"])
    plot_roc(df["fpr"],df["tpr"],ax=ax[0])
    
# Plot the PR curves
j=0
for f in iglob("/Volumes/sivleyrm/pdbmap/results/multiROC_analysis_06-29-2016/clinvar-exac/*pr*.txt.gz"):
    j+=1
    df = pd.read_csv(f,sep='\t',header=None,names=["prec","rec"])
    plot_pr(df["prec"],df["rec"],ax=ax[1])

print "Number of proteins:  %3d"%i
print "Mean ROC AUC:       %.2f"%(plot_roc.sum_rocauc/float(i))
print "Mean PR  AUC:       %.2f"%(plot_pr.sum_prauc/float(j))

plt.tight_layout()
plt.show()
```
