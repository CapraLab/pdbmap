```python
import MySQLdb as mysql
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway,spearmanr
%matplotlib inline
```

```python
con = mysql.connect("chgr2.accre.vanderbilt.edu","sivleyrm","global-trifecta","ripleysK_results")
```

```python
sql = """SELECT a.unp,a.pdbid,a.chain,nresidues,nsnvs,t,k,p,z,q,method,resolution
            FROM pdb_exac_missense_univariate a 
            inner join pdbmap_v12.Structure b
            on a.pdbid=b.pdbid
            where b.label='uniprot-pdb'"""
df = pd.read_sql(sql,con)
```

```python
# Structures types
print df.groupby("method").apply(len).sort_values(ascending=False)
```

```python
# Only look at correlations for X-ray diffraction structures
tdf = df[df["method"]=="x-ray diffraction"]
# Remove all structures where resolution was not recorded
tdf = tdf[tdf["resolution"]!=-1]

fig,ax = plt.subplots(2,1,figsize=(20,10),sharex=True)

# Test for correlation with the protein-summary Z
print "  Z-Score: R=% 7.2g; P=% 5.2g"%(spearmanr(tdf["resolution"],tdf["z"]))
sns.regplot(x="resolution",y="z",data=tdf,ax=ax[0])
ax[0].set_xlabel("Structure Resolution",fontsize=20)
ax[0].set_ylabel("Ripley's K Protein Z-Score",fontsize=20)
for tick in ax[0].xaxis.get_major_ticks():
    tick.label.set_fontsize(18)
for tick in ax[0].yaxis.get_major_ticks():
    tick.label.set_fontsize(18)

# Test for correlation with the magnitude of Z
tdf["z"] = tdf["z"].apply(abs)
print "|Z-Score|: R=% 7.2g; P=% 5.2g"%(spearmanr(tdf["resolution"],tdf["z"]))
sns.regplot(x="resolution",y="z",data=tdf,ax=ax[1])
ax[1].set_xlabel("Structure Resolution",fontsize=20)
ax[1].set_ylabel("Ripley's K Protein |Z-Score|",fontsize=20)
for tick in ax[1].xaxis.get_major_ticks():
    tick.label.set_fontsize(18)
for tick in ax[1].yaxis.get_major_ticks():
    tick.label.set_fontsize(18)
ax[1].set_ylim([-0.1,12])
plt.tight_layout()
plt.show()
plt.close(fig)

# Test for resolution differences between significant and non-significant structures
print "F=% 4.2g; P=% 5.2g"%(f_oneway(tdf.ix[tdf["q"]>=0.1,"resolution"],tdf.ix[tdf["q"]<0.1,"resolution"]))
fig,ax = plt.subplots(1,2,figsize=(20,10),sharey=True)
sns.boxplot(tdf.ix[tdf["q"]>=0.1,"resolution"],ax=ax[0],orient='vertical',color='cornflowerblue')
sns.boxplot(tdf.ix[tdf["q"]<0.1,"resolution"],ax=ax[1],orient='vertical',color='cornflowerblue')
ax[0].set_xlabel("Not Significant",fontsize=20)
ax[1].set_xlabel("Significant",fontsize=20)
ax[0].set_ylabel("Resolution",fontsize=20)
ax[1].set_ylabel("")
for tick in ax[0].yaxis.get_major_ticks():
    tick.label.set_fontsize(18)
plt.tight_layout()
plt.show()
```

```python
# Only look at correlations for X-ray diffraction structures
tdf = df[df["method"]=="x-ray diffraction"]
fig,ax = plt.subplots(2,1,figsize=(20,10))

# Positive Z-Scores
# Test for correlation with the magnitude of Z
ttdf = tdf[tdf["z"]>0]
print "Z-Score: R=% 7.2g; P=% 6.2g"%(spearmanr(ttdf["resolution"],ttdf["z"]))
sns.regplot(x="resolution",y="z",data=ttdf,ax=ax[0])
ax[0].set_xlabel("Resolution",fontsize=20)
ax[0].set_ylabel("Ripley's K Protein |Z-Score|",fontsize=20)
for tick in ax[0].xaxis.get_major_ticks():
    tick.label.set_fontsize(18)
for tick in ax[0].yaxis.get_major_ticks():
    tick.label.set_fontsize(18)
ax[0].set_title("Proteins with Positive Z-Scores (Clustering)",fontsize=25)
# ax[0].set_ylim([-0.1,12])

# Negative Z-Scores
# Test for correlation with the magnitude of Z
ttdf = tdf[tdf["z"]<0]
print "Z-Score: R=% 7.2g; P=% 6.2g"%(spearmanr(ttdf["resolution"],ttdf["z"]))
sns.regplot(x="resolution",y="z",data=ttdf,ax=ax[1])
ax[1].set_xlabel("Resolution",fontsize=20)
ax[1].set_ylabel("Ripley's K Protein |Z-Score|",fontsize=20)
for tick in ax[1].xaxis.get_major_ticks():
    tick.label.set_fontsize(18)
for tick in ax[1].yaxis.get_major_ticks():
    tick.label.set_fontsize(18)
ax[1].set_title("Proteins with Negative Z-Scores (Dispersion)",fontsize=25)
# ax[1].set_ylim([-0.1,12])

plt.tight_layout()
plt.show()

print "\nConclusion: Protein size is significantly correlated with dispersion,"
print " but not with clustering."
```

```python
# Test for differences between x-ray crystallography and solution nmr
tdf = df.copy()

# Z-Score
fig,ax = plt.subplots(1,2,figsize=(20,10),sharey=True)
print "Z-Score:   F=%6.2f; P=% 5.2g"%(f_oneway(tdf.ix[tdf["method"]=="x-ray diffraction","z"],
                                     tdf.ix[tdf["method"]=="solution nmr","z"]))
sns.boxplot(tdf.ix[tdf["method"]=="x-ray diffraction","z"],ax=ax[0],orient='vertical',color='cornflowerblue')
sns.boxplot(tdf.ix[tdf["method"]=="solution nmr","z"],ax=ax[1],orient='vertical',color='cornflowerblue')
ax[0].set_ylabel("Protein Z-Score",fontsize=20)
ax[1].set_ylabel("")
for tick in ax[0].yaxis.get_major_ticks():
    tick.label.set_fontsize(18)
plt.tight_layout()
plt.show()

# |Z-Score|
fig,ax = plt.subplots(1,2,figsize=(20,10),sharey=True)
tdf['z'] = tdf['z'].apply(abs)
print "|Z-Score|: F=%6.2f; P=% 5.2g"%(f_oneway(tdf.ix[tdf["method"]=="x-ray diffraction","z"],
                                     tdf.ix[tdf["method"]=="solution nmr","z"]))
sns.boxplot(tdf.ix[tdf["method"]=="x-ray diffraction","z"],ax=ax[0],orient='vertical',color='cornflowerblue')
sns.boxplot(tdf.ix[tdf["method"]=="solution nmr","z"],ax=ax[1],orient='vertical',color='cornflowerblue')
ax[0].set_ylabel("Protein |Z-Score|",fontsize=20)
ax[1].set_ylabel("")
for tick in ax[0].yaxis.get_major_ticks():
    tick.label.set_fontsize(18)
plt.tight_layout()
plt.show()

# Number of Residues
fig,ax = plt.subplots(1,2,figsize=(20,10),sharey=True)
print "Residues:  F=%6.2f; P=% 5.2g"%(f_oneway(tdf.ix[tdf["method"]=="x-ray diffraction","nresidues"],
                                     tdf.ix[tdf["method"]=="solution nmr","nresidues"]))
sns.boxplot(tdf.ix[tdf["method"]=="x-ray diffraction","nresidues"],ax=ax[0],orient='vertical',color='cornflowerblue')
sns.boxplot(tdf.ix[tdf["method"]=="solution nmr","nresidues"],ax=ax[1],orient='vertical',color='cornflowerblue')
ax[0].set_ylabel("Number of Residues",fontsize=20)
ax[1].set_ylabel("")
for tick in ax[0].yaxis.get_major_ticks():
    tick.label.set_fontsize(18)
plt.tight_layout()
plt.show()

# Number of SNVs
fig,ax = plt.subplots(1,2,figsize=(20,10),sharey=True)
print "SNVs:      F=%6.2f; P=% 5.2g"%(f_oneway(tdf.ix[tdf["method"]=="x-ray diffraction","nsnvs"],
                                     tdf.ix[tdf["method"]=="solution nmr","nsnvs"]))
sns.boxplot(tdf.ix[tdf["method"]=="x-ray diffraction","nsnvs"],ax=ax[0],orient='vertical',color='cornflowerblue')
sns.boxplot(tdf.ix[tdf["method"]=="solution nmr","nsnvs"],ax=ax[1],orient='vertical',color='cornflowerblue')
ax[0].set_ylabel("Number of SNVs",fontsize=20)
ax[1].set_ylabel("")
for tick in ax[0].yaxis.get_major_ticks():
    tick.label.set_fontsize(18)
plt.tight_layout()
plt.show()

# Peak Distance Threshold
fig,ax = plt.subplots(1,2,figsize=(20,10),sharey=True)
print "Peak Dist: F=%6.2f; P=% 5.2g"%(f_oneway(tdf.ix[tdf["method"]=="x-ray diffraction","t"],
                                     tdf.ix[tdf["method"]=="solution nmr","t"]))
sns.boxplot(tdf.ix[tdf["method"]=="x-ray diffraction","t"],ax=ax[0],orient='vertical',color='cornflowerblue')
sns.boxplot(tdf.ix[tdf["method"]=="solution nmr","t"],ax=ax[1],orient='vertical',color='cornflowerblue')
ax[0].set_ylabel("Peak Distance Threshold",fontsize=20)
ax[1].set_ylabel("")
for tick in ax[0].yaxis.get_major_ticks():
    tick.label.set_fontsize(18)
ax[0].set_xlabel("X-Ray Diffraction",fontsize=20)
ax[1].set_xlabel("Solution NMR",fontsize=20)
plt.tight_layout()
plt.show()
```
