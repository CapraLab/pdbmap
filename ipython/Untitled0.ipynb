```python
import numpy as np
from scipy.stats import describe
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def pi(snps):
    # First dimension should be samples
    # Second dimension should be variants
    # Calculates the hamming distance between two sample sequences
    return sum([(i!=j).sum() for i in snps for j in snps])/float(len(snps)**3)
```

```python
# Simulate individual SNP frequency in a population
snps = np.array([np.random.uniform(0,1,1000)<(AF/1000.) for AF in xrange(0,1000,5)]).astype(np.int)
```

```python
mafs = np.array([AF/1000. for AF in xrange(0,1000,5)])
actual_mafs = np.apply_along_axis(sum,1,snps)/1000.
plt.scatter(mafs,actual_mafs)
plt.xlabel("Simulated MAF")
plt.ylabel("Actual MAF")
plt.title("MAF Simulation Validation")
```

```python
# Measure nucleotide diversity for a single site at different frequencies
pis = np.apply_along_axis(pi,1,snps)
```

```python
# Plot Pi by MAF for individual sites
plt.scatter(actual_mafs,pis)
plt.xlabel("Minor Allele Frequency")
plt.ylabel("Nucleotide Diversity")
plt.title("Effect of MAF on Nucleotide Diversity")
```

```python
# Measure nucleotide diversity across all sites
pis = pi(snps)
print "Nucleotide diversity: %2.4f nucleotide differences per site"%pis
```

```python
# Simulate an "abundance" of rare variation
# i.e. simulate an equally sized matrix containing only rare variants (<0.5%)
rare = np.array([np.random.uniform(0,1,1000)<(AF/100000.) for AF in xrange(0,1000,5)]).astype(np.int)
# Bind this matrix to the original matrix
snps = np.r_[rare,snps]
```

```python
rare = np.array([AF/100000. for AF in xrange(0,1000,5)])
mafs = np.r_[rare,mafs]
actual_mafs = np.apply_along_axis(sum,1,snps)/1000.
plt.scatter(mafs,actual_mafs)
plt.xlabel("Simulated MAF")
plt.ylabel("Actual MAF")
plt.title("MAF Simulation Validation")
```

```python
# Measure nucleotide diversity across all sites
# with an abundance of rare variation
pis = pi(snps)
print "Nucleotide diversity: %2.4f nucleotide differences per site"%pis
```

```python
# Average number of variants per sample
print "Average variants per sample (out of %d sites): %2.2f"%(snps.shape[1],snps.sum(axis=1).mean())
```
