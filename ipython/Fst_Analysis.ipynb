```python
# Load libraries
%matplotlib inline
import glob,gzip,os,operator
import numpy as np
from numpy.random import randn
from itertools import combinations
import random
import pandas as pd
from scipy import stats
from scipy.spatial import KDTree
from sklearn.metrics import confusion_matrix
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patheffects as PathEffects
pd.set_option('display.max_columns', 500)
```

```python
def conf_matrix(ax,grp1,grp2,lab1,lab2):
    # Calculate the confusion matrix
    grp1,grp2 = grp1.align(grp2,join="outer",axis=0)
    grp1 = grp1.fillna(False)
    grp2 = grp2.fillna(False)
    if not (grp1 & grp2).sum():
        cm = np.zeros((2,2))
        cm[0][0] = len(grp1)
    else:
        cm = confusion_matrix(grp1,grp2)
    ax.text(0,0,'{0:,}'.format(cm[0][0]),fontsize=40,horizontalalignment='center',verticalalignment='center',color='white',path_effects=[PathEffects.withStroke(linewidth=3,foreground="black")])
    ax.text(0,1,'{0:,}'.format(cm[1][0]),fontsize=40,horizontalalignment='center',verticalalignment='center',color='white',path_effects=[PathEffects.withStroke(linewidth=3,foreground="black")])
    ax.text(1,0,'{0:,}'.format(cm[0][1]),fontsize=40,horizontalalignment='center',verticalalignment='center',color='white',path_effects=[PathEffects.withStroke(linewidth=3,foreground="black")])
    ax.text(1,1,'{0:,}'.format(cm[1][1]),fontsize=40,horizontalalignment='center',verticalalignment='center',color='white',path_effects=[PathEffects.withStroke(linewidth=3,foreground="black")])
    cm[0][0] = 0 # improve visualization of significant agreement/disagreement
    # Show confusion matrix in a separate window
    ax.imshow(cm,cmap=plt.cm.jet,interpolation='nearest')
#     ax.set_title("%s vs %s"%(lab1,lab2),fontsize=35)
    ax.set_ylabel(lab1,fontsize=30)
    ax.set_xlabel(lab2,fontsize=30)
    ax.set_xlim([-0.5,1.5])
    ax.set_ylim([-0.5,1.5])
    ax.set_xticks([0,1])
    ax.set_yticks([0,1])
    ax.set_xticklabels(["Not Significant","Significant"])
    ax.set_yticklabels(["Not Significant","Significant"],rotation='vertical')
    ax.grid(False,'both','both')
```

```python
def unlist_col(col):
    return np.array([str(c).split(',') for c in col])
```

```python
resdir = "/Volumes/sivleyrm/pdbmap/results/fst_analysis_20150619"
shcols = ["fst","fst'","nsnp","snplist","maflist","fstlist","unpac","hgnc"]
def load_fst(metric,unq_cols):
    print "header: %s"%' '.join(unq_cols+shcols)
    df = pd.read_csv("%s/fst_by_%s.txt"%(resdir,metric),sep=" ",header=None,names=unq_cols+shcols,encoding="utf-8",index_col=False)
    df["fst"] = df["fst"].apply(lambda x: np.max([x,0]))
    # Disregard bins with fewer than 2 SNPs
    df = df[df["nsnp"]>1]
    # As well as SNPs with an Fst <= 0
    df = df[df["fst"]>0]
    df["snplist"] = unlist_col(df["snplist"])
    df["maflist"] = unlist_col(df["maflist"])
    df["fstlist"] = unlist_col(df["fstlist"])
#     df.sort("unpac")
    print "%d %s bins"%(len(df),metric)
    return df
```

```python
fst_by_gene   = load_fst("gene",["gene"])
# fst_by_unp    = load_fst("uniprot",["unp"])
fst_by_pfam   = load_fst("pfam",["gene","pfam"])
# fst_by_ipfam  = load_fst("ipfam",["gene","pfamblock"])
fst_by_scop   = load_fst("scop",["gene","scop"])
# fst_by_iscop  = load_fst("iscop",["gene","scopblock"])
# fst_by_sec    = load_fst("secondary",["structid","biounit","ss"])
# fst_by_isec   = load_fst("isecondary",["structid","biounit","chain","ssblock"])
fst_by_sphere = load_fst("sphere",["structid","biounit","model","chain","seqid"])
```

```python
# fst_grp = [fst_by_gene,fst_by_unp,fst_by_pfam,fst_by_ipfam,fst_by_scop,fst_by_iscop,fst_by_isec,fst_by_sphere]
# fst_grp = [fst_by_gene,fst_by_pfam,fst_by_ipfam,fst_by_scop,fst_by_iscop,fst_by_isec,fst_by_sphere]
fst_grp = [fst_by_gene,fst_by_pfam,fst_by_scop,fst_by_sphere]
```

```python
# fst_labs = ["by-gene","by-pfam","by-ipfam","by-scop","by-iscop","by-iss","by-sphere"]
fst_labs = ["by-gene","by-pfam","by-scop","by-sphere"]
```

```python
# Number of UniProt proteins covered by each method
for i,fst in enumerate(fst_grp):
    print "%8s\t%2.2f"%(fst_labs[i],len(fst.groupby("unpac")))
```

```python
# Average SNP count per metric
for i,fst in enumerate(fst_grp):
    print "%8s\t%2.2f"%(fst_labs[i],fst["nsnp"].mean())
```

```python
# Mean percent difference in maximum Fst and bin Fst
print "Mean Percent Difference Between Max Individual Fst and Binned Fst"
fst_diff = [np.mean((fst["fst"]-fst_by_sphere["fstlist"].apply(lambda x: np.max(np.array(x,dtype=np.float64))))/fst["fst"]) for fst in fst_grp]
print "Method  \tPerDiff\tnSNPs"
for i,diff in enumerate(fst_diff):
    print "%8s\t% 0.2f\t%0.2f"%(fst_labs[i],diff,fst_grp[i]["nsnp"].mean())
```

```python
# Correlation between Max Individual Fst and Binned Fst
print " Method \tMax-Bin\tMed-Bin\tMax-Med"
for i,fst in enumerate(fst_grp):
    x = fst["fst"]
    y = fst["fstlist"].apply(lambda x: np.max(np.array(x,dtype=np.float64)))
    z = fst["fstlist"].apply(lambda x: np.median(np.array(x,dtype=np.float64)))
    print "%8s\t%2.2f\t%2.2f\t%2.2f"%(fst_labs[i],np.corrcoef(x,y)[0][1],np.corrcoef(x,z)[0][1],np.corrcoef(y,z)[0][1])
```

```python
# Calculate the % decrease from max individual Fst to bin Fst
fig,axes = plt.subplots(len(fst_grp),1,figsize=(10,50))
for i,fst in enumerate(fst_grp):
    ax = axes[i]
#     ax.set_xlim([-100,100])
    fst  = fst[fst["fstlist"].apply(lambda x: np.max(np.array(x,dtype=np.float64)))>fst["fst"]]
    fmax = fst["fstlist"].apply(lambda x: np.max(np.array(x,dtype=np.float64)))
    x = (fmax-fst["fst"])/fmax
    x.hist(ax=ax,bins=50)
    ax.set_title("%s (%d)"%(fst_labs[i],len(x)))
    ax.set_xlabel("% Decrease from Max Individual Fst to Bin Fst")
plt.tight_layout()
```

```python
# Calculate the % increase from max individual Fst to bin Fst
fig,axes = plt.subplots(len(fst_grp),1,figsize=(10,50))
for i,fst in enumerate(fst_grp):
    ax = axes[i]
#     ax.set_xlim([-100,100])
    fst  = fst[fst["fstlist"].apply(lambda x: np.max(np.array(x,dtype=np.float64)))<fst["fst"]]
    if fst.empty:
        continue
    fmax = fst["fstlist"].apply(lambda x: np.max(np.array(x,dtype=np.float64)))
    x = (fst["fst"]-fmax)/fst["fst"]
    x.hist(ax=ax,bins=50)
    ax.set_title("%s (%d)"%(fst_labs[i],len(x)))
    ax.set_xlabel("% Increase from Max Individual Fst to Bin Fst")
plt.tight_layout()
```

```python
# Print high-Fst spheres with completely duplicate spheres (exact same SNPs) removed to give a better sense for the region
sig = (fst_by_sphere["fst"]>0.05) & (fst_by_sphere["fst'"]>0.05)
print "%d unique spheres with Fst > 0.05 and Fst' > 0.05"%len(fst_by_sphere[sig].drop_duplicates(["structid","fst"]))
print "%d proteins containing spheres with Fst > 0.05 and Fst' > 0.05"%len(fst_by_sphere[sig].groupby("unpac"))
fst_by_sphere["mfst"] = fst_by_sphere["fstlist"].apply(lambda x: np.array(x,dtype=np.float64).mean())
cols = ["structid","hgnc","unpac","seqid","fst","fst'","mfst","nsnp","fstlist"]
fst_by_sphere.ix[sig,cols].sort(["nsnp","mfst","fst'"],ascending=False).drop_duplicates(["structid","fst"])
```

```python
# Generate an 8x8 confusion matrix
# Calculate the confusion matrix
fig,axes = plt.subplots(len(sig_grp),len(sig_grp),figsize=(50,50))
for i,grp1 in enumerate(sig_grp):
    for j,grp2 in enumerate(sig_grp):
        if j <= i:
            grp1,grp2 = grp1.align(grp2,fill_value=False)
            conf_matrix(axes[i][j],grp1,grp2,fst_labs[i],fst_labs[j])
        else:
            axes[i][j].set_frame_on(False)
            axes[i][j].axes.get_xaxis().set_visible(False)
            axes[i][j].axes.get_yaxis().set_visible(False)
plt.tight_layout()
```
