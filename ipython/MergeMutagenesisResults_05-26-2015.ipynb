```python
# Load libraries
%matplotlib inline
import glob,gzip,os
import numpy as np
from numpy.random import randn
from itertools import combinations
import random
import pandas as pd
from scipy import stats
from scipy.spatial import KDTree
from sklearn.metrics import confusion_matrix
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 500)
```

```python
# Read the mutation disease classifications
dclass = {'WT':-1}
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/1kg3.txt','rb') as fin:
  dclass.update(dict((m.strip(),0) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/unclassified.txt','rb') as fin:
  dclass.update(dict((m.strip(),1) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/dyskeratosis.txt','rb') as fin:
  dclass.update(dict((m.strip(),2) for m in fin))
with open('/Volumes/sivleyrm/pdbmap/temp/rtel1/published_pathogenic.txt','rb') as fin:
  dclass.update(dict((m.strip(),3) for m in fin))
dcolor = {-2:"orange",-1:"black",0:"darkgray",1:"green",2:"orangered",3:"red"}
```

```python
resdir  = "/Volumes/sivleyrm/pdbmap/results/Q9NZ71_mutagenesis_20150526"
names   = []
results = []
for f in glob.glob("%s/*.results"%resdir):
  # Record the structid and biounit
  fname = os.path.basename(f)
  if fname in ["score.results","wt-rmsd.results","rwt-rmsd.results","ddg.results"]: continue
  structid,biounit,_ = fname.split('.')[0].split('_')
  # Read the Wild-Type and Relaxed Wild-Type scores for this structure
  with open(f,'rb') as fin:
    wt  = float(fin.readline().split(':')[-1].strip())
    rwt = float(fin.readline().split(':')[-1].strip())
  # Extract results for this structure
  try:
    df = pd.read_csv(f,header=2,sep='\t',comment="#")
  except:
    print "\nException: %s\n"%f
    raise
  # Add structid and biounit columns
  sid = "%s_%s"%(structid,biounit)
  df["Structid"] = [sid]*len(df)
  # Reduce to minimum scores and RSMDs
  df = df[["Mutation","Structid","Unrelaxed_Score","Min_Score","Min_MT_RMSD","rWT_RMSD"]]
  # Calculate ddG
  df["ddG"]  = df["Unrelaxed_Score"] - rwt
  df["rddG"] = df["Min_Score"] - rwt
  # Record the WT and rWT scores
  df = df.append({"Mutation":"WT","Unrelaxed_Score":rwt,"Structid":sid},ignore_index=True)
  # Index by structid, biounit, and mutation
  df.set_index(["Structid","Mutation"],inplace=True)
  results.append(df)
# Merge results
df = pd.concat(results)
df = df.reset_index()
print "Number of Mutation+Structure Pairs Before Pruning: %d"%len(df)
```

```python
# Optional: Remove bad structures
print "\nRemoving structures prone to moving:"
bad_structs = ['ENSP00000359035_1','ENSP00000353332_4','ENSP00000359020_2','ENSP00000424307_3']
for struct in bad_structs:
  print "Removing %s"%struct
  df = df[df["Structid"]!=struct]
print ''
print "Number of Mutation+Structure Pairs After Pruning: %d"%len(df)
```

```python
# Pivot by structure
df = df.reindex_axis(sorted(df.columns),axis=1)

cls = [dclass[d] if d[0]!='r' else dclass[d[1:]] for d in df["Mutation"]]
labels = ["1000 Genomes","Unclassified","Dyskeratosis","IPF","WT"]
df["DiseaseClass"] = [labels[c] for c in cls]
```

```python
# Plot the Unrelaxed Mutant scores
plt.clf()
plt.figure(figsize=(15,10))
sns.boxplot(df["Unrelaxed_Score"],groupby=df["DiseaseClass"])
plt.title("Energy Scores of Unrelaxed RTEL1 Mutations")
plt.xlabel("Disease Classification")
plt.ylabel("Rosetta Energy Score")
plt.tight_layout()
plt.savefig("%s/mt_score.pdf"%resdir)
plt.savefig("%s/mt_score.png"%resdir)
plt.show()
```

```python
# Plot ddG from Relaxed Wild Type to Unelaxed Mutant by disease class
plt.clf()
plt.figure(figsize=(15,10))
sns.boxplot(df["ddG"],groupby=df["DiseaseClass"])
plt.ylabel("Delta-Delta-G from rWT to MT")
plt.xlabel("Disease Classification")
plt.xticks(rotation="vertical")
plt.title("Energetic Effect of Unrelaxed RTEL1 Mutations by Disease Class")
plt.legend()
plt.tight_layout()
plt.savefig("%s/ddG_bydclass_boxplot.pdf"%resdir)
plt.savefig("%s/ddG_bydclass_boxplot.png"%resdir)
```
