```python
# Imports, configurations, and initializations
from pandas import DataFrame,read_table
import pandas as pd
import numpy as np
from scipy import stats
# Import and initiailize matplotlib parameters
%matplotlib inline
import matplotlib.pylab as plt
import matplotlib as mpl
mpl.rcParams.update({'font.size': 14, 'font.family': 'serif'})
```

```python
df = read_table('1kg_clustering_coefficient.txt',header=None,sep='\t',error_bad_lines=False,
                names=["type","structid","biounit",
                       "all","nres","snpres","nsnp","pdcoef","pd.sig","npdcoef","npd.sig","dscoef15","ds15.sig","ndscoef15","nds15.sig",
                       "amr","amr_nres","amr_snpres","amr_nsnp","amr_pdcoef","amr_pd.sig","amr_npdcoef","amr_npd.sig","amr_dscoef15","amr_ds15.sig","amr_ndscoef15","amr_nds15.sig",
                       "asn","asn_nres","asn_snpres","asn_nsnp","asn_pdcoef","asn_pd.sig","asn_npdcoef","asn_npd.sig","asn_dscoef15","asn_ds15.sig","asn_ndscoef15","asn_nds15.sig",
                       "afr","afr_nres","afr_snpres","afr_nsnp","afr_pdcoef","afr_pd.sig","afr_npdcoef","afr_npd.sig","afr_dscoef15","afr_ds15.sig","afr_ndscoef15","afr_nds15.sig",
                       "eur","eur_nres","eur_snpres","eur_nsnp","eur_pdcoef","eur_pd.sig","eur_npdcoef","eur_npd.sig","eur_dscoef15","eur_ds15.sig","eur_ndscoef15","eur_nds15.sig"])
```

```python
# print "PDBMap contains 117,298 structures/models"
print "Model / Structure counts:\n",df['type'].value_counts()
print "NaN dscoef15 counts:",np.sum(np.isnan(df['dscoef15']))
print "INF dscoef15 counts:",np.sum(np.isinf(df['dscoef15']))
```

## Correlation between populations on value of Inverse Pairwise Distance Coefficient

```python
## Start by comparing the ALL-cluster with the AFR-cluster.
## Red:  > in ALL
## Blue: > in AFR
tdf = df[["npdcoef","afr_npdcoef","nsnp","afr_nsnp"]].replace(np.inf,1)
# Manually assign NA to 0.0
tdf = tdf.replace(np.nan,0)
# Manually round negative values to 0
tdf[tdf<0] = 0
# Represent structures with 20+ SNPs differences w/ 20
tdf.loc[:,"dsnp"] = tdf.nsnp-tdf.afr_nsnp
tdf.dsnp[tdf.dsnp>10] = 10
tdf.dsnp[tdf.dsnp<-10] = -10
# Force -10..10 spectrum centered on 0 using two dummy structures
tdf.loc[len(tdf)] = [0,0,0,0,-10]
tdf.loc[len(tdf)] = [0,0,0,0,10] 

# Correlation matrix between populations with respect to npdcoef
plt.figure(figsize=(10,10),dpi=300)
plt.scatter(tdf.afr_npdcoef,tdf.npdcoef,marker='o',
            c=tdf.dsnp,cmap=plt.get_cmap('seismic'))
plt.ylabel('NPD Coefficient (ALL)')
plt.xlabel('NPD Coefficient (AFR)')
plt.tight_layout()
del tdf
```

```python
## And compare the EUR-cluster with the AFR-cluster
## Red:  > in EUR
## Blue: > in AFR
tdf = df[["eur_npdcoef","afr_npdcoef","eur_nsnp","afr_nsnp"]].replace(np.inf,1)
# Manually assign NA to 0.0
tdf = tdf.replace(np.nan,0)
# Manually round negative values to 0
tdf[tdf<0] = 0
# Represent structures with 20+ SNPs differences w/ 20
tdf.loc[:,"dsnp"] = tdf.eur_nsnp-tdf.afr_nsnp
tdf.dsnp[tdf.dsnp>10] = 10
tdf.dsnp[tdf.dsnp<-10] = -10
# Force -10..10 spectrum centered on 0 using two dummy structures
tdf.loc[len(tdf)] = [0,0,0,0,-10]
tdf.loc[len(tdf)] = [0,0,0,0,10] 

# Linear model to test associations
from scipy import stats
import statsmodels.formula.api as smf
print smf.ols("afr_npdcoef ~ eur_npdcoef + dsnp",data=tdf).fit().summary()

# Correlation matrix between populations with respect to npdcoef
plt.figure(figsize=(10,10),dpi=300)
plt.scatter(tdf.afr_npdcoef,tdf.eur_npdcoef,marker='o',
            c=tdf.dsnp,cmap=plt.get_cmap('seismic'))
plt.ylabel('NPD Coefficient (EUR)')
plt.xlabel('NPD Coefficient (AFR)')
plt.tight_layout()
del tdf
```

```python
# Manually assign infinite values to 1.0
cols = ["npdcoef","amr_npdcoef","asn_npdcoef","afr_npdcoef","eur_npdcoef","nsnp",
        "amr_nsnp","asn_nsnp","afr_nsnp","eur_nsnp"]
tdf  = df[["npdcoef","amr_npdcoef","asn_npdcoef","afr_npdcoef","eur_npdcoef",
           "nsnp","amr_nsnp","asn_nsnp","afr_nsnp","eur_nsnp"]].replace(np.inf,1)
names = ["NPD ALL","NPD AMR","NPD ASN","NPD AFR","NPD EUR"]
# Represent structures with 20+ SNPs differences w/ 20
for i in range(5,10):
    for j in range(5,10):
        tdf.loc[:,5*i+j-20,] = tdf.iloc[:,i]-tdf.iloc[:,j]
        tdf.iloc[:,5*i+j-20,][tdf.iloc[:,5*i+j-20,]>10] = 10
        tdf.iloc[:,5*i+j-20,][tdf.iloc[:,5*i+j-20,]<-10] = -10
# Force -10..10 spectrum centered on 0 using two dummy structures
tdf.loc[len(tdf)] = [0 for i in range(10)]+[-10 for i in range(25)]
tdf.loc[len(tdf)] = [0 for i in range(10)]+[10 for i in range(25)]
# Manually assign NA to 0.0
tdf = tdf.replace(np.nan,0)
# Manually round negative values to 0
tdf[tdf<0] = 0

# Correlation matrix between populations with respect to npdcoef
fig, axes = plt.subplots(figsize=(16,15),nrows=5,ncols=5,dpi=300)
histlim = None
for i in range(5):
    for j in range(5):
        if i == j:
            axes[i][j].hist(tdf.iloc[:,i])
            # Set ylim of all histograms equal to the first
            histlim = axes[i][j].get_ylim() if not histlim else histlim
            axes[i][j].set_ylim(histlim)
            axes[i][j].set_xlim(0,0.15)
            if i == 0:
                axes[i][j].set_ylabel(names[i])
                axes[i][j].set_xticks([])
            elif j == 4:
                axes[i][j].set_xlabel(names[j])
                axes[i][j].set_yticks([])
            else:
                axes[i][j].set_xticks([])
                axes[i][j].set_yticks([])
        else:
#             colc = cols[i+5]+cols[j+5]
            axes[i][j].scatter(tdf.iloc[:,j],tdf.iloc[:,i],marker='o',alpha=0.4,
                               c=tdf.iloc[:,10+(5*i)+j],cmap=plt.get_cmap('seismic'))
            axes[i][j].set_xlim(0,0.15)
            axes[i][j].set_ylim(0,0.15)
            axes[i][j].set_xticks([])
            axes[i][j].set_yticks([])
            if j == 0: # Left column
                axes[i][j].set_ylabel(names[i])
                axes[i][j].set_yticks([0,0.05,0.1,0.15])
            if i == 4: # Bottom row
                axes[i][j].set_xlabel(names[j])
                axes[i][j].set_xticks([0,0.05,0.1,0.15])
fig.tight_layout()
```

## Correlation between populations on p-value of Inverse Pairwise Distance Coefficient

```python
# Manually assign infinite values to 1.0
tdf = df[["npd.sig","amr_npd.sig","asn_npd.sig","afr_npd.sig","eur_npd.sig"]].replace(np.inf,1)
# Manually assign NA to 1.0
tdf = tdf.replace(np.nan,1)

# Correlation matrix between populations with respect to npdcoef p-value
pd.tools.plotting.scatter_matrix(-np.log(df[["npd.sig","amr_npd.sig","asn_npd.sig","afr_npd.sig","eur_npd.sig"]]),
                                 diagonal='kde',alpha=0.1,figsize=(10,10))
plt.tight_layout()
del tdf
```

## Concordance between populations on significance of Inverse Pairwise Distance Coefficient

```python
# Correlation matrix between populations with respect to npdcoef significance
pops = ['','amr_','asn_','afr_','eur_']
conc_mat = [[len(df[(df['%snpd.sig'%pop1]<0.05) & (df['%snpd.sig'%pop2]<0.05)]) for pop1 in pops] for pop2 in pops]
print '\tALL\tAMR\tASN\tAFR\tEUR'
for i,pop in enumerate(['ALL','AMR','ASN','AFR','EUR']):
    print pop+'\t'+'\t'.join(str(x) for x in conc_mat[i])
# Convert to z-score
conc_mat = stats.zscore(conc_mat)
print '\n\t ALL\t AMR\t ASN\t AFR\t EUR'
for i,pop in enumerate(['ALL','AMR','ASN','AFR','EUR']):
    print pop+'\t'+'\t'.join("% 2.2f"%x for x in conc_mat[i])
```
