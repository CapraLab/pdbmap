```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy.stats import spearmanr,pearsonr
```

```python
# Load the results from the MONOMERIC colocalization analysis
df = pd.read_csv("/Volumes/sivleyrm/pdbmap/results/Colocalization_KCNQ1_2016-07-29-17-54/KCNQ1_data.txt",
                 header=0,sep='\t')
df["mutation"] = df["ref"].str.cat(df["pos"].astype(str)).str.cat(df["alt"])
# Add the experimental results from the homotetramer experiments
tdf1 = pd.read_csv("/Volumes/sivleyrm/pdbmap/data/kcnq1/variants/deleterious.txt",
                 names=["pos","ref","alt","I%WT"],sep='\t')
tdf1["I%WT"] = tdf1["I%WT"].astype(np.float64)
tdf2 = pd.read_csv("/Volumes/sivleyrm/pdbmap/data/kcnq1/variants/deleterious.txt",
                 names=["pos","ref","alt","I%WT"],sep='\t')
tdf2["I%WT"] = tdf2["I%WT"].astype(np.float64)
tdf = pd.concat((tdf1,tdf2))
tdf["mutation"] = tdf["ref"].str.cat(tdf["pos"].astype(str)).str.cat(tdf["alt"])
df = df.merge(tdf,on="mutation")
```

```python
# Plot the correlation
tdf = df.dropna()
print "Pearson  Correlation: %.2f (p=%.1g)"%pearsonr(tdf["I%WT"],tdf["pathprox"])
print "Spearman Correlation: %.2f (p=%.1g)"%spearmanr(tdf["I%WT"],tdf["pathprox"])
plt.figure(figsize=(20,7))
sns.regplot(df["pathprox"],df["I%WT"],scatter_kws={"s":50,"alpha":0.5,"c":"darkred"})
plt.title("Correlation Between Pathogenic Proximity and I%WT (Homotetramer + KCNE)",fontsize=25,y=1.01)
plt.ylabel("I%WT + KCNE1",fontsize=18)
plt.xlabel("Pathogenic Proximity",fontsize=18)
plt.axhline(0.,ls="--",color="darkgrey",lw=3)
plt.tight_layout()
plt.show()
```
