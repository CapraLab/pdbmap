```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy.stats import spearmanr
```

```python
# Load the results from the colocalization analysis
df = pd.read_csv("/Volumes/sivleyrm/pdbmap/results/Colocalization_KCNQ1_Homotetramer_Severe_2016-07-25-15-36/KCNQ1_Homotetramer_Severe_data.txt",
                 header=0,sep='\t')
df["mutation"] = df["ref"].str.cat(df["pos"].astype(str)).str.cat(df["alt"])
# Add the experimental results from the homotetramer experiments
tdf = pd.read_csv("/Volumes/sivleyrm/pdbmap/data/kcnq1/variants/homo_experimental.txt",
                 names=["mutation","I%WT"],sep='\t')
tdf["I%WT"] = tdf["I%WT"].astype(np.float64)
df = df.merge(tdf,on="mutation")
```

```python
# Plot the correlation
print "Spearman Correlation: %.2f (p=%.1g)"%spearmanr(df["I%WT"],df["pathprox"])
plt.figure(figsize=(20,7))
sns.regplot(df["pathprox"],df["I%WT"],scatter_kws={"s":50,"alpha":0.5})
plt.title("Correlation Between I%WT and Pathogenic Proximity (Homotetramer)",fontsize=25,y=1.01)
plt.ylabel("I%WT + KCNE1",fontsize=18)
plt.xlabel("Pathogenic Proximity",fontsize=18)
plt.axhline(0.,ls="--",color="darkgrey",lw=3)
plt.tight_layout()
plt.show()
```
