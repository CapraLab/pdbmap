```python
import numpy as np
import pandas as pd
from pi_util import *
%matplotlib inline
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 18})
import seaborn as sns
from copy import deepcopy
from scipy.stats import ks_2samp
%matplotlib inline
```

```python
def plot_distribution(df):
    plt.figure(figsize=(20,7))
    
    NUM_STEPS = 500
    ITER_STEP = NUM_STEPS/50.
    
    # Plot structure discovery as a function of random sphere selection
    idx = np.random.permutation(df.index)
    total_structs = len(df.drop_duplicates(["structid","biounit"]))
    nstructs = [len(df.loc[idx[:i],:].drop_duplicates(["structid","biounit"])) for i in range(0,len(idx),len(idx)/NUM_STEPS)[:NUM_STEPS]]
    x = deepcopy(nstructs)
    plt.scatter(np.array(range(NUM_STEPS))/ITER_STEP,nstructs,color="darkblue",label="random")
    plt.annotate("Random 5%%: %d (%.2f%%)"%(nstructs[20],nstructs[20]*100./nstructs[-1]),(2.75,nstructs[18]))

    # Plot structure discovery as a function deltaPi percentile
    lpcntls = [np.percentile(df["dpi"],i/ITER_STEP) for i in range(0,  NUM_STEPS, 1)]
    upcntls = [np.percentile(df["dpi"],i/ITER_STEP) for i in range(2*NUM_STEPS,NUM_STEPS,-1)]
    nstructs = [len(df[(df["dpi"]<=lpcntls[i]) | (df["dpi"]>=upcntls[i])].drop_duplicates(["structid","biounit"])) for i in range(NUM_STEPS)]
    y = deepcopy(nstructs)
    plt.scatter(np.array(range(NUM_STEPS))/ITER_STEP,nstructs,color="darkred",label="deltaPi-ordered")
    plt.annotate("5th Percentile: %d (%.2f%%)"%(nstructs[20],nstructs[20]*100./nstructs[-1]),(2.75,nstructs[18]))
    plt.legend(loc="lower right")
    plt.title("Structure Discovery as a Function of deltaPi Two-Sided Percentile Threshold")
    plt.xlabel("Delta-Nucleotide Diversity Percentile (Top or Bottom)")
    plt.ylabel("Number of Structures")
    plt.xticks(range(0,50,5),range(0,100,10))
    plt.xlim([0,50])
    plt.ylim([0,total_structs])

    # 5th percentile with respect to the entire distribution (below 2.5th and above 97.5th)
    plt.axvline(2.5,color='darkred')

    # Test the distributions for equality
    D,p = ks_2samp(x,y)
    print "Two-sided 2-Sample KS Test: D=%.3f; p<%.1e"%(D,p)
    return D,p
```

## All Missense Variants

```python
# Check for distribution biases over the entire dataset
df  = load_results("all_pi.txt")
df  = collapse(df)
D,p = plot_distribution(df)
```

```python
# Reduce to localized spheres and check for bias
df  = load_results("all_pi.txt")
df  = filter_fet(df,0.05,adj=False)
df  = collapse(df)
D,p = plot_distribution(df)
```
