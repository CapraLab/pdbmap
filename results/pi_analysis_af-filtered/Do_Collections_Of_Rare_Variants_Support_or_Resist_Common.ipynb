```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 18})
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("all/pi_all.txt",header=0,sep='\t')
# Store the entire dataset for future reference
aadf = df.copy()
print "Total spheres examined:\t\t%d"%len(df)
# Identify spheres with the same set of SNPs
ndup = (df.duplicated(["#structid","biounit","nbrsnps1","nbrsnps2"])).values.sum()
print "Number of SNP-set duplicates:\t%d"%ndup
print "Number of unique SNP sets:\t%d"%(len(df)-ndup)
# Are there spheres where neighbor SNPs listed as NaN? If so, make sure their SNP count is 0
nansnp1 = (df["nbrsnps1"].astype(str)=="nan").values 
nansnp2 = (df["nbrsnps2"].astype(str)=="nan").values
df.ix[nansnp1,"snpcnt1"] = 0
df.ix[nansnp2,"snpcnt2"] = 0
# Filter any spheres with fewer than 2 SNPs
df = df[(df["snpcnt1"]>1) | (df["snpcnt2"]>1)]
print "Total spheres after requiring 2+ SNPs: %d"%len(df)
# Filter any spheres with allele counts less than 10
df = df[(df["ac1"]>0) | (df["ac2"]>0)]
print "Total spheres after requiring 0+ alleles per sphere: %d"%len(df)
# Filter any spheres with a value of 0
df["absdpi"] = abs(df["dpi"])
df = df[df["absdpi"]>0]
print "Total spheres after removing non-differentiated: %d"%len(df)
# Select the sphere with the minimum value of deltaPi
df = df.sort("absdpi",ascending=True).replace(np.nan,-999).groupby(["#structid","biounit","nbrsnps1","nbrsnps2"],as_index=False).nth(0)
# Select the sphere with the median value of deltaPi
# df = df.groupby(["#structid","biounit","nbrsnps1","nbrsnps2"]).apply(lambda x: x[x["dpi"]==x["dpi"].median()])
# df = df.replace(-999,np.nan)
print "Total spheres after collapse:\t%d"%len(df)
# Identify spheres where deltaPi is NaN. These shouldn't exist.
df = df[~np.isnan(df["pi1"])]
df = df[~np.isnan(df["pi2"])]
print "Total spheres after filtering NaN:\t%d"%len(df)
adf = df
```

```python
df = pd.read_csv("rare/pi_rare.txt",header=0,sep='\t')
# Store the entire dataset for future reference
ardf = df.copy()
print "Total spheres examined:\t\t%d"%len(df)
# Identify spheres with the same set of SNPs
ndup = (df.duplicated(["#structid","biounit","nbrsnps1","nbrsnps2"])).values.sum()
print "Number of SNP-set duplicates:\t%d"%ndup
print "Number of unique SNP sets:\t%d"%(len(df)-ndup)
# Are there spheres where neighbor SNPs listed as NaN? If so, make sure their SNP count is 0
nansnp1 = (df["nbrsnps1"].astype(str)=="nan").values 
nansnp2 = (df["nbrsnps2"].astype(str)=="nan").values
df.ix[nansnp1,"snpcnt1"] = 0
df.ix[nansnp2,"snpcnt2"] = 0
# Filter any spheres with fewer than 2 SNPs
df = df[(df["snpcnt1"]>1) | (df["snpcnt2"]>1)]
print "Total spheres after requiring 2+ SNPs: %d"%len(df)
# Filter any spheres with allele counts less than 10
df = df[(df["ac1"]>0) | (df["ac2"]>0)]
print "Total spheres after requiring 0+ alleles per sphere: %d"%len(df)
# Filter any spheres with a value of 0
df["absdpi"] = abs(df["dpi"])
df = df[df["absdpi"]>0]
print "Total spheres after removing non-differentiated: %d"%len(df)
# Select the sphere with the minimum value of deltaPi
df = df.sort("absdpi",ascending=True).replace(np.nan,-999).groupby(["#structid","biounit","nbrsnps1","nbrsnps2"],as_index=False).nth(0)
# Select the sphere with the median value of deltaPi
# df = df.groupby(["#structid","biounit","nbrsnps1","nbrsnps2"]).apply(lambda x: x[x["dpi"]==x["dpi"].median()])
# df = df.replace(-999,np.nan)
print "Total spheres after collapse:\t%d"%len(df)
# Identify spheres where deltaPi is NaN. These shouldn't exist.
df = df[~np.isnan(df["pi1"])]
df = df[~np.isnan(df["pi2"])]
print "Total spheres after filtering NaN:\t%d"%len(df)
rdf = df
```

## Identify structures with spheres in the extreme 5% of the deltaPi distribution and localized (FET)

```python
p025 = np.percentile(rdf["dpi"].values, 2.5)
p975 = np.percentile(rdf["dpi"].values,97.5)
extRare = rdf[(rdf['dpi']>p975) | (rdf['dpi']<p025)]
extRare = extRare[extRare["fetp"]<0.01]
print "Number of spheres in the extreme of the Rare-SNP analysis: %d"%len(extRare)
rarestructs = extRare[["#structid","biounit"]].drop_duplicates().values
allcnt = len(rdf[["#structid","biounit"]].drop_duplicates())
extcnt = len(extRare[["#structid","biounit"]].drop_duplicates())
print "%d of %d (%.2f%%) rare-only structures contain spheres in the extreme 5%%"%(extcnt,allcnt,extcnt*100./allcnt)
```

```python
p025 = np.percentile(adf["dpi"].values, 2.5)
p975 = np.percentile(adf["dpi"].values,97.5)
extAll = adf[(adf['dpi']>p975) | (adf['dpi']<p025)]
extAll = extAll[extAll["fetp"]<0.01]
print "Number of spheres in the extreme of the All-SNP analysis: %d"%len(extAll)
allstructs = extAll[["#structid","biounit"]].drop_duplicates().values
allcnt = len(adf[["#structid","biounit"]].drop_duplicates())
extcnt = len(extAll[["#structid","biounit"]].drop_duplicates())
print "%d of %d (%.2f%%) all-only structures contain spheres in the extreme 5%%"%(extcnt,allcnt,extcnt*100./allcnt)
```

## How many of the extreme structures are shared or unique to the ALL- or RARE-only analyses?

```python
rarestructs = [tuple(x) for x in rarestructs]
allstructs  = [tuple(x) for x in allstructs]
print "Number of structures with spheres in the top 5%% of the rare-only analysis:  %4d"%len(rarestructs)
print "Number of structures with spheres in the top 5%% of the all-only analysis:   %4d"%len(allstructs)
unionstructs = set(allstructs).union(rarestructs)
interstructs = set(allstructs).intersection(rarestructs)
interpct     = len(interstructs)*100. / len(unionstructs)
print "Number of shared structures between the all- and rare-only analysis:        %4d (%.2f%%)"%(len(interstructs),interpct)
rareonly = set(rarestructs).difference(allstructs)
rarepct  = len(rareonly)*100. / len(rarestructs)
print "Number of structures unique to the rare-only analysis:                      %4d (%.2f%%)"%(len(rareonly),rarepct)
allonly = set(allstructs).difference(rarestructs)
allpct  = len(allonly)*100. / len(allstructs)
print "Number of structures unique to the all-only analysis:                       %4d (%.2f%%)"%(len(allonly),allpct)
```

## For spheres that were extreme in both analyses, do their effect sizes and directions agree?

```python
# Merge the all and rare-only analysis results (pre-filtered)
mdf = aadf.merge(ardf,on=["#structid","biounit","model","chain","seqid","icode"],how='outer',suffixes=('_all','_rare'))
# Identify spheres in the extreme 5% of the all analysis
abool = (mdf["dpi_all"]  < np.percentile(adf.ix[np.isfinite(adf["dpi"]),"dpi"].values,2.5))  | (mdf["dpi_all"]  > np.percentile(adf.ix[np.isfinite(adf["dpi"]),"dpi"].values,97.5))
print "Number of extreme spheres from All-SNP analysis:  %d"%sum(abool)
rbool = (mdf["dpi_rare"] < np.percentile(rdf.ix[np.isfinite(rdf["dpi"]),"dpi"].values,2.5)) | (mdf["dpi_rare"] > np.percentile(rdf.ix[np.isfinite(rdf["dpi"]),"dpi"].values,97.5))
print "Number of extreme spheres from Rare-SNP analysis: %d"%sum(rbool)
print '--'
print "Number of unioned extreme spheres: %d"%sum(abool | rbool)
print "Number of intersected extreme spheres: %d"%sum(abool & rbool)
# Assign all extremes to "blue", rare-only extremes to "red", and shared extremes to "green"
cdict = {(False,False):'darkgrey',(True,False):'darkblue',(False,True):'darkred',(True,True):'darkgreen'}
mdf["col"] = [cdict[(abool[i],rbool[i])] for i in range(len(abool))]
# Subset the dataframe to extreme spheres
sdf   = mdf[(abool | rbool)]
# Collapse spheres from the all-snp analysis into non-redundant snp-sets
sdf["absdpi"] = abs(sdf["dpi_all"])
asdf = sdf[sdf["col"]=="darkblue"].sort("absdpi",ascending=False).replace(np.nan,-999).groupby(["#structid","biounit","nbrsnps1_all","nbrsnps2_all"],as_index=False).nth(0)
asdf = asdf.replace(-999,np.nan)
# Collapse spheres from the rare-snp analysis into non-redundant snp-sets
sdf["absdpi"] = abs(sdf["dpi_rare"])
rsdf = sdf[sdf["col"]=="darkred"].sort("absdpi",ascending=False).replace(np.nan,-999).groupby(["#structid","biounit","nbrsnps1_rare","nbrsnps2_rare"],as_index=False).nth(0)
rsdf = rsdf.replace(-999,np.nan)
# Shared spheres will collapse correctly with either key, b/c that's how they were originally generated
sdf["absdpi"] = abs(sdf["dpi_all"])
ssdf = sdf[sdf["col"]=="darkgreen"].sort("absdpi",ascending=False).replace(np.nan,-999).groupby(["#structid","biounit","nbrsnps1_all","nbrsnps2_all"],as_index=False).nth(0)
ssdf = ssdf.replace(-999,np.nan)
# Concatenate the datafram back together
sdf = pd.concat([asdf,rsdf,ssdf])
# Perform the same reduction on the original data for plotting only
rmdf = mdf.copy()
# Collapse spheres from the all-snp analysis into non-redundant snp-sets
rmdf["absdpi"] = abs(rmdf["dpi_all"])
armdf = rmdf[rmdf["col"]=="darkblue"].sort("absdpi",ascending=False).replace(np.nan,-999).groupby(["#structid","biounit","nbrsnps1_all","nbrsnps2_all"],as_index=False).nth(0)
armdf = armdf.replace(-999,np.nan)
# Collapse spheres from the rare-snp analysis into non-redundant snp-sets
rmdf["absdpi"] = abs(rmdf["dpi_rare"])
rrmdf = rmdf[rmdf["col"]=="darkred"].sort("absdpi",ascending=False).replace(np.nan,-999).groupby(["#structid","biounit","nbrsnps1_rare","nbrsnps2_rare"],as_index=False).nth(0)
rrmdf = rrmdf.replace(-999,np.nan)
# Shared sig/insig spheres will collapse correctly with either key, b/c that's how they were originally generated
rmdf["absdpi"] = abs(rmdf["dpi_all"])
srmdf = rmdf[(rmdf["col"]=="darkgreen") | (rmdf["col"]=="darkgrey")].sort("absdpi",ascending=False).replace(np.nan,-999).groupby(["#structid","biounit","nbrsnps1_all","nbrsnps2_all"],as_index=False).nth(0)
srmdf = srmdf.replace(-999,np.nan)
# Concatenate the datafram back together
rmdf = pd.concat([armdf,rrmdf,srmdf])

print "Spheres chosen from all-extreme:    %4d"%(sdf["col"]=="darkblue").sum()
print "Spheres chosen from rare-extreme:   %4d"%(sdf["col"]=="darkred").sum()
print "Spheres chosen from shared extreme: %4d"%(sdf["col"]=="darkgreen").sum()
print "*Some* redundant spheres have been introduced, but counts are very comparable."
print '--'
print "Spheres with NaN deltaPi values for the All-SNP results:  %4d"%np.isnan(sdf["dpi_all"]).sum()
print "Spheres with NaN deltaPi values for the Rare-SNP results: %4d"%np.isnan(sdf["dpi_rare"]).sum()
print "Setting Rare-SNP deltaPi to 0 for spheres with no rare variants"
sdf.ix[np.isnan(sdf["dpi_rare"]),"dpi_rare"] = 0.

# Report the off-plot outliers
outliers = sdf.ix[(sdf["dpi_all"]>0.03) | (sdf["dpi_all"]<-0.03) | (sdf["dpi_rare"]>0.015) | (sdf["dpi_rare"]<-0.015),
         ["#structid","biounit","model","chain","seqid","dpi_all","dpi_rare","col"]]
print "Spheres (1 & 2) outside the plot range: %d"%len(outliers)
print outliers

pexall = len(sdf.ix[sdf["dpi_rare"]==0,"dpi_all"]) * 100. / len(sdf)
pexrar = len(sdf.ix[(sdf["dpi_all"]==sdf["dpi_rare"]),"dpi_rare"]) * 100. / len(sdf)
print "Percentage of extreme All-SNP spheres with no *rare* variants:    %.2f%%"%pexall
print "Percentage of extreme Rare-SNP spheres with no *common* variants: %.2f%%"%pexrar

plt.figure(figsize=(25,15))
trmdf = rmdf[rmdf["dpi_rare"]!=0]
plt.hexbin(trmdf["dpi_all"],trmdf["dpi_rare"],gridsize=50,bins='log',mincnt=0,cmap=plt.cm.YlOrRd_r)
plt.title("Non-Empty, Non-Redundant, Rare-Containing Sphere Density (%d)"%len(trmdf))
plt.xlabel("Pi(AFR) > Pi(EUR)%sAll-SNP%sPi(EUR) > Pi(AFR)"%(''.join([' ' for j in range(50)]),''.join([' ' for j in range(50)])))
plt.ylabel("Pi(AFR) > Pi(EUR)%sRare-SNP%sPi(EUR) > Pi(AFR)"%(''.join([' ' for j in range(50)]),''.join([' ' for j in range(50)])))
plt.colorbar().set_label('log10(Sphere Count)')
plt.axvline(0,lw=2,color='.3',ls='-')
plt.axhline(0,lw=2,color='.3',ls='-')
plt.plot([-0.0075, 0.015], [-0.0075, 0.015],lw=2,ls="-", c=".3")
plt.show()

# # Decrease 0-tolerance to |dpi|>0.0001 for both rare and all
# trmdf = trmdf[(abs(trmdf["dpi_all"])>0.0001) & (abs(trmdf["dpi_rare"])>0.0001)]
# q1 = len(trmdf[(trmdf["dpi_all"]> 0.0001) & (trmdf["dpi_rare"]> 0.0001)]) * 100. / len(trmdf)
# q2 = len(trmdf[(trmdf["dpi_all"]<-0.0001) & (trmdf["dpi_rare"]> 0.0001)]) * 100. / len(trmdf)
# q3 = len(trmdf[(trmdf["dpi_all"]<-0.0001) & (trmdf["dpi_rare"]<-0.0001)]) * 100. / len(trmdf)
# q4 = len(trmdf[(trmdf["dpi_all"]> 0.0001) & (trmdf["dpi_rare"]<-0.0001)]) * 100. / len(trmdf)
# plt.annotate("Q1: %.2f%%"%q1, xy=( 0.04, 0.005),color="white")
# plt.annotate("Q2: %.2f%%"%q2, xy=(-0.03, 0.005),color="white")
# plt.annotate("Q3: %.2f%%"%q3, xy=(-0.03,-0.005),color="white")
# plt.annotate("Q4: %.2f%%"%q4, xy=( 0.04,-0.005),color="white")

# Plot the spheres, isolate spheres that contains only-rare or only-common SNPs
fig,axes = plt.subplots(5,1,figsize=(20,50))
classes = ["Not Extreme","All-SNP Extreme","Rare-SNP Extreme","Shared Extreme"]
for i,c in enumerate(["darkgrey","darkblue","darkred","darkgreen"]):
    # Plot the entire dataset (not just the extremes)
    nspheres = len(mdf)
    axes[0].set_xlim([-0.03,0.03])
    axes[0].set_ylim([-0.015,0.015])
    axes[0].scatter(mdf.ix[mdf["col"]==c,"dpi_all"],mdf.ix[mdf["col"]==c,"dpi_rare"],alpha=0.3,c=c,label=classes[i])
    axes[0].set_title("All Non-Empty Spheres (%d)"%nspheres)
    axes[0].set_xlabel("Pi(AFR) > Pi(EUR)%sAll-SNP%sPi(EUR) > Pi(AFR)"%(''.join([' ' for j in range(70)]),''.join([' ' for j in range(70)])))
    axes[0].set_ylabel("Pi(AFR) > Pi(EUR)%sRare-SNP%sPi(EUR) > Pi(AFR)"%(''.join([' ' for j in range(25)]),''.join([' ' for j in range(25)])))
    axes[0].axvline(0,lw=0.5,color='.3',ls='--')
    axes[0].axhline(0,lw=0.5,color='.3',ls='--')
    axes[0].plot([-0.02, 0.02], [-0.02, 0.02],lw=0.5,ls="--", c=".3")
    axes[0].plot([-0.02, 0.02], [ 0.02,-0.02],lw=0.5,ls="--", c=".3")
    # Plot the non-redundant dataset (not just the extremes)
    nspheres = len(rmdf)
    axes[1].scatter(rmdf.ix[rmdf["col"]==c,"dpi_all"],rmdf.ix[rmdf["col"]==c,"dpi_rare"],alpha=0.3,c=c,label=classes[i])
    axes[1].set_title("All Non-Empty, Non-Redundant Spheres (%d)"%nspheres)
    axes[1].set_xlabel("Pi(AFR) > Pi(EUR)%sAll-SNP%sPi(EUR) > Pi(AFR)"%(''.join([' ' for j in range(70)]),''.join([' ' for j in range(70)])))
    axes[1].set_ylabel("Pi(AFR) > Pi(EUR)%sRare-SNP%sPi(EUR) > Pi(AFR)"%(''.join([' ' for j in range(25)]),''.join([' ' for j in range(25)])))
    axes[1].set_xlim([-0.03,0.03])
    axes[1].set_ylim([-0.015,0.015])
    axes[1].axvline(0,lw=0.5,color='.3',ls='--')
    axes[1].axhline(0,lw=0.5,color='.3',ls='--')
    axes[1].plot([-0.02, 0.02], [-0.02, 0.02],lw=0.5,ls="--", c=".3")
    axes[1].plot([-0.02, 0.02], [ 0.02,-0.02],lw=0.5,ls="--", c=".3")
    # Plot the rare-only and common-only spheres
    nspheres = len(sdf[((sdf["dpi_all"]==sdf["dpi_rare"]) | (sdf["dpi_rare"]==0))])
    axes[2].scatter(sdf.ix[((sdf["dpi_all"]==sdf["dpi_rare"]) | (sdf["dpi_rare"]==0)) & (sdf["col"]==c),"dpi_all"],
                    sdf.ix[((sdf["dpi_all"]==sdf["dpi_rare"]) | (sdf["dpi_rare"]==0)) & (sdf["col"]==c),"dpi_rare"],
                    c=sdf.ix[sdf["col"]==c,"col"],alpha=0.8,label=classes[i])
    axes[2].set_title("Extreme spheres with exclusively rare or exclusively common SNPs (%d)"%nspheres)
    axes[2].set_xlabel("Pi(AFR) > Pi(EUR)%sAll-SNP%sPi(EUR) > Pi(AFR)"%(''.join([' ' for j in range(70)]),''.join([' ' for j in range(70)])))
    axes[2].set_ylabel("Pi(AFR) > Pi(EUR)%sRare-SNP%sPi(EUR) > Pi(AFR)"%(''.join([' ' for j in range(25)]),''.join([' ' for j in range(25)])))
    axes[2].set_xlim([-0.03,0.03])
    axes[2].set_ylim([-0.015,0.015])
    axes[2].axvline(0,lw=0.5,color='.3',ls='--')
    axes[2].axhline(0,lw=0.5,color='.3',ls='--')
    axes[2].plot([-0.01, 0.01], [-0.01, 0.01],lw=0.5,ls="--", c=".3")
    # Plot the spheres with both rare and common SNPs
    nspheres = len(sdf[((sdf["dpi_all"]!=sdf["dpi_rare"]) & (sdf["dpi_rare"]!=0))])
    axes[3].scatter(sdf.ix[((sdf["dpi_all"]!=sdf["dpi_rare"]) & (sdf["dpi_rare"]!=0)) & (sdf["col"]==c),"dpi_all"],
                    sdf.ix[((sdf["dpi_all"]!=sdf["dpi_rare"]) & (sdf["dpi_rare"]!=0)) & (sdf["col"]==c),"dpi_rare"],
                    c=sdf.ix[sdf["col"]==c,"col"],alpha=0.8,label=classes[i])
    axes[3].set_title("Extreme spheres with both rare and common SNPs (%d)"%nspheres)
    axes[3].set_xlabel("Pi(AFR) > Pi(EUR)%sAll-SNP%sPi(EUR) > Pi(AFR)"%(''.join([' ' for j in range(70)]),''.join([' ' for j in range(70)])))
    axes[3].set_ylabel("Pi(AFR) > Pi(EUR)%sRare-SNP%sPi(EUR) > Pi(AFR)"%(''.join([' ' for j in range(25)]),''.join([' ' for j in range(25)])))
    axes[3].set_xlim([-0.03,0.03])
    axes[3].set_ylim([-0.015,0.015])
    axes[3].axvline(0,lw=0.5,color='.3',ls='--')
    axes[3].axhline(0,lw=0.5,color='.3',ls='--')
    axes[3].axhline( 0.0001,lw=0.5,color='r',ls='--') # Plot the "effective 0"
    axes[3].axhline(-0.0001,lw=0.5,color='r',ls='--') # Plot the "effective 0"
    axes[3].plot([-0.02, 0.02], [-0.02, 0.02],lw=0.5,ls="--", c=".3")
    axes[3].plot([-0.02, 0.02], [ 0.02,-0.02],lw=0.5,ls="--", c=".3")
    # Plot the spheres with at least two rare SNPs and at least one common SNP
    nspheres = len(sdf[((sdf["dpi_all"]!=sdf["dpi_rare"]) & (sdf["dpi_rare"]!=0)) & ((sdf["snpcnt1_rare"]>1) | (sdf["snpcnt2_rare"]>1)) & ((sdf["snpcnt1_all"]-sdf["snpcnt1_rare"]>0) | (sdf["snpcnt2_all"]-sdf["snpcnt2_rare"]>0))])
    axes[4].scatter(sdf.ix[((sdf["dpi_all"]!=sdf["dpi_rare"]) & (sdf["dpi_rare"]!=0)) & (sdf["col"]==c) & ((sdf["snpcnt1_rare"]>1) | (sdf["snpcnt2_rare"]>1)) & ((sdf["snpcnt1_all"]-sdf["snpcnt1_rare"]>0) | (sdf["snpcnt2_all"]-sdf["snpcnt2_rare"]>0)),"dpi_all"],
                    sdf.ix[((sdf["dpi_all"]!=sdf["dpi_rare"]) & (sdf["dpi_rare"]!=0)) & (sdf["col"]==c) & ((sdf["snpcnt1_rare"]>1) | (sdf["snpcnt2_rare"]>1)) & ((sdf["snpcnt1_all"]-sdf["snpcnt1_rare"]>0) | (sdf["snpcnt2_all"]-sdf["snpcnt2_rare"]>0)),"dpi_rare"],
                    c=sdf.ix[sdf["col"]==c,"col"],alpha=0.8,label=classes[i])
    axes[4].set_title("Extreme spheres with both 2+ rare and 1+ common SNPs (%d)"%nspheres)
    axes[4].set_xlabel("Pi(AFR) > Pi(EUR)%sAll-SNP%sPi(EUR) > Pi(AFR)"%(''.join([' ' for j in range(70)]),''.join([' ' for j in range(70)])))
    axes[4].set_ylabel("Pi(AFR) > Pi(EUR)%sRare-SNP%sPi(EUR) > Pi(AFR)"%(''.join([' ' for j in range(25)]),''.join([' ' for j in range(25)])))
    axes[4].set_xlim([-0.03,0.03])
    axes[4].set_ylim([-0.015,0.015])
    axes[4].axvline(0,lw=0.5,color='.3',ls='--')
    axes[4].axhline(0,lw=0.5,color='.3',ls='--')
    axes[4].axhline( 0.0001,lw=0.5,color='r',ls='--') # Plot the "effective 0"
    axes[4].axhline(-0.0001,lw=0.5,color='r',ls='--') # Plot the "effective 0"
    axes[4].plot([-0.02, 0.02], [-0.02, 0.02],lw=0.5,ls="--", c=".3")
    axes[4].plot([-0.02, 0.02], [ 0.02,-0.02],lw=0.5,ls="--", c=".3")
plt.legend()
plt.tight_layout()
plt.show()
```
