```python
# The assignment (index match) between the sph and structs arrays are uniform random
# Each of the 100 structures has the same probability of containing a sphere of any value
colors = ["red","blue","green","yellow","purple","orange","brown","pink","grey","white"]
plt.figure(figsize=(10,10))
nsids = 3208.
nsphs = 42103.
asph,asid = None,None
for j in range(10):
    # Define 1000 random normal sphere values
    sph = np.random.randn(42103)
    # Define 1000 random uniform integers in the range 0-100
    structs = np.random.randint(0,3208,42103)
    try:
        asph = np.concatenate(asph,sph)
        asid = np.concatenate(asid,structs)
    except:
        asph = sph
        asid = structs
    # Define the dataframe
    df = pd.DataFrame(np.array([sph,structs]).T,columns=["dpi","structid"])
    nstructs = [len(df.iloc[:i,:].drop_duplicates("structid")) for i in range(len(df),0,-1)]
    plt.scatter(range(1,len(df)),nstructs,color=colors[j])
    plt.xlabel("Sphere Percentile")
    plt.ylabel("Percentage of Structures")
    plt.xlim([0,len(df)])
```
