```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr,pearsonr
from pi_util import *
%matplotlib inline
```

```python
df = load_results("all_pi.txt")
df = collapse(df)
```

```python
print "Spearman correlation CMAF1-PI1:",spearmanr(df["cmaf1"].values,df["pi1"].values)
print "Spearman correlation CMAF2-PI2:",spearmanr(df["cmaf2"].values,df["pi2"].values)
print "Spearman correlation dMAF-dPI: ",spearmanr(df["cmaf1"].values-df["cmaf2"].values,df["dpi"].values)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df["cmaf1"],df["pi1"])
plt.show()
```
