```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("rare/pi_rare.txt",header=0,sep='\t')
print "Total spheres examined:\t\t%d"%len(df)
# Identify spheres with the same set of SNPs
ndup = (df.duplicated(["#structid","biounit","nbrsnps1","nbrsnps2"])).values.sum()
print "Number of SNP-set duplicates:\t%d"%ndup
print "Number of unique SNP sets:\t%d"%(len(df)-ndup)
# Are there spheres where neighbor SNPs listed as NaN? If so, make sure their SNP count is 0
nansnp1 = (df["nbrsnps1"].astype(str)=="nan").values 
nansnp2 = (df["nbrsnps2"].astype(str)=="nan").values
df.ix[nansnp1,"snpcnt1"] = 0
df.ix[nansnp2,"snpcnt2"] = 0
# Filter any spheres with fewer than 2 SNPs
df = df[(df["snpcnt1"]>1) | (df["snpcnt2"]>1)]
print "Total spheres after requiring 2+ SNPs: %d"%len(df)
# Filter any spheres with allele counts less than 10
df = df[(df["ac1"]>0) | (df["ac2"]>0)]
print "Total spheres after requiring 0+ alleles per sphere: %d"%len(df)
# Filter any spheres with a value of 0
df["absdpi"] = abs(df["dpi"])
df = df[df["absdpi"]>0]
print "Total spheres after removing non-differentiated: %d"%len(df)
# Select the sphere with the minimum value of deltaPi
df = df.sort("absdpi",ascending=True).replace(np.nan,-999).groupby(["#structid","biounit","nbrsnps1","nbrsnps2"],as_index=False).nth(0)
# Select the sphere with the median value of deltaPi
# df = df.groupby(["#structid","biounit","nbrsnps1","nbrsnps2"]).apply(lambda x: x[x["dpi"]==x["dpi"].median()])
# df = df.replace(-999,np.nan)
print "Total spheres after collapse:\t%d"%len(df)
# Identify spheres where deltaPi is NaN. These shouldn't exist.
df = df[~np.isnan(df["pi1"])]
df = df[~np.isnan(df["pi2"])]
print "Total spheres after filtering NaN:\t%d"%len(df)
```

```python
# Check the FET p-value distribution for uniformality before subsetting the data
# First, note the proportion of spheres with empty expectation cells
num_p1  = (df["fetp"]==1).sum()
num_nan = ((df["fetp"]==1) & (np.isnan(df["fet_or"]))).sum()
num_inf = ((df["fetp"]==1) & (np.isinf(df["fet_or"]))).sum()
print "Proportion of spheres with p=1:\t\t\t%d/%d (%.2f%%)"%(num_p1,len(df),num_p1*100./len(df))
print "Proportion of p=1 spheres with NaN OR:\t\t%d/%d (%.2f%%)"%(num_nan,num_p1,num_nan*100./num_p1)
print "Proportion of p=1 spheres with infinite OR:\t%d/%d (%.2f%%)"%(num_inf,num_p1,num_inf*100./num_p1)
tdf = df[~np.isnan(df["fet_or"])]
print '--'
sig_fet = (tdf["fetp"]<0.01).sum()
print "Proportion of testable spheres with p<0.01: %d/%d (%.2f%%)"%(sig_fet,len(tdf),sig_fet*100./len(tdf))
# What are my Bonferroni thresholds for p<0.05 and p<0.01?
bft5 = 0.05 / len(tdf)
bft1 = 0.01 / len(tdf)
print "Bonferroni threshold for p<0.05 and %d tests: %.1e"%(len(tdf),bft5)
print "Bonferroni threshold for p<0.01 and %d tests: %.1e"%(len(tdf),bft1)
sig_fet = (tdf["fetp"]<bft1).sum()
print "Proportion of testable spheres with BFp<0.01 (%.1e): %d/%d (%.2f%%)"%(bft1,sig_fet,len(tdf),sig_fet*100./len(tdf))
plt.figure(figsize=(20,7))
plt.scatter(sorted(-np.log10(np.random.random(len(tdf)))),sorted(-np.log10(tdf["fetp"])),c="blue")
plt.axhline(-np.log10(0.01),color="blue",label="p<0.01")
plt.axhline(-np.log10(bft1),color="red",label="p<%.1e"%bft1)
plt.legend(loc="lower right")
plt.xlabel("Expected P-Value")
plt.ylabel("Observed P-Value")
plt.plot([0,6],[0,6])
plt.show()
plt.figure(figsize=(20,7))
plt.hist(tdf.ix[(tdf["fetp"]>0.01) & (~np.isinf(tdf["fet_or"])),"fetp"].values,bins=150)
plt.title("Histogram of Nominally Insignificant Spheres")
plt.show()
# Does the p-value histogram make sense given that ~16% of our data has invalid parameters for the test and had to be removed?
```

```python
# Identify spheres with the same set of SNPs
ndup = (df.duplicated(["#structid","biounit","nbrsnps1","nbrsnps2"])).values.sum()
print "Number of SNP-set duplicates:\t%d"%ndup
print "Number of unique SNP sets:\t%d"%(len(df)-ndup)
# Select only the sphere with the most extreme value of deltaPi
fig,ax = plt.subplots(1,1,figsize=(20,7))
ax.set_xlabel("Total Spheres per Structure")
plt.hist(df.groupby(["#structid","biounit"]).size(),color="blue",label='all',bins=150,alpha=0.3)
df["absdpi"] = abs(df["dpi"])
df = df.sort("absdpi",ascending=False).replace(np.nan,-999).groupby(["#structid","biounit","nbrsnps1","nbrsnps2"],as_index=False).nth(0)
df = df.replace(-999,np.nan)
ax2 = ax.twiny()
ax2.hist(df.groupby(["#structid","biounit"]).size(),color="red",label='representative',bins=150,alpha=0.3)
ax2.set_xlim([0,350])
ax2.set_xlabel("Representative Spheres per Structure")
plt.legend(loc="upper right")
print "Total spheres after collapse:\t%d"%len(df)
# Remove spheres where snpcount<1 but Pi has a value; these seem to be caused by 1000 Genomes genotype formatting issue
l = len(df)
df = df[~((df["snpcnt1"]<1) & (df["pi1"]!=0))]
df = df[~((df["snpcnt2"]<1) & (df["pi2"]!=0))]
print "%d SNPs removed because Pi was calculated in the absence of variation"%(l-len(df))
# Remove spheres where deltaPi is NaN. These shouldn't exist.
l = len(df)
df = df[~np.isnan(df["pi1"])]
df = df[~np.isnan(df["pi2"])]
print "%d SNPs removed because Pi1 or Pi2 was NaN"%(l-len(df))
print "Total spheres after filtering:\t%d"%len(df)
# Are there spheres where neighbor SNPs listed as NaN? If so, make sure their SNP count is 0
nansnp1 = (df["nbrsnps1"].astype(str)=="nan").values
nansnp2 = (df["nbrsnps2"].astype(str)=="nan").values
df.ix[nansnp1,"snpcnt1"] = 0
df.ix[nansnp2,"snpcnt2"] = 0
# Also replace the NaN with an empty string
df.ix[nansnp1,"nbrsnps1"] = ''
df.ix[nansnp2,"nbrsnps2"] = ''
# And finally, convert the dtype to string
df[["nbrsnps1","nbrsnps2"]] = df[["nbrsnps1","nbrsnps2"]].astype(str)
print "Total spheres for Pi analysis:\t%d"%len(df)
```

```python
from scipy.stats import fisher_exact
# Don't need to test the outcome of an empty sphere, b/c they are not tested
print "2 empty outside, 1 empty inside",fisher_exact([[2,0],[0,0]])
print "2 empty outside, 0 empty inside",fisher_exact([[2,2],[0,0]])
print "0 empty outside, 0 empty inside",fisher_exact([[2,1],[1,1]])
print "0 empty outside, 1 empty inside",fisher_exact([[2,0],[1,1]])
print "0 empty outside, 1 empty inside",fisher_exact([[2,0],[2,2]])
print "0 empty outside, 1 empty inside",fisher_exact([[2,0],[100,100]])
# So we get NaN values when there is no expected odds ratio
# And we get inf values when one of the observed cells is 0
# But we still get a p-value for the infinite values, determined by the magnitude of the other cell
# And those p-values are very similar to what you get if the 0-cell has been a 1
plt.scatter(range(50),[fisher_exact([[i,0],[50,150]])[1] for i in range(50)],color="blue")
plt.scatter(range(50),[fisher_exact([[i,1],[50,150]])[1] for i in range(50)],color="red")
plt.show()
# So we will need to find another way of quantifying the "significance" of 
# spheres without external variation in the structure, but everything else
# should be fine to look at as-is.
```

```python
tdf = df[~np.isnan(df["fet_or"])]
# What are my Bonferroni thresholds for p<0.05 and p<0.01?
bft5 = 0.05 / len(tdf)
bft1 = 0.01 / len(tdf)
print "Bonferroni threshold for p<0.05 and %d tests: %.1e"%(len(tdf),bft5)
print "Bonferroni threshold for p<0.01 and %d tests: %.1e"%(len(tdf),bft1)

# Do my p-value follow the uniform expectation?
# Should they? We've filtered out empty spheres, and all spheres with no expectation
# We've also filtered down to representative spheres, and not the entire set that was tested
# Perhaps we should look at the entire sphere set when evaluating the p-value distribution
# Moving this to the top
plt.figure(figsize=(20,7))
plt.scatter(sorted(-np.log10(np.random.random(len(tdf)))),sorted(-np.log10(tdf["fetp"])),c="blue")
plt.axhline(-np.log10(bft5),color="blue")
plt.axhline(-np.log10(bft1),color="red")
plt.title("All deltaPi values")
plt.xlabel("Expected P-Value")
plt.ylabel("Observed P-Value")
plt.plot([0,6],[0,6])

# Hell no they don't...
# What if we just look at the extreme-valued deltaPi spheres?
tdf = extDPI[~np.isnan(extDPI["fet_or"])]
plt.figure(figsize=(20,7))
plt.scatter(sorted(-np.log10(np.random.random(len(tdf)))),sorted(-np.log10(tdf["fetp"])),c="blue")
plt.axhline(-np.log10(bft5),color="blue")
plt.axhline(-np.log10(bft1),color="red")
plt.title("Extreme top and bottom 5% of deltaPi values")
plt.xlabel("Expected P-Value")
plt.ylabel("Observed P-Value")
plt.plot([0,6],[0,6])
plt.show()
```

```python
# The Fisher Exact Test is identifying regions with extreme deviations from expectation *within* the structure
# If we see that BF-significant FET spheres are skewed towards extreme deltaPi values, then it implies that 
# regions that deviate from expectation do so drastically. If not, it implies that deviations from expectation
# are modest, and would not be detected by global deltaPi thresholds, and only by comparison within a structure.
plt.figure(figsize=(20,7))
plt.hist(df.ix[df["fetp"]<bft1,"dpi"].values,bins=150)
plt.title("Distribution of deltaPi for FET Bonferroni-Significant Spheres")
plt.xlim([-.015,.015])
plt.show()
```

```python
# I want to know if Pi values are distributed evenly over all structures, 
# or if some structures are enriched or depleted for these values
# I want to consider the entire range of values, so I want to test if the distribution
# of Pi(EUR), Pi(AFR), and deltaPi for each structure is the same as the global distribution
# To do this, I will use an ANOVA, but I'm not sure that I've met all the assumptions, so
# I will also use Kruskal's
from scipy.stats import kruskal,f_oneway
# ANOVA - Pi(EUR)
plt.figure(figsize=(20,7))
anovasEUR = pd.DataFrame([f_oneway(df.ix[(df["#structid"]==sid) & (df["biounit"]==bio),"pi1"],
                          df.ix[~((df["#structid"]==sid) & (df["biounit"]==bio)),"pi1"])
                       for _,(sid,bio) in df[["#structid","biounit"]].drop_duplicates().iterrows()],columns=["F","pval"])
plt.hist(anovasEUR["pval"].values,bins=150,color="blue",label="Pi(EUR)",alpha=0.33)
# ANOVA - Pi(AFR)
anovasAFR = pd.DataFrame([f_oneway(df.ix[(df["#structid"]==sid) & (df["biounit"]==bio),"pi2"],
                                   df.ix[~((df["#structid"]==sid) & (df["biounit"]==bio)),"pi2"])
                       for _,(sid,bio) in df[["#structid","biounit"]].drop_duplicates().iterrows()],columns=["F","pval"])
plt.hist(anovasAFR["pval"].values,bins=150,color="red",label="Pi(AFR)",alpha=0.33)
# ANOVA - deltaPi
anovasDPI = pd.DataFrame([f_oneway(df.ix[(df["#structid"]==sid) & (df["biounit"]==bio),"dpi"],
                                   df.ix[~((df["#structid"]==sid) & (df["biounit"]==bio)),"dpi"])
                       for _,(sid,bio) in df[["#structid","biounit"]].drop_duplicates().iterrows()],columns=["F","pval"])
plt.hist(anovasDPI["pval"].values,bins=150,color="green",label="deltaPi",alpha=0.33)
plt.legend()
plt.show()
# The raw Pi values have some really strange artifacts with p between ~0.5 and 0.8
# The deltaPi values look better, with only a slight uptick near 1.0, but otherwise a roughly uniform distribution
# with an enrichment near 0
```
