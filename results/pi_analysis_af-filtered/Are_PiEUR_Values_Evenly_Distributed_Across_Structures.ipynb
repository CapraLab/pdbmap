```python
import numpy as np
import pandas as pd
from scipy.stats import kruskal
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 18})
import seaborn as sns
from copy import deepcopy
from scipy.stats import ks_2samp
%matplotlib inline
```

## Read, Reduce, and QC the Rare-SNP-Only Dataset

```python
df = pd.read_csv("rare/pi_rare.txt",header=0,sep='\t')
print "Total spheres examined:\t%d"%len(df)
print "Total biological assemblies examined:\t%d"%len(df[["#structid","biounit"]].drop_duplicates())
```

```python
# Identify spheres with the same set of SNPs
ndup = (df.duplicated(["#structid","biounit","nbrsnps1","nbrsnps2"])).values.sum()
print "Number of SNP-set duplicates:\t%d"%ndup
print "Number of unique SNP sets:\t%d"%(len(df)-ndup)
# Select only the sphere with the most extreme value of deltaPi
fig,ax = plt.subplots(1,1,figsize=(20,7))
ax.set_xlabel("Total Spheres per Structure")
plt.hist(df.groupby(["#structid","biounit"]).size(),color="blue",label='all',bins=150,alpha=0.3)
df["abspi"] = abs(df["pi1"])
df = df.sort("abspi",ascending=False).replace(np.nan,-999).groupby(["#structid","biounit","nbrsnps1","nbrsnps2"],as_index=False).nth(0)
df = df.replace(-999,np.nan)
ax2 = ax.twiny()
ax2.hist(df.groupby(["#structid","biounit"]).size(),color="red",label='representative',bins=150,alpha=0.3)
ax2.set_xlim([0,350])
ax2.set_xlabel("Representative Spheres per Structure")
plt.legend(loc="upper right")
print "Total spheres after collapse:\t%d"%len(df)
# Remove spheres where snpcount<1 but Pi has a value; these seem to be caused by 1000 Genomes genotype formatting issue
l = len(df)
df = df[~((df["snpcnt1"]<1) & (df["pi1"]!=0))]
print "%d SNPs removed because Pi was calculated in the absence of variation"%(l-len(df))
# Remove spheres where deltaPi is NaN. These shouldn't exist.
l = len(df)
df = df[~np.isnan(df["pi1"])]
print "%d SNPs removed because Pi1 was NaN"%(l-len(df))
print "Total spheres after filtering:\t%d"%len(df)
# Are there spheres where neighbor SNPs listed as NaN? If so, make sure their SNP count is 0
nansnp1 = (df["nbrsnps1"].astype(str)=="nan").values
df.ix[nansnp1,"snpcnt1"] = 0
# Also replace the NaN with an empty string
df.ix[nansnp1,"nbrsnps1"] = ''
# And finally, convert the dtype to string
df[["nbrsnps1","nbrsnps2"]] = df[["nbrsnps1","nbrsnps2"]].astype(str)
print "Total spheres for Pi analysis:\t%d"%len(df)
maxsize = df.groupby(["#structid","biounit"]).size().max()
maxdf   = df.groupby(["#structid","biounit"]).filter(lambda x: len(x)>=maxsize)
```

## Do the structures with significantly high Pi(EUR) distributions or spheres with extreme values of Pi(EUR) overlap with the genes identified by Tennessen et al.?

```python
# Kruskal-Wallis - Pi(EUR)
kruskalPi = pd.DataFrame([kruskal(df.ix[(df["#structid"]==sid) & (df["biounit"]==bio),"pi1"],
                          df.ix[~((df["#structid"]==sid) & (df["biounit"]==bio)),"pi1"])
                       for _,(sid,bio) in df[["#structid","biounit"]].drop_duplicates().iterrows()],columns=["H","p"])
# Test against Tennessen
bft1 = 0.01 / len(df.drop_duplicates(["#structid","biounit"]))
kdf  = df[["#structid","biounit"]].drop_duplicates()[(kruskalPi["p"]<bft1).values]
p95 = np.percentile(df["pi1"].values,95)
pdf  = df[df["pi1"]>p95]
# Do these overlap with Tennessen?
tennessen = [sid.strip() for sid in open("/Users/mikesivley/sivleyrm/pdbmap/temp/pi_analysis/Tennessen2012_SolvedPDB.txt")]
# How many of these structures were used as representatives (were they included in the analysis?)
# How many of the significant structures overlapped the Tennesseen dataset?
sigcnt = len(kdf.drop_duplicates(["#structid","biounit"]))
extcnt = len(pdf.drop_duplicates(["#structid","biounit"]))
print "Evaluation of %4d structures with >0 non-redundant spheres:"%len(kruskalPi)
print "Structures passing Bonferroni-corrected p<0.01 (p<%.1e): %3d (%.2f%%)"%(bft1,sigcnt,sigcnt*100./len(kruskalPi))
print "Structures with extreme 5%% values of Pi(EUR):  %3d (%.2f%%)"%(extcnt,extcnt*100./len(kruskalPi))
tencnt = len(df[df["#structid"].isin(tennessen)].drop_duplicates("#structid"))
print "All-structure overlap with Tennessen genes: %3d of %3d"%(tencnt,len(tennessen))
print "Sig-structure overlap with Tennessen genes: %3d of %3d"%(len(kdf[kdf["#structid"].isin(tennessen)].drop_duplicates("#structid")),tencnt)
print "Ext-structure overlap with Tennessen genes: %3d of %3d"%(len(pdf[pdf["#structid"].isin(tennessen)].drop_duplicates("#structid")),tencnt)
pdf[pdf["#structid"].isin(tennessen)].drop_duplicates("#structid")
```

## How many structures contain spheres in the extreme 5% of the deltaPi distribution?

```python
extPi = df[(df['pi1']>p95)]
rarestructs = extPi[["#structid","biounit"]].drop_duplicates().values
allcnt = len(df[["#structid","biounit"]].drop_duplicates())
extcnt = len(extPi[["#structid","biounit"]].drop_duplicates())
print "%d of %d (%.2f%%) structures contain spheres in the extreme 5%%"%(extcnt,allcnt,extcnt*100./allcnt)
```

```python
# Structure discovery is defined as the number of structures observed in the extreme Xth sphere percentile
plt.figure(figsize=(20,7))

# Plot structure discovery as a function of random sphere selection
idx = np.random.permutation(df.index)
nstructs = [len(df.loc[idx[:i],:].drop_duplicates(["#structid","biounit"])) for i in range(0,len(idx),len(idx)/800)[:800]]
x = deepcopy(nstructs)
plt.scatter(np.array(range(800))/8.,nstructs,color="darkblue",label="random",alpha=0.3)
plt.annotate("Random 5%%: %d (%.2f%%)"%(nstructs[20],nstructs[20]*100./nstructs[-1]),(5,nstructs[18]))

# Plot structure discovery as a function deltaPi percentile
upcntls = [np.percentile(df["pi1"],i/8.) for i in range(800,0,-1)]
nstructs = [len(df[(df["pi1"]>=upcntls[i])].drop_duplicates(["#structid","biounit"])) for i in range(800)]
y = deepcopy(nstructs)
plt.scatter(np.array(range(800))/8.,nstructs,color="darkred",label="deltaPi-ordered",alpha=0.3)
plt.annotate("5th Percentile: %d (%.2f%%)"%(nstructs[20],nstructs[20]*100./nstructs[-1]),(5,nstructs[18]))
plt.legend(loc="lower right")
plt.title("Structure Discovery as a Function of Pi(EUR) One-Sided Percentile Threshold")
plt.xlabel("European Nucleotide Diversity Percentile")
plt.ylabel("Number of Structures")
plt.xticks(range(100,0,-5),range(100,0,-5))
plt.xlim([0,100])
plt.ylim([0,3500])

# 5th percentile with respect to the entire distribution (below 2.5th and above 97.5th)
plt.axvline(5,color='darkred')

# Test the distributions for equality
D,p = ks_2samp(x,y)
print "Two-sided 2-Sample KS Test: D=%.3f; p<%.1e"%(D,p)
```
